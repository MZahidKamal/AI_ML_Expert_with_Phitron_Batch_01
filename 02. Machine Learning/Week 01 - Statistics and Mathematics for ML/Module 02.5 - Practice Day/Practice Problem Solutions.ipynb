```python
import numpy as np
from scipy import stats



# Original Data list.
marks = [50, 60, 65, 70, 75, 80, 85, 90]
# print(type(marks))


# Now we'll convert the original data list into a Numpy Array.
marks = np.array(marks)
print(f"Original Marks: {marks}")


# Now we'll find the mean, median and mode of this Numpy Array.
mean_val = np.mean(marks)
median_val = np.median(marks)
mode_result = stats.mode(marks, keepdims=True)
mode_val = mode_result.mode[0]
original_skew_value = stats.skew(marks)


print(f"Mean value of Marks: {mean_val}")
print(f"Median value of Marks: {median_val}")
print(f"Mode value of Marks: {mode_val}")
print(f"Skew value of Marks: {original_skew_value}")


# Now we'll find the index of 90 in the Numpy Array, and then will replace 90 by 900.
# If the occurrence of 90 is more than once, then we'll replace all the occurrences of 90 with 900.
marks[np.where(marks==90)] = 900
print(f"--Then again--\nMarks after replacing 90 with 900: {marks}")


# Now again we'll find the mean, median and mode of this Numpy Array.
mean_val = np.mean(marks)
median_val = np.median(marks)
mode_result = stats.mode(marks, keepdims=True)
mode_val = mode_result.mode[0]
outlier_skew_value = stats.skew(marks)


print(f"Mean value of Marks: {mean_val}")
print(f"Median value of Marks: {median_val}")
print(f"Mode value of Marks: {mode_val}")
print(f"Skew value of Marks with outlier: {outlier_skew_value}")
```

**Effect or outlier:**
> In the original Marks array, there were no outliers. So the values of Mean, Median, and Mode were all normal. But after adding an outlier, the Mean value changed a lot — it shifted towards the outlier. This is because Mean is very sensitive to outliers; whenever an outlier is present, the Mean value changes in an abnormal way. And there was no change in the Mode value at all, since no number's repeating frequency changed.

> The number 900 is positioned on the right side of the data distribution. So if a graph is drawn, its "tail" will get pulled and stretched towards the right. This is called Positive Skewness. Therefore, it can be said that Positive Skewness has been created due to the outlier.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Creating Numpy Array from the original data array.
temps = np.array([29, 31, 33, 33, 32, 31, 30])

# Now we'll find the variance and standard deviation of this Numpy Array.'
variance_val = np.var(temps)
std_dev_val = np.std(temps)

# Now we'll print the variance and standard deviation.
print(f"Variance: {variance_val:.2f}")
print(f"Standard Deviation: {std_dev_val:.2f}")

# 2. Visualization
sns.histplot(temps, kde=True, bins=10)
plt.axvline(np.mean(temps), color='red', linestyle='--', label=f'Mean: {np.mean(temps):.2f}')

plt.title('Temperature Distribution')
plt.xlabel('Temperature (°C)')

plt.legend()
plt.show()
```

### Part B – Percentiles, IQR and Z-Score
#### Problem 03

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# --- Task 3: Cat Weights Analysis ---
weights = np.array([2.5, 3.0, 3.2, 3.3, 3.4, 3.5, 3.6, 3.9, 4.0, 4.5])

# Calculating Percentiles and IQR
p25 = np.percentile(weights, 25)
p50 = np.percentile(weights, 50) # Median
p75 = np.percentile(weights, 75)
iqr = p75 - p25

# Finding Outliers using 1.5*IQR rule
lower_bound = p25 - (1.5 * iqr)
upper_bound = p75 + (1.5 * iqr)
outliers = weights[(weights < lower_bound) | (weights > upper_bound)]

print(f"Task 3: P25={p25}, P50={p50}, P75={p75}, IQR={iqr}")
print(f"Outliers: {outliers if len(outliers)>0 else 'None'}")

# Boxplot Visualization for Outliers
sns.boxplot(x=weights)

plt.title('Cat Weights Boxplot (Checking Outliers)')
plt.xlabel('Weight (kg)')

plt.show()
```

### Part B – Percentiles, IQR and Z-Score
#### Problem 04

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats


# --- Task 4: Z-Score Calculations ---
mu = 50
sigma = 10
x_val = 65

# Z for x = 65
z_score = (x_val - mu) / sigma

# x for Z = -2.5
z_given = -2.5
x_calculated = mu + (z_given * sigma)

print(f"Task 4: Z-score for x=65 is {z_score}")
print(f"Value x for Z=-2.5 is {x_calculated}")

# Normal Distribution Visualization for Z-score
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
y = stats.norm.pdf(x, mu, sigma)
plt.plot(x, y, label='Normal Distribution')
plt.axvline(x_val, color='red', linestyle='--', label=f'x=65 (Z={z_score})')
plt.fill_between(x, y, where=(x >= x_val), color='red', alpha=0.3)

plt.title('Z-score Interpretation (x=65)')

plt.legend()
plt.show()
```

## **সার্ভে ডেটা এবং প্রোবাবিলিটি বিশ্লেষণ**

মোট জনসংখ্যা (Sample Space), $n(S) = ১০০$

### **১. মৌলিক প্রোবাবিলিটি বের করার নিয়ম:**

* **P(A) [Pizza পছন্দ করার সম্ভাবনা]:** $$P(A) = \frac{n(A)}{n(S)} = \frac{৪০}{১০০} = 0.4$$
* **P(B) [Burgers পছন্দ করার সম্ভাবনা]:** $$P(B) = \frac{n(B)}{n(S)} = \frac{৫০}{১০০} = 0.5$$
* **P(A ∩ B) [উভয়টি পছন্দ করার সম্ভাবনা]:** $$P(A \cap B) = \frac{n(A \cap B)}{n(S)} = \frac{২০}{১০০} = 0.2$$

### **২. কন্ডিশনাল প্রোবাবিলিটি (Conditional Probability):**

* **P(A | B):** এর অর্থ হলো— যারা বার্গার পছন্দ করে, তাদের মধ্যে পিৎজা পছন্দ করার সম্ভাবনা কত?
    $$P(A | B) = \frac{P(A \cap B)}{P(B)} = \frac{0.2}{0.5} = 0.4$$
    *(অথবা সরাসরি সংখ্যা দিয়ে: $\frac{২০}{৫০} = 0.4$)*

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

# Data from the survey
total_people = 100
like_pizza = 40
like_burger = 50
like_both = 20

# Probability Calculations
p_a = like_pizza / total_people
p_b = like_burger / total_people
p_a_and_b = like_both / total_people
p_a_given_b = p_a_and_b / p_b

print(f"P(A): {p_a}")
print(f"P(B): {p_b}")
print(f"P(A ∩ B): {p_a_and_b}")
print(f"P(A | B): {p_a_given_b}")



# Calculated values for Venn Diagram
only_pizza = like_pizza - like_both
only_burger = like_burger - like_both
neither = total_people - (only_pizza + only_burger + like_both)

# Visualization
plt.figure(figsize=(8, 6))
venn = venn2(
	subsets=(only_pizza, only_burger, like_both),
	set_labels=('Pizza (A)', 'Burgers (B)')
)

plt.title(f"Food Preference Survey (N={total_people})")
plt.annotate(f'Neither: {neither}', xy=(-0.6, 0.5))
plt.show()
```

### **গুরুত্বপূর্ণ Keyword ও ব্যাখ্যা:**

* **Intersection ($A \cap B$):** এটি 'And' লজিক কাজ করে, অর্থাৎ যারা একই সাথে দুটি ক্যাটাগরিতেই পড়ে।
* **Conditional Probability ($P(A|B)$):** এটি আমাদের স্যাম্পল স্পেসকে ছোট করে দেয়। এখানে আমরা পুরো ১০০ জনের বদলে শুধু ওই ৫০ জনকে দেখছি যারা বার্গার পছন্দ করে।
* **Independence Check:** লক্ষ্য করুন, এখানে $P(A|B) = 0.4$ এবং $P(A) = 0.4$। যেহেতু এই দুটি সমান, তাই আমরা বলতে পারি **Pizza** এবং **Burger** পছন্দ করার ঘটনা দুটি একে অপরের থেকে **Independent** (স্বাধীন)।

## **Bayes' Theorem: Spam Filter Analysis**

### **১. প্রদত্ত তথ্য (Given Data):**
* মোট ইমেল (Total) = ১০০০
* স্প্যাম ইমেল, **P(Spam)** = ১০০ / ১০০০ = **০.১**
* নন-স্প্যাম ইমেল, **P(notSpam)** = ৯০০ / ১০০০ = **০.৯**
* স্প্যাম ইমেলে "free" আছে, **P(Free | Spam)** = ৪০ / ১০০ = **০.৪**
* নন-স্প্যাম ইমেলে "free" আছে, **P(Free | notSpam)** = ৬০ / ৯০০ ≈ **০.০৬৬**

### **২. Bayes' Theorem ফর্মুলা:**
$$P(Spam | Free) = \frac{P(Free | Spam) \times P(Spam)}{P(Free)}$$

এখানে, $P(Free)$ হলো যেকোনো ইমেলে "free" থাকার মোট সম্ভাবনা:
$$P(Free) = [P(Free | Spam) \times P(Spam)] + [P(Free | notSpam) \times P(notSpam)]$$
$$P(Free) = (০.৪ \times ০.১) + (০.০৬৬ \times ০.৯) = ০.০৪ + ০.০৬ = ০.১$$

### **৩. চূড়ান্ত হিসাব:**
$$P(Spam | Free) = \frac{০.০৪}{০.১} = 0.4$$
**ফলাফল:** অর্থাৎ, কোনো ইমেলে "free" শব্দটি থাকলে সেটি স্প্যাম হওয়ার সম্ভাবনা **৪০%**।

## **স্প্যাম ফিল্টারের জন্য এর গুরুত্ব (Interpretation)**

* **Probability Update:** শুরুতে আমরা জানতাম মাত্র ১০% ইমেল স্প্যাম। কিন্তু যখনই ইমেলে "free" শব্দটি পাওয়া গেল, Bayes' Theorem ব্যবহার করে আমাদের বিশ্বাস (Belief) আপডেট হলো এবং স্প্যাম হওয়ার সম্ভাবনা ১০% থেকে বেড়ে **৪০%** হয়ে গেল।
* **Decision Making:** একটি স্প্যাম ফিল্টার সাধারণত একটি নির্দিষ্ট **Threshold** (যেমন: ৫০% বা ৮০%) ব্যবহার করে। যেহেতু এখানে সম্ভাবনা ৪০%, ফিল্টারটি হয়তো শুধু "free" শব্দটির ওপর ভিত্তি করে এটিকে সরাসরি স্প্যাম বক্সে পাঠাবে না, বরং আরও কিছু কিওয়ার্ড চেক করবে।
* **False Positives:** এখানে দেখা যাচ্ছে নন-স্প্যাম ইমেলেও "free" শব্দটি (৬০ বার) আছে। Bayes' Theorem ফিল্টারকে সাহায্য করে যেন গুরুত্বপূর্ণ ইমেল ভুল করে স্প্যামে না চলে যায়।

```python
# Bayes' Theorem Implementation
p_spam = 100 / 1000
p_not_spam = 900 / 1000
p_free_given_spam = 40 / 100
p_free_given_not_spam = 60 / 900

# Total probability of 'free'
p_free = (p_free_given_spam * p_spam) + (p_free_given_not_spam * p_not_spam)

# P(Spam | Free)
p_spam_given_free = (p_free_given_spam * p_spam) / p_free

print(f"Probability that an email is Spam, given it contains 'free': {float(p_spam_given_free * 100):.0f}%")
```

### Part B – Conditional Probability and Bayes
#### Problem 03
---

### **১. প্রদত্ত তথ্য (Given Data):**
* **Prevalence (Base Rate):** $P(Disease) = \frac{1}{10,000} = 0.0001$
* **Sensitivity:** $P(+ \mid Disease) = 0.99$
* **Specificity:** $P(- \mid \neg Disease) = 0.99$
* **False Positive Rate:** $P(+ \mid \neg Disease) = 1 - 0.99 = 0.01$


### **২. $P(Disease \mid +)$ ক্যালকুলেশন:**
আমাদের বের করতে হবে পজিটিভ রেজাল্ট আসার পর ব্যক্তিটির আসলেই রোগ থাকার সম্ভাবনা কত।

**মোট পজিটিভ হওয়ার সম্ভাবনা, $P(+)$:**
$$P(+) = [P(+ \mid Disease) \times P(Disease)] + [P(+ \mid \neg Disease) \times P(\neg Disease)]$$
$$P(+) = (0.99 \times 0.0001) + (0.01 \times 0.9999)$$
$$P(+) = 0.000099 + 0.009999 = 0.010098$$

**বেয়েজ থিওরেম অনুযায়ী $P(Disease \mid +)$:**
$$P(Disease \mid +) = \frac{P(+ \mid Disease) \times P(Disease)}{P(+)}$$
$$P(Disease \mid +) = \frac{0.000099}{0.010098} \approx 0.0098 \text{ which is } \mathbf{0.98\%}$$


### **৩. Base-rate Effect এর আলোচনা:**
ফলাফলটি খেয়াল করুন—টেস্টটি **৯৯% নির্ভুল** হওয়া সত্ত্বেও রেজাল্ট পজিটিভ আসার পর রোগ থাকার সম্ভাবনা **১% এরও কম**! একেই বলা হয় **Base-rate fallacy** বা **Base-rate effect**।

* **কেন এমন হলো?** যেহেতু সমাজে রোগটি অত্যন্ত বিরল (১০০০০ জনে মাত্র ১ জন), তাই সুস্থ মানুষের সংখ্যা অনেক বেশি। ফলে টেস্টের মাত্র ১% ভুল (False Positive) হলেও, সুস্থ মানুষের বিশাল সংখ্যার কারণে "ভুল পজিটিভ" রিপোর্ট সংখ্যায় "আসল পজিটিভ" রিপোর্টের চেয়ে অনেক বেশি হয়ে যায়।
* **Spam Filter-এর সাথে মিল:** আপনার আগের স্প্যাম ফিল্টার উদাহরণেও আমরা দেখেছি, বিরল ঘটনার ক্ষেত্রে শুধু একটি কিওয়ার্ড বা টেস্টের ওপর ভিত্তি করে সিদ্ধান্ত নিলে ভুলের সম্ভাবনা বেশি থাকে।

```python
p_disease = 1/10000
p_pos_given_disease = 0.99  # Sensitivity
p_neg_given_no_disease = 0.99 # Specificity
p_pos_given_no_disease = 1 - p_neg_given_no_disease # False Positive Rate

# Total P(+)
p_positive = (p_pos_given_disease * p_disease) + (p_pos_given_no_disease * (1 - p_disease))

# P(D | +)
p_disease_given_pos = (p_pos_given_disease * p_disease) / p_positive

print(f"Probability of disease given a positive test: {p_disease_given_pos:.4f}, which is {float(p_disease_given_pos * 100)}%")
```

### **১. রেজাল্ট সামারি, মেট্রিকস এবং মডেলের আচরণ (Model Behavior)**
* **Accuracy (৮৯%):** মডেলটি মোট কত শতাংশ ক্ষেত্রে সঠিক প্রেডিকশন দিয়েছে।
* **Precision/PPV (৬০%):** মডেল যখন কোনো কিছুকে "পজিটিভ" বলছে, তার ৬০% আসলে সঠিক।
* **Recall/Sensitivity (৮০%):** মোট পজিটিভ কেসগুলোর মধ্যে মডেল ৮০% খুঁজে বের করতে পেরেছে।
* **F1 Score (০.৬৮৫৭):** প্রিসিশন এবং রিকল-এর মধ্যে একটি ব্যালেন্স।
* **Specificity (৯০.৫%):** নেগেটিভ কেসগুলোকে সঠিকভাবে নেগেটিভ হিসেবে চিহ্নিত করার ক্ষমতা।
* **Prevalence (১৫%):** ডেটাসেটে পজিটিভ কেসের হার।

### **২. মডেলের আচরণ বিশ্লেষণ (Explanation):**
* **High Accuracy vs Low Precision:** খেয়াল করুন, মডেলের Accuracy অনেক বেশি (৮৯%) মনে হলেও Precision কিন্তু তুলনামূলক কম (৬০%)। এর কারণ হলো **Imbalanced Data**। যেহেতু ডেটাসেটে নেগেটিভ কেস অনেক বেশি (৮৫০টি), মডেলটি নেগেটিভ প্রেডিক্ট করে সহজেই ভালো একুরেসি পাচ্ছে।
* **Recall vs Precision:** মডেলটির Recall (৮০%) ভালো, অর্থাৎ এটি অধিকাংশ পজিটিভ কেস ধরতে পারে। তবে এটি করতে গিয়ে অনেক নেগেটিভ কেসকেও ভুল করে পজিটিভ (False Positive = ৮০) বলে দিচ্ছে, যা Precision কমিয়ে দিয়েছে।
* **F1 Score-এর গুরুত্ব:** এখানে একুরেসি না দেখে F1 Score দেখা বেশি যুক্তিযুক্ত, কারণ এটি আমাদের বলছে যে মডেলটি পজিটিভ ক্লাস প্রেডিক্ট করার ক্ষেত্রে মাঝারি মানের পারফর্ম করছে।
* **NPV (৯৬%):** মডেলটি যখন কোনো কিছুকে "নেগেটিভ" বলছে, তা সঠিক হওয়ার সম্ভাবনা অনেক বেশি।

### **গুরুত্বপূর্ণ Keyword:**
* **False Positive (FP):** মডেলটি ভুল করে ৮০ জনকে পজিটিভ বলেছে যারা আসলে নেগেটিভ ছিল।
* **False Negative (FN):** ৩০ জন পজিটিভ রোগীকে মডেলটি চিনতে পারেনি।

```python
# Confusion Matrix Data
tp = 120  # True Positive
fn = 30   # False Negative
fp = 80   # False Positive
tn = 770  # True Negative
total = tp + fn + fp + tn

# Calculations
accuracy = (tp + tn) / total
precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1_score = 2 * (precision * recall) / (precision + recall)
specificity = tn / (tn + fp)
npv = tn / (tn + fn)
prevalence = (tp + fn) / total

# Output results
metrics = {
    "Accuracy": accuracy,
    "Precision (PPV)": precision,
    "Recall (Sensitivity)": recall,
    "F1 Score": f1_score,
    "Specificity": specificity,
    "NPV": npv,
    "Prevalence": prevalence
}

for metric, value in metrics.items():
    print(f"{metric}: {value:.4f}")
```

#### **মেডিকেল ট্রায়াজ (Triage) কন্টেক্সটে এটি কেন গুরুত্বপূর্ণ?**
মেডিকেল ট্রায়াজের মূল উদ্দেশ্য হলো রোগীদের অবস্থার গুরুত্ব অনুযায়ী অগ্রাধিকার দেওয়া। এখানে কিছু চিন্তার বিষয় আছে:

* **Urgent (U) রোগীদের ঝুঁকি:** লক্ষ্য করলে দেখবে, Urgent ক্লাসের **Recall** মাত্র **৬৬.৭%**। এর মানে হলো, প্রতি ১০০ জন জরুরি রোগীর মধ্যে ৩৩ জনকেই মডেলটি সঠিকভাবে চিনতে পারছে না। মেডিকেল ট্রায়াজে এটি একটি বড় ভুল (False Negative), কারণ গুরুতর অসুস্থ রোগী যদি সঠিক সময়ে চিকিৎসা না পান, তবে তার জীবনহানি হতে পারে।
* **ভুল চিকিৎসার সম্ভাবনা (Precision):** Urgent ক্লাসের **Precision** মাত্র **৫৩.৩%**। অর্থাৎ, যাদের জরুরি চিকিৎসা দেওয়া হচ্ছে, তাদের প্রায় অর্ধেকই আসলে জরুরি রোগী নন। এর ফলে হাসপাতালের সীমিত রিসোর্স (যেমন: আইসিইউ বেড বা ডাক্তার) অপচয় হতে পারে।
* **সফলতা:** মডেলটি 'Normal' রোগীদের আলাদা করার ক্ষেত্রে দক্ষ হওয়ায় হাসপাতালের ভিড় কমাতে সাহায্য করবে, কারণ যারা সাধারণ রোগী তাদের দ্রুত চিহ্নিত করা যাচ্ছে।
* **মেডিকেল মডেলে Recall-কে গুরুত্ব দাও:** যখন কোনো রোগের ডায়াগনোসিস করবে, তখন Recall যেন বেশি থাকে সেদিকে নজর দেবে। কারণ রোগী সুস্থ মানুষকে 'অসুস্থ' বলা (False Positive) যতটা না ক্ষতিকর, একজন অসুস্থ মানুষকে 'সুস্থ' বলে বাড়ি পাঠিয়ে দেওয়া (False Negative) তার চেয়ে অনেক বেশি বিপজ্জনক।
* **F1 Score কেন দেখবে:** শুধু Accuracy দেখলে অনেক সময় আসল চিত্র বোঝা যায় না। যেমন তোমার মডেলে Overall Accuracy **৭৪.৭৫%** হলেও Urgent ক্লাসের অবস্থা কিন্তু বেশ নাজুক। তাই সবসময় ক্লাসভিত্তিক F1 Score চেক করবে।
* **Trade-off বোঝা:** Precision বাড়াতে গেলে অনেক সময় Recall কমে যায়। কাজের ধরন অনুযায়ী তোমাকে ঠিক করতে হবে কোনটি বেশি জরুরি—রিসোর্স বাঁচানো (Precision) নাকি জীবন বাঁচানো (Recall)।
