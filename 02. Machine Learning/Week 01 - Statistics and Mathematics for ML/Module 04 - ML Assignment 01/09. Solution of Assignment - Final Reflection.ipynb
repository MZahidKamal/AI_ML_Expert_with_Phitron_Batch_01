### Q6.4 — Short Answer [5 Marks]

**Question:** In 4 to 6 sentences, describe how the three modules connect in this assignment.
> Mention:
> * One idea from Module 1 or 2 that you used
> * One idea from Module 3 that you used
> * How these ideas together help you understand a dataset more deeply

---

**Answer:**

> From Module 1 and 2, I used descriptive statistics and classification metrics — computing the mean, standard deviation, and proportion of the positive class helped me understand the basic shape and balance of the dataset, while accuracy, precision, and recall revealed that the model was performing only slightly better than random guessing. From Module 3, I applied standardization and min-max scaling to the `monthly_income` column, which showed how the same data looks very different depending on the scaling method — standardization centers values around zero while min-max compresses them into a fixed range of 0 to 1. I also used one-hot encoding for `city_type` and ordinal encoding for `satisfaction_level`, which demonstrated that choosing the wrong encoding for a feature can introduce false relationships into the data. Finally, computing Euclidean and Manhattan distances on scaled vectors showed me that the scale of a feature directly affects how "close" or "far" two data points appear to be, which is why preprocessing is not just a technical step but a necessary foundation for any meaningful analysis or modeling.

> Module 1 এবং 2 থেকে আমি descriptive statistics এবং classification metrics ব্যবহার করেছি — mean, standard deviation, এবং positive class-এর অনুপাত বের করে dataset-এর মূল গঠন ও ভারসাম্য বুঝতে পেরেছি, আর accuracy, precision, এবং recall থেকে দেখা গেছে যে model-টি random guessing-এর চেয়ে সামান্যই ভালো কাজ করছে। Module 3 থেকে আমি `monthly_income` column-এ standardization এবং min-max scaling প্রয়োগ করেছি, যা দেখিয়েছে যে scaling পদ্ধতি ভিন্ন হলে একই data একদম আলাদা দেখায় — standardization মানগুলোকে শূন্যকে কেন্দ্র করে সাজায়, আর min-max সেগুলোকে ০ থেকে ১-এর মধ্যে সংকুচিত করে। এছাড়া `city_type`-এর জন্য one-hot encoding এবং `satisfaction_level`-এর জন্য ordinal encoding ব্যবহার করেছি, যা দেখিয়েছে যে কোনো feature-এর জন্য ভুল encoding বেছে নিলে data-তে মিথ্যা সম্পর্ক তৈরি হতে পারে। সবশেষে, scaled vector-এ Euclidean এবং Manhattan distance হিসাব করে বুঝলাম যে কোনো feature-এর scale সরাসরি প্রভাব ফেলে দুটি data point কতটা "কাছে" বা "দূরে" দেখাবে তার উপর — এই কারণেই preprocessing শুধু একটি কারিগরি ধাপ নয়, বরং যেকোনো অর্থবহ analysis বা modeling-এর জন্য এটি একটি অপরিহার্য ভিত্তি।
