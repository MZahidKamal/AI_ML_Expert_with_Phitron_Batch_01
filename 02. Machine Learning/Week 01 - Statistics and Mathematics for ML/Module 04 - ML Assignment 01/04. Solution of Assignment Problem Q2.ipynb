## Q2. Proportion of Positive Class

---

### 🔍 এই Question-এ কী চাওয়া হয়েছে?

`true_label` column-এ **1 = positive**, **0 = negative**।
মোট কতটি positive আছে এবং সেটি মোট sample-এর কত অংশ (proportion) — সেটি বের করতে হবে।
তারপর Q2.2-এ 1-2 বাক্যে বলতে হবে dataset balanced কিনা।

### 🎯 এখান থেকে কী অর্জন করব?

**Class imbalance** বোঝার দক্ষতা তৈরি হবে — যদি একটি class অনেক বেশি থাকে, model সেটার দিকে biased হয়ে যায়।

### 🧠 কীভাবে চিন্তা করতে হবে?

- Proportion ≈ **0.5** → dataset **balanced** — দুটো class প্রায় সমান
- Proportion **>> 0.5 বা << 0.5** → dataset **imbalanced** — একটি class বেশি
- এই dataset-এ: positive = **52**, negative = **48**, proportion = **0.52** → প্রায় balanced


### Q2.1 — Positive Class-এর Proportion হিসাব করা

```python
import pandas as pd
import numpy as np

# Now we'll load the dataset
df = pd.read_csv("02. Provided - module123_data.csv")
label_col = "true_label"
# Here we've loaded the data and set the label column name.

# Now we'll compute positive count, total, and proportion
positive_count    = (df[label_col] == 1).sum()
total_count       = df.shape[0]
positive_proportion = positive_count / total_count
# Here we've calculated: positive_count=52, total=100, proportion=0.52

print("Positive count:", positive_count)
print("Total samples:", total_count)
print("Proportion of positive class:", positive_proportion)
```

`(df[label_col] == 1)` → True/False array তৈরি করে, `.sum()` সব True গুলো গণনা করে।
`df.shape[0]` → মোট row সংখ্যা (100)।
`positive_count / total_count` → proportion = 52/100 = **0.52**।
