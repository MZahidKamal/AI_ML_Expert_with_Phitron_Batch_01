## Q5. One-Hot and Ordinal Encoding

---

### 🔍 এই Question-এ কী চাওয়া হয়েছে?

দুটো categorical column encode করতে হবে:

| Column | Type | Encoding |
|---|---|---|
| `city_type` | **Nominal** (Urban, Rural, Suburban) | **One-Hot Encoding** |
| `satisfaction_level` | **Ordinal** (Low < Medium < High) | **Ordinal Encoding** |

- **Q5.1** → `city_type`-এ One-Hot Encoding করা
- **Q5.2** → সেই encoded columns-কে main df-এ attach করা
- **Q5.3** → `satisfaction_level`-এ Ordinal Encoding করা
- **Q5.4** → কেন কোনটিতে কোন encoding — 2-3 বাক্যে explain করা

### 🎯 এখান থেকে কী অর্জন করব?

Real dataset-এ text data-কে ML-এর জন্য number-এ রূপান্তর করার পুরো process শিখব।

### 🧠 কীভাবে চিন্তা করতে হবে?

- `city_type`: Urban, Rural, Suburban — এগুলোর মধ্যে কোনো ক্রম নেই → **One-Hot**
- `satisfaction_level`: Low < Medium < High — স্পষ্ট ক্রম আছে → **Ordinal**

**Dataset-এ actual distribution:**
- `city_type`: Urban=39, Rural=32, Suburban=29
- `satisfaction_level`: Medium=36, Low=35, High=29


### Q5.1 — One-Hot Encoding for `city_type` [10 Marks]

```python
import pandas as pd
import numpy as np

# Now we'll load the dataset
df = pd.read_csv("02. Provided - module123_data.csv")
# Here we've loaded the data into df.

# Now we'll apply one-hot encoding to city_type using pandas
city_ohe = pd.get_dummies(df["city_type"], prefix="city_type")
# Here we've created 3 new binary columns: city_type_Rural, city_type_Suburban, city_type_Urban.

print("One-Hot Encoded city_type (first 5 rows):")
print(city_ohe.head())
```

`pd.get_dummies(df["city_type"], prefix="city_type")` → `city_type` column-এর প্রতিটি unique value-এর জন্য একটি করে নতুন binary column তৈরি হয়।
`prefix="city_type"` → নতুন column-এর নাম হয় `city_type_Rural`, `city_type_Suburban`, `city_type_Urban`।
প্রতিটি row-এ ঠিক **একটিমাত্র column-এ True**, বাকি দুটো False।


### Q5.2 — One-Hot Encoded Columns df-এ Attach করা [5 Marks]

```python
# Now we'll attach the one-hot columns to the main dataframe
city_ohe_int = city_ohe.astype(int)
df = pd.concat([df, city_ohe_int], axis=1)
# Here we've added the 3 new binary columns to df. True/False converted to 1/0.

print("df with One-Hot columns (first 5 rows):")
print(df[["city_type", "city_type_Rural",
          "city_type_Suburban", "city_type_Urban"]].head())
```

`astype(int)` → True → **1**, False → **0** — ML model number বোঝে, True/False নয়।
`pd.concat([df, city_ohe_int], axis=1)` → দুটো DataFrame-কে পাশাপাশি (column-wise) জোড়া লাগায়।
এখন main `df`-এ তিনটি নতুন column যুক্ত হয়েছে।


### Q5.3 — Ordinal Encoding for `satisfaction_level` [10 Marks]

```python
# Now we'll define the ordinal mapping: Low=0, Medium=1, High=2
satisfaction_map = {"Low": 0, "Medium": 1, "High": 2}
# Here we've set the correct order — Low is smallest, High is largest.

# Now we'll apply the mapping to create a new encoded column
df["satisfaction_encoded"] = df["satisfaction_level"].map(satisfaction_map)
# Here we've created 'satisfaction_encoded' with numeric values 0, 1, or 2.

print("Ordinal Encoded satisfaction_level (first 8 rows):")
print(df[["satisfaction_level", "satisfaction_encoded"]].head(8).to_string(index=False))
```

`{"Low": 0, "Medium": 1, "High": 2}` → Python dictionary দিয়ে mapping define করা।
`.map(satisfaction_map)` → প্রতিটি row-এর text value-কে dictionary অনুযায়ী number-এ বদলে দেয়।
এই approach-এ ক্রম সংরক্ষিত থাকে: Low(0) < Medium(1) < High(2)।


### Q5.4 — Short Answer [5 Marks]

> **Question:** In 2 to 3 sentences, explain why one-hot encoding is suitable for `city_type` and why ordinal encoding is suitable for `satisfaction_level`.

---

**Answer:**

> One-hot encoding is suitable for `city_type` because the three categories — Urban, Rural, and Suburban — have no natural order or ranking between them, and assigning numbers like 0, 1, 2 would incorrectly imply that one city type is "greater" than another, which could mislead the model. Ordinal encoding is suitable for `satisfaction_level` because the three values — Low, Medium, and High — follow a clear and meaningful order where Low < Medium < High, so mapping them to 0, 1, and 2 correctly preserves that relationship and allows the model to understand that higher numbers represent greater satisfaction.

> `city_type`-এর জন্য one-hot encoding উপযুক্ত কারণ এর তিনটি category — Urban, Rural, এবং Suburban — এর মধ্যে কোনো স্বাভাবিক ক্রম বা ranking নেই, আর ০, ১, ২-এর মতো সংখ্যা বরাদ্দ করলে ভুলভাবে মনে হবে যে এক ধরনের city অন্যটির চেয়ে "বড়", যা model-কে বিভ্রান্ত করতে পারে। `satisfaction_level`-এর জন্য ordinal encoding উপযুক্ত কারণ এর তিনটি মান — Low, Medium, এবং High — একটি স্পষ্ট ও অর্থবহ ক্রম অনুসরণ করে যেখানে Low < Medium < High, তাই এগুলোকে ০, ১, এবং ২-এ map করলে সেই সম্পর্কটা সঠিকভাবে বজায় থাকে এবং model বুঝতে পারে যে বড় সংখ্যা মানে বেশি satisfaction।
