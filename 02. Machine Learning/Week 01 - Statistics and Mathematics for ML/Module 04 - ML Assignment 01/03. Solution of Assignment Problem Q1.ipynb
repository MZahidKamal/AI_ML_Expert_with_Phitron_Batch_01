## Q1. Descriptive Statistics on a Numeric Feature

---

### 🔍 এই Question-এ কী চাওয়া হয়েছে?

`daily_screen_time_min` column-এর উপর `.describe()` run করে তার **count, mean, std, min, max** দেখতে হবে। তারপর **Q1.2**-এ 2-3 বাক্যে comment করতে হবে।

### 🎯 এখান থেকে কী অর্জন করব?

একটি numeric column-এর **central tendency** (mean) ও **spread** (std) দেখে বুঝতে পারব data কতটা ছড়িয়ে আছে এবং কোনো extreme মান (outlier) আছে কিনা।

### 🧠 কীভাবে চিন্তা করতে হবে?

- **mean vs max দূরত্ব বড়?** → outlier থাকতে পারে
- **std বড়?** → data অনেক ছড়িয়ে আছে
- **min কি 0 বা negative?** → data-র nature বুঝতে সাহায্য করে


### Q1.1 — Numeric Column Select ও `.describe()` Run করা

```python
import pandas as pd
import numpy as np

# Now we'll load the dataset and select the numeric column
df = pd.read_csv("02. Provided - module123_data.csv")
num_col = "daily_screen_time_min"
# Here we've loaded the dataset and chosen our target column.

# Now we'll run describe() to get summary statistics
df[num_col].describe()
# Here we've generated count, mean, std, min, 25%, 50%, 75%, max for the column.
```

`df[num_col].describe()` → একটি numeric column-এর সব গুরুত্বপূর্ণ statistics একসাথে দেখায়:
- **count** → কতটি non-null মান আছে (এখানে 100)
- **mean** → গড় মান (181.89 মিনিট)
- **std** → standard deviation — data কতটা ছড়িয়ে আছে (68.89 মিনিট)
- **min / max** → সর্বনিম্ন (60) ও সর্বোচ্চ (299) মান
- **25%, 50%, 75%** → quartile — data-র distribution বোঝায়
