### Vector কী?
ML-এ প্রতিটি data point একটি **vector** — সংখ্যার একটি ordered list।
যেমন একজন মানুষের `[Height, Weight, Age]` = `[170, 65, 25]` একটি vector।

### Dot Product:
দুটো vector-এর **corresponding element** গুলো গুণ করে সব যোগ করা হয়।

$$\mathbf{a} \cdot \mathbf{b} = \sum_{i} a_i \times b_i = a_1 b_1 + a_2 b_2 + a_3 b_3$$

**ML-এ ব্যবহার:** দুটো vector কতটা **similar direction**-এ আছে তা মাপে। Neural network-এর প্রতিটি layer-এ dot product ব্যবহার হয়।

### L1 Norm (Manhattan Distance):
সব element-এর **absolute value-এর যোগফল**।

$$\|\mathbf{a}\|_1 = \sum_{i} |a_i| = |a_1| + |a_2| + |a_3|$$

**ML-এ ব্যবহার:** **Lasso Regression**-এ — sparse model তৈরিতে সাহায্য করে।

### L2 Norm (Euclidean Distance):
সব element-এর **বর্গের যোগফলের বর্গমূল**।

$$\|\mathbf{a}\|_2 = \sqrt{\sum_{i} a_i^2} = \sqrt{a_1^2 + a_2^2 + a_3^2}$$

**ML-এ ব্যবহার:** **Ridge Regression**, **KNN**, **distance calculation**-এ — সবচেয়ে বেশি ব্যবহৃত norm।

## Step 1: Vector Define করা

```python
import numpy as np

a = np.array([3, -1, 2])
b = np.array([4,  0, -2])

print("Vector a:", a)
print("Vector b:", b)
```

`np.array()` দিয়ে Python list-কে NumPy vector-এ রূপান্তর করা হয়েছে।
NumPy vector-এ mathematical operation করা অনেক সহজ ও দ্রুত।


---

## Task (a): Dot Product — হাতে হিসাব

$$\mathbf{a} \cdot \mathbf{b} = (a_1 \times b_1) + (a_2 \times b_2) + (a_3 \times b_3)$$

$$= (3 \times 4) + ((-1) \times 0) + (2 \times (-2))$$

$$= 12 + 0 + (-4)$$

$$= \boxed{8}$$


```python
dot_manual = (a[0]*b[0]) + (a[1]*b[1]) + (a[2]*b[2])
dot_numpy  = np.dot(a, b)

print(f"Manual dot product : {dot_manual}")
print(f"NumPy  dot product : {dot_numpy}")
print(f"Match: {dot_manual == dot_numpy}")
```

`np.dot(a, b)` — NumPy-র built-in dot product function।
`a[0]`, `a[1]`, `a[2]` দিয়ে vector-এর individual element access করা হয়েছে।
দুটো result মিলে গেলে `Match: True` দেখাবে।


---

## Task (b): L1 Norm — হাতে হিসাব

$$\|\mathbf{a}\|_1 = |3| + |-1| + |2| = 3 + 1 + 2 = \boxed{6}$$


```python
l1_manual = abs(a[0]) + abs(a[1]) + abs(a[2])
l1_numpy  = np.linalg.norm(a, ord=1)

print(f"Manual L1 Norm : {l1_manual}")
print(f"NumPy  L1 Norm : {l1_numpy}")
print(f"Match: {l1_manual == l1_numpy}")
```

`abs()` — Python-এর built-in absolute value function।
`np.linalg.norm(a, ord=1)` — `ord=1` parameter দিয়ে L1 Norm নির্দিষ্ট করা হয়।
`np.linalg` হলো NumPy-র **linear algebra** module।


---

## Task (b): L2 Norm — হাতে হিসাব

$$\|\mathbf{a}\|_2 = \sqrt{3^2 + (-1)^2 + 2^2} = \sqrt{9 + 1 + 4} = \sqrt{14} \approx \boxed{3.742}$$


```python
l2_manual = np.sqrt(a[0]**2 + a[1]**2 + a[2]**2)
l2_numpy  = np.linalg.norm(a, ord=2)

print(f"Manual L2 Norm : {l2_manual:.4f}")
print(f"NumPy  L2 Norm : {l2_numpy:.4f}")
print(f"sqrt(14)       : {np.sqrt(14):.4f}")
```

`np.sqrt()` — NumPy-র square root function।
`**2` — Python-এ power operator (বর্গ করা)।
`np.linalg.norm(a, ord=2)` — `ord=2` দিয়ে L2 Norm। `ord` না লিখলেও default L2 হয়।

## Task (c): NumPy Snippet — সব একসাথে Verify

এটি একটি clean summary print — সব হিসাব একনজরে দেখা যাচ্ছে।

### L1 vs L2 Norm — পার্থক্য মনে রাখার সহজ উপায়:

| | L1 Norm | L2 Norm |
|---|---|---|
| **Formula** | $\sum |a_i|$ | $\sqrt{\sum a_i^2}$ |
| **Result for a** | 6 | ≈ 3.742 |
| **Geometry** | Manhattan (grid) distance | Euclidean (straight line) distance |
| **ML use** | Lasso Regression | Ridge Regression, KNN |
| **Outlier sensitivity** | কম sensitive | বেশি sensitive (বর্গ করার কারণে) |
