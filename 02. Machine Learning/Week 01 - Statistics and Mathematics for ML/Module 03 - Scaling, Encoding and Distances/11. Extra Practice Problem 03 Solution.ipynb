# Problem 3 — Nominal vs Ordinal and Encoding

---

## 🔍 Problem-এ কী চাওয়া হয়েছে?

আমাদের কাছে ৪ জন মানুষের একটি dataset আছে:

| ID | City | Education_Level | Satisfaction |
|---|---|---|---|
| 1 | Dhaka | High School | Low |
| 2 | Chattogram | Bachelor | Medium |
| 3 | Rajshahi | Master | High |
| 4 | Dhaka | Bachelor | Medium |

তিনটি কাজ করতে হবে:

| Task | কী করতে হবে |
|---|---|
| **(a)** | কোন column **Nominal** আর কোনটি **Ordinal** — সেটা চিহ্নিত করা |
| **(b)** | `City` column-এ **One-Hot Encoding** করা — হাতে হিসাব করে দেখানো |
| **(c)** | `Education_Level`-এ **Ordinal Encoding** করা: High School=0, Bachelor=1, Master=2 |


---

## 🎯 এই কাজ থেকে আমরা কী অর্জন করতে পারব?

- **Nominal** ও **Ordinal** data-র মধ্যে পার্থক্য স্পষ্টভাবে বুঝতে পারব।
- ML model text data সরাসরি বুঝতে পারে না — **encoding** কেন দরকার সেটা বুঝব।
- **One-Hot Encoding** কখন ব্যবহার করতে হয় এবং এটি কীভাবে কাজ করে জানব।
- **Ordinal Encoding** কখন ব্যবহার করতে হয় এবং কীভাবে order assign করতে হয় জানব।


---

## 🧠 আমরা যা শিখেছি, সেই আলোকে কীভাবে চিন্তা করতে হবে?

### Nominal vs Ordinal চেনার নিয়ম:

> **মূল প্রশ্ন: এই column-এর values-এর মধ্যে কি কোনো স্বাভাবিক ক্রম (natural order) আছে?**

| যদি... | তাহলে... | উদাহরণ |
|---|---|---|
| **না**, কোনো ক্রম নেই — শুধু আলাদা আলাদা category | **Nominal** | City: Dhaka, Chattogram, Rajshahi — কোনোটি "বড়" বা "ছোট" নয় |
| **হ্যাঁ**, একটা স্পষ্ট ক্রম আছে | **Ordinal** | Education: High School < Bachelor < Master |

### কোন Encoding কখন ব্যবহার করব:

| Data Type | Encoding | কারণ |
|---|---|---|
| **Nominal** | **One-Hot Encoding** | কোনো ক্রম নেই — সংখ্যা দিলে model ভুল ক্রম শিখবে |
| **Ordinal** | **Ordinal Encoding** | ক্রম আছে — সংখ্যা দিয়ে সেই ক্রম প্রকাশ করা যায় |

### One-Hot Encoding কীভাবে কাজ করে:
- প্রতিটি unique category-র জন্য একটি নতুন column তৈরি হয়।
- যে row-এ সেই category আছে সেখানে **1**, বাকি সব জায়গায় **0**।
- `City`-তে ৩টি unique value (Dhaka, Chattogram, Rajshahi) → ৩টি নতুন column।


---

## 🛠️ Problem Solve করার Approach

**Step 1:** Dataset তৈরি করা।

**Step 2 (Task a):** প্রতিটি column বিশ্লেষণ করে Nominal/Ordinal চিহ্নিত করা।

**Step 3 (Task b):** `City`-তে One-Hot Encoding — হাতের হিসাব দেখানো, তারপর code দিয়ে করা।

**Step 4 (Task c):** `Education_Level`-এ Ordinal Encoding করা।

**Step 5:** Final encoded dataset একসাথে দেখা।


## Step 1: Dataset তৈরি করা

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

df = pd.DataFrame({
    'ID':              [1, 2, 3, 4],
    'City':            ['Dhaka', 'Chattogram', 'Rajshahi', 'Dhaka'],
    'Education_Level': ['High School', 'Bachelor', 'Master', 'Bachelor'],
    'Satisfaction':    ['Low', 'Medium', 'High', 'Medium']
})

print(df.to_string(index=False))
```

`pd.DataFrame` দিয়ে dictionary থেকে সরাসরি table তৈরি হয়েছে।
প্রতিটি key একটি column, প্রতিটি list সেই column-এর values।


---

## Task (a): Nominal vs Ordinal চিহ্নিত করা

| Column | Type | কারণ |
|---|---|---|
| `City` | **Nominal** | Dhaka, Chattogram, Rajshahi — কোনো স্বাভাবিক ক্রম নেই |
| `Education_Level` | **Ordinal** | High School → Bachelor → Master — স্পষ্ট ক্রম আছে |
| `Satisfaction` | **Ordinal** | Low → Medium → High — স্পষ্ট ক্রম আছে |

> **Note:** `Satisfaction`-ও Ordinal, কিন্তু এই problem-এ শুধু `City` ও `Education_Level` encode করতে বলা হয়েছে।


```python
column_types = {
    'City':            'Nominal  — no natural order',
    'Education_Level': 'Ordinal  — High School < Bachelor < Master',
    'Satisfaction':    'Ordinal  — Low < Medium < High'
}

for col, dtype in column_types.items():
    print(f"{col:20s} → {dtype}")
```

এটি একটি সহজ **reference print** — কোন column কী ধরনের তা একনজরে দেখা যাচ্ছে।


---

## Task (b): One-Hot Encoding — `City` column

### হাতে হিসাব:

`City`-তে ৩টি unique value: **Dhaka**, **Chattogram**, **Rajshahi**
→ ৩টি নতুন column তৈরি হবে: `City_Dhaka`, `City_Chattogram`, `City_Rajshahi`

| ID | City | City_Dhaka | City_Chattogram | City_Rajshahi |
|---|---|---|---|---|
| 1 | Dhaka      | **1** | 0 | 0 |
| 2 | Chattogram | 0 | **1** | 0 |
| 3 | Rajshahi   | 0 | 0 | **1** |
| 4 | Dhaka      | **1** | 0 | 0 |

**নিয়ম:** প্রতিটি row-এ ঠিক একটিমাত্র column-এ 1 থাকবে, বাকি সব 0।


```python
city_encoded = pd.get_dummies(df['City'], prefix='City')
print("One-Hot Encoded City:" + "\n" + city_encoded.to_string(index=False))

df_with_ohe = pd.concat([df[['ID', 'City']], city_encoded], axis=1)
print("DataFrame with One-Hot Encoded City:")
print(df_with_ohe.to_string(index=False))
```

`pd.get_dummies()` — pandas-এর built-in One-Hot Encoding function।
- `prefix='City'` → নতুন column-এর নাম হয় `City_Dhaka`, `City_Chattogram` ইত্যাদি।

`pd.concat(..., axis=1)` → দুটো DataFrame-কে পাশাপাশি (column-wise) জোড়া লাগায়।

**লক্ষ্য করো:** ID 1 ও ID 4 উভয়ই Dhaka → দুটো row-এই `City_Dhaka = 1`।


---

## Task (c): Ordinal Encoding — `Education_Level` column

### হাতে হিসাব:

নির্ধারিত mapping:
- `High School` → **0**
- `Bachelor`    → **1**
- `Master`      → **2**

| ID | Education_Level | Education_Encoded |
|---|---|---|
| 1 | High School | **0** |
| 2 | Bachelor    | **1** |
| 3 | Master      | **2** |
| 4 | Bachelor    | **1** |

**নিয়ম:** ক্রম অনুযায়ী সংখ্যা assign করা হয় — ছোট সংখ্যা মানে নিচের স্তর।


```python
education_order = [['High School', 'Bachelor', 'Master']]

enc = OrdinalEncoder(categories=education_order)
df['Education_Encoded'] = enc.fit_transform(df[['Education_Level']]).astype(int)

print("Ordinal Encoded Education_Level:")
print(df[['ID', 'Education_Level', 'Education_Encoded']].to_string(index=False))
```

`OrdinalEncoder(categories=...)` — sklearn-এর Ordinal Encoding function।
- `categories` parameter-এ list-of-lists দিতে হয় — ভেতরের list-এ **ক্রম অনুযায়ী** category সাজানো থাকে।
- `fit_transform()` → data দেখে শেখে, তারপর transform করে।
- `.astype(int)` → output default float হয়, তাই int-এ convert করা হয়েছে।


---

## Step 5: Final Encoded Dataset একসাথে দেখা


```python
final_df = pd.concat([
    df[['ID', 'Education_Level', 'Education_Encoded', 'Satisfaction']],
    city_encoded
], axis=1)

final_df = final_df[
    ['ID', 'City_Dhaka', 'City_Chattogram', 'City_Rajshahi', 'Education_Level', 'Education_Encoded', 'Satisfaction']
]

print("Final Encoded Dataset:")
print(final_df.to_string(index=False))
```

এই final dataset-এ:
- `City` → ৩টি binary column-এ রূপান্তরিত (**One-Hot Encoding**)।
- `Education_Level` → একটি numeric column-এ রূপান্তরিত (**Ordinal Encoding**)।
- এখন কোনো column-এ text নেই — ML model সরাসরি এই data দিয়ে কাজ করতে পারবে।

### ⚠️ গুরুত্বপূর্ণ মনে রাখার বিষয়:
- **One-Hot Encoding**-এ column সংখ্যা বাড়ে — অনেক unique category থাকলে dataset বড় হয়।
- **Ordinal Encoding**-এ ভুল order দিলে model ভুল শিখবে — তাই order সতর্কতার সাথে দিতে হবে।
