## Part 1 — Probability-র মূল ধারণা

### Probability কী?

Probability হলো **অনিশ্চয়তার গণিত** — কোনো ঘটনা ঘটার সম্ভাবনা কতটুকু তা পরিমাপ করে।

ML-এ model কখনো absolute সিদ্ধান্ত নেয় না — বরং data-র উপর ভিত্তি করে **likelihood estimate** করে।

| Situation | Probability | অর্থ |
|---|---|---|
| Email spam হওয়ার সম্ভাবনা | 0.9 | Model 90% confident এটি spam |
| Loan default হওয়ার সম্ভাবনা | 0.2 | 20% risk of default |
| Patient-এর diabetes থাকার সম্ভাবনা | 0.6 | আরও পরীক্ষার দরকার হতে পারে |

**Probability-র সীমা:**

| মান | অর্থ |
|---|---|
| 0 | ঘটনাটি কখনোই ঘটবে না |
| 0.5 | ঘটবে বা ঘটবে না — সমান সম্ভাবনা |
| 1 | নিশ্চিতভাবে ঘটবে |

$$0 \leq P(E) \leq 1$$

---

### Sample Space কী?

একটি experiment-এর **সব সম্ভাব্য outcome-এর সমষ্টি**।

| Experiment | Sample Space |
|---|---|
| Coin toss | S = {Heads, Tails} |
| Dice roll | S = {1, 2, 3, 4, 5, 6} |

$$P(\text{Event}) = \frac{\text{Favorable outcomes}}{\text{Total outcomes}}$$

> 🔑 ML-এ Sample Space = সব সম্ভাব্য data points-এর universe। Dataset সেই জগৎ define করে যেখান থেকে model শেখে।

---

### Events ও তাদের ধরন

Event হলো Sample Space-এর একটি **subset** — যে outcome-টি আমরা দেখতে চাই।

| ধরন | অর্থ | উদাহরণ |
|---|---|---|
| Simple event | একটিমাত্র outcome | 3 পাওয়া |
| Compound event | একাধিক outcome | জোড় সংখ্যা পাওয়া |
| Complementary event | বাকি সব outcome | জোড় না পাওয়া |

$$P(E^c) = 1 - P(E)$$

## Part 2 — Basic Probability Rules (চারটি মূল নিয়ম)

---

### Rule 1: Probability Range Rule
$$0 \leq P(E) \leq 1$$
> ⚠️ ML model যদি 1-এর বেশি probability দেয় → math বা scaling-এ ভুল আছে।

---

### Rule 2: Complement Rule
$$P(E) + P(E^c) = 1$$
**উদাহরণ:** User product buy করবে → P = 0.7, তাই buy করবে না → P = 0.3

> 💡 Logistic regression ঠিক এভাবেই কাজ করে।

---

### Rule 3: Addition Rule (OR Events)

**Mutually exclusive** (একসাথে হতে পারে না) হলে:
$$P(A \text{ or } B) = P(A) + P(B)$$

**একসাথে হতে পারলে** (General Addition Rule):
$$P(A \text{ or } B) = P(A) + P(B) - P(A \cap B)$$

> ⚠️ Overlap না বাদ দিলে double-counting হবে।

---

### Rule 4: Multiplication Rule (AND Events)

A ও B **independent** হলে:
$$P(A \cap B) = P(A) \times P(B)$$

**উদাহরণ:** পরপর দুটো heads পাওয়া → 0.5 × 0.5 = 0.25

> 💡 Naive Bayes এই independence assumption ব্যবহার করে — তাই নাম "Naive"।

---

## Part 3 — Conditional Probability

### Conditional Probability কী?

**"B ঘটেছে"** জানার পর **"A ঘটার সম্ভাবনা"** কত।

$$P(A \mid B) = \frac{P(A \cap B)}{P(B)}$$

> 💡 **কেন P(B) দিয়ে ভাগ করি?** B ঘটেছে জানার পর sample space আর পুরোটা নয় — শুধু B-র মধ্যে সীমাবদ্ধ হয়ে যায়।

**Simple vs Conditional:**
- Simple: P(pizza কেনা) — কোনো শর্ত নেই
- Conditional: P(pizza কেনা | Friday night) — শর্ত আছে

---

### Visual উদাহরণ — Contingency Table

**100 জন Student:**

| | Tea | Not Tea | Total |
|---|---|---|---|
| Coffee | 30 | 30 | 60 |
| Not Coffee | 10 | 30 | 40 |

**প্রশ্ন:** Tea পছন্দ করে জানলে — Coffee পছন্দ করার সম্ভাবনা কত?

$$P(\text{Coffee} \mid \text{Tea}) = \frac{30}{40} = 0.75$$

40 দিয়ে ভাগ কারণ — Tea পছন্দ করে এমন 40 জনই এখন আমাদের নতুন sample space।

---

### Venn Diagram — Overlap মানেই Conditional Probability

```
   ╭──────────╮   ╭──────────╮
   │  Math    │///│  Science │
   │  (40)    │(40│  (30)    │
   │          ////│          │
   ╰──────────╯   ╰──────────╯
              ↑
         Overlap = P(A∩B)
```

---

### ML-এ Conditional Probability-র ব্যবহার

| Application | Conditional Probability |
|---|---|
| Spam detection | P(Spam \| "lottery" শব্দ আছে) |
| Medical diagnosis | P(Disease \| Test Positive) |
| Recommender systems | P(User likes item \| Similar items পছন্দ করেছে) |

## Part 4 — Independence ও Conditional Independence

### Independence কী?

A ও B **independent** যদি — একটি ঘটলেও অন্যটির সম্ভাবনা না বদলায়।

$$P(A \mid B) = P(A) \quad \text{এবং} \quad P(A \cap B) = P(A) \times P(B)$$

**Independent vs Dependent:**

| | Independent | Dependent |
|---|---|---|
| সম্পর্ক | কোনো সম্পর্ক নেই | একটি অন্যটিকে প্রভাবিত করে |
| Venn Diagram | কোনো overlap নেই | Overlap আছে |

```
Independent (No Overlap):     Dependent (Overlap আছে):
  ╭──────╮   ╭──────╮           ╭─────╮╭─────╮
  │  A   │   │  B   │           │  A  ││  B  │
  │      │   │      │           │    ╭╯╰╮    │
  ╰──────╯   ╰──────╯           ╰────╯  ╰────╯
```

---

### Conditional Independence কী?

A এবং B **conditionally independent given C** — যদি C জানা থাকলে B আর A সম্পর্কে কোনো নতুন তথ্য না দেয়।

$$P(A \mid B, C) = P(A \mid C)$$

**উদাহরণ — Ice Cream:**
- A = ice cream কেনা, B = sunny কিনা, C = Temperature
- Temperature (C) জানা থাকলে — sunny কিনা (B) জানলে বাড়তি কিছু জানা যায় না

**উদাহরণ — Exam Pass:**
- C = মোট কত ঘণ্টা পড়েছে জানলে → বন্ধুদের সাথে পড়েছে কিনা জানলে আর বাড়তি তথ্য আসে না

**ML-এ ব্যবহার:**

| Model | কীভাবে ব্যবহার করে |
|---|---|
| Naive Bayes | Class জানলে সব features independent ধরে |
| Bayesian Networks | Variables-এর মধ্যে conditional independence map করে |
| Hidden Markov Models | Current state জানলে past irrelevant ধরে |

## Part 5 — Bayes' Theorem ও Base-Rate Effect

### Bayes' Theorem কী?

নতুন evidence পেলে **আগের belief আপডেট** করার গাণিতিক পদ্ধতি।

$$P(A \mid B) = \frac{P(B \mid A) \times P(A)}{P(B)}$$

**তিনটি ধাপ:**

| ধাপ | নাম | অর্থ |
|---|---|---|
| 1 | Prior P(A) | আগে থেকে যা বিশ্বাস করি |
| 2 | Likelihood P(B\|A) | Belief সত্যি হলে এই evidence দেখার সম্ভাবনা |
| 3 | Posterior P(A\|B) | Evidence দেখার পরে নতুন belief |

> 💡 **এক কথায়:** Prior + নতুন Evidence = Posterior

---

### Base-Rate Effect — গুরুত্বপূর্ণ উদাহরণ

**পরিস্থিতি:**
- রোগটি হয় মাত্র 1000 জনে 1 জনের (0.1%)
- Test 99% accurate
- Test positive এলো — আসলেই রোগ আছে?

🧠 আমাদের gut বলে → **"99%!"** — কিন্তু এটা ভুল।

**সঠিক হিসাব (1000 জনের মধ্যে):**

| ফলাফল | সংখ্যা |
|---|---|
| আসলেই রোগ আছে + test positive ✅ | 1 জন |
| রোগ নেই, কিন্তু test positive ❌ (false positive) | ~10 জন |
| মোট positive | 11 জন |

$$P(\text{Disease} \mid \text{Positive}) = \frac{1}{11} \approx 9\%$$

> 😲 Test 99% accurate হওয়ার পরেও সত্যিকারের সম্ভাবনা মাত্র ~9%!

---

### Real-Life উদাহরণ — Sports Car

- শহরে চোরাই গাড়ির হার: 10,000-এ 1টি (খুবই rare)
- চোরাই গাড়ির মধ্যে sports car: 80%
- সাধারণ গাড়ির মধ্যে sports car: 10%

**হিসাব (10,000 গাড়ির মধ্যে):**
- Stolen sports cars = 80% × 1 = **0.8**
- Non-stolen sports cars = 10% × 9,999 = **~1,000**

$$P(\text{Stolen} \mid \text{Sports Car}) = \frac{0.8}{0.8 + 1000} \approx 0.08\%$$

> ⚠️ Base-Rate Effect: ঘটনাটি যখন **extremely rare**, তখন শক্তিশালী clue দিয়েও সম্ভাবনা বেশি হয় না।

---

### Spam Detection-এ Bayes' Theorem

$$P(\text{Spam} \mid \text{Word}) \propto P(\text{Word} \mid \text{Spam}) \times P(\text{Spam})$$

- "lottery" শব্দটি বেশিরভাগ spam-এ থাকে → P(Word|Spam) বেশি
- কিন্তু spam-এর base rate মাত্র 5% হলে → 100% certain হওয়া যাবে না

> 💡 ML model-কে শেখায় — **সবসময় skeptical থাকো, base rate কখনো ignore করো না।**

## Part 6 — Confusion Matrix ও Model Evaluation Metrics

### Confusion Matrix কী?

Model কোথায় সঠিক, কোথায় ভুল — তার **report card**।

| | Predicted Positive | Predicted Negative |
|---|---|---|
| **Actual Positive** | True Positive (TP) ✅ | False Negative (FN) ❌ |
| **Actual Negative** | False Positive (FP) ❌ | True Negative (TN) ✅ |

**মনে রাখার সহজ উপায়:**
- **True** → model সঠিক সিদ্ধান্ত নিয়েছে
- **False** → model ভুল সিদ্ধান্ত নিয়েছে
- **Positive** → model বলেছে "আছে"
- **Negative** → model বলেছে "নেই"

---

### উদাহরণ — 100 জন Diabetes Test

| ফলাফল | সংখ্যা |
|---|---|
| TP: diabetes আছে + model ধরেছে | 9 |
| FN: diabetes আছে কিন্তু model miss করেছে | 1 |
| FP: সুস্থ কিন্তু model ভুলভাবে diabetic বলেছে | 5 |
| TN: সুস্থ + model সঠিকভাবে সুস্থ বলেছে | 85 |

---

### Sensitivity (Recall) — আসল Positive কতটা ধরা পড়লো?

$$\text{Sensitivity} = \frac{TP}{TP + FN} = \frac{9}{9+1} = 90\%$$

> 🚒 **Analogy:** High sensitivity smoke detector → সামান্য ধোঁয়াতেও বেজে ওঠে, আগুন কখনো miss করবে না।
> ⚠️ Healthcare-এ FN সবচেয়ে বিপজ্জনক — রোগীকে miss করা যাবে না।

---

### Specificity — আসল Negative কতটা সঠিকভাবে চিহ্নিত হলো?

$$\text{Specificity} = \frac{TN}{TN + FP} = \frac{85}{85+5} = 94.4\%$$

> 🛡 **Analogy:** High specificity security guard → নির্দোষ মানুষকে ভুলভাবে আটকায় না।

---

### False Positive Rate (FPR) ও False Negative Rate (FNR)

$$\text{FPR} = \frac{FP}{FP + TN} = 1 - \text{Specificity}$$

$$\text{FNR} = \frac{FN}{FN + TP} = 1 - \text{Sensitivity}$$

**Trade-off:**

| ক্ষেত্র | বেশি ক্ষতিকর কোনটা |
|---|---|
| 🏥 Cancer detection | FNR — রোগী miss করা বিপজ্জনক |
| 📧 Spam filter | FPR — important email spam-এ যাওয়া সমস্যা |

---

### PPV ও NPV — Prediction কতটা বিশ্বাসযোগ্য?

**PPV (Positive Predictive Value) = Precision:**
$$\text{PPV} = \frac{TP}{TP + FP} = \frac{9}{9+5} = 64\%$$

**NPV (Negative Predictive Value):**
$$\text{NPV} = \frac{TN}{TN + FN} = \frac{85}{85+1} = 98.8\%$$

| Metric | মান | অর্থ |
|---|---|---|
| PPV | 64% | Positive prediction মাত্র 64% সঠিক |
| NPV | 98.8% | Negative prediction প্রায় সবসময় সঠিক |

> ⚠️ Model "সুস্থ" বলতে ভালো — কিন্তু "রোগী" ধরতে দুর্বল। এটাই imbalanced dataset-এর সাধারণ সমস্যা।

---

### সব Metrics এক নজরে

| Metric | Formula | কী পরিমাপ করে |
|---|---|---|
| Sensitivity/Recall | TP/(TP+FN) | আসল positive কতটা ধরা পড়ছে |
| Specificity | TN/(TN+FP) | আসল negative কতটা সঠিকভাবে চিহ্নিত |
| FPR | FP/(FP+TN) | সুস্থকে ভুলভাবে রোগী বলার হার |
| FNR | FN/(FN+TP) | রোগীকে ভুলভাবে সুস্থ বলার হার |
| PPV/Precision | TP/(TP+FP) | Positive prediction কতবার সত্যিই সঠিক |
| NPV | TN/(TN+FN) | Negative prediction কতবার সত্যিই সঠিক |

## Part 8 — Class Imbalance ও তার প্রভাব

### Class Imbalance কী?

একটি class-এ অন্যটির চেয়ে **অনেক বেশি sample** থাকা।

| Dataset | Imbalance |
|---|---|
| Spam detection | বেশিরভাগ email spam নয় |
| Disease prediction | বেশিরভাগ মানুষ সুস্থ |
| Credit card fraud | বেশিরভাগ transaction safe |

---

### Accuracy কেন Misleading?

**উদাহরণ — Fraud Detection:**
10,000টি transaction → মাত্র 50টি fraudulent (0.5%)

Model যদি সবসময় **"no fraud"** বলে:
- Accuracy = **99.5%** ✅ (দেখতে ভালো)
- কিন্তু model সম্পূর্ণ useless ❌

> True Negatives score-কে dominate করে — তাই Accuracy বেশি দেখায়।

---

### Confusion Matrix-এ Imbalance-এর প্রভাব

```
Dataset: 990টি "No Fraud" + 10টি "Fraud"

            Pred Positive   Pred Negative
Actual Pos  TP (কম)         FN (miss হয়)
Actual Neg  FP (false alarm) TN (অনেক বেশি ← এটাই Accuracy inflate করে)
```

---

### সমাধান — Better Metrics ব্যবহার করো

| Metric | কী পরিমাপ করে |
|---|---|
| Precision | Predicted positive-এর মধ্যে কতটা সত্যিই সঠিক |
| Recall (Sensitivity) | আসল positive-এর মধ্যে কতটা ধরা পড়েছে |
| F1-Score | Precision ও Recall-এর balanced measure (পরে আসবে) |

**Context অনুযায়ী:**

| Scenario | বেশি গুরুত্বপূর্ণ | কারণ |
|---|---|---|
| 🏥 Medical diagnosis | Recall | রোগী miss করা বিপজ্জনক |
| 📧 Spam detection | Precision | Legit email block করা সমস্যা |

| Confusion Matrix | TP, FP, FN, TN দিয়ে model evaluate করা |
| Class Imbalance | Accuracy alone যথেষ্ট নয় — Precision ও Recall দেখো |
