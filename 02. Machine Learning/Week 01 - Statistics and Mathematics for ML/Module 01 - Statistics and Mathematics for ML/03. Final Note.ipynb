## Part 1 — Measures of Central Tendency (কেন্দ্রীয় প্রবণতার পরিমাপ)

Data-র "center" কোথায় সেটা বোঝার জন্য তিনটি measure ব্যবহার করা হয়।

---

### 1. Mean (গড়)

সব মান যোগ করে মোট সংখ্যা দিয়ে ভাগ করলে Mean পাওয়া যায়।

$$\text{Mean} = \frac{\sum x_i}{n}$$

**উদাহরণ:** `[10, 12, 14, 16, 18, 100]`
$$\text{Mean} = \frac{10+12+14+16+18+100}{6} = 28.3$$

> ⚠️ **সমস্যা:** Outlier (যেমন `100`) থাকলে Mean অনেকটা বেড়ে যায় — আসল "center" আর represent করে না।

---

### 2. Median (মধ্যমান)

Data ছোট থেকে বড় সাজিয়ে ঠিক মাঝের মানটিই Median। এটি data-কে দুটি সমান ভাগে ভাগ করে।

- সংখ্যা **বেজোড়** হলে → মাঝের মানটি নাও
- সংখ্যা **জোড়** হলে → মাঝের দুটির গড় নাও

**উদাহরণ:** Sorted: `[10, 12, 14, 16, 18, 100]`
$$\text{Median} = \frac{14+16}{2} = 15$$

> ✅ Outlier থাকলেও Median প্রভাবিত হয় না — তাই এটিকে **robust measure** বলে।

---

### 3. Mode (সর্বোচ্চ পুনরাবৃত্তি)

Dataset-এ যে মান **সবচেয়ে বেশিবার** আসে সেটাই Mode।

**উদাহরণ:** `[2, 2, 3, 3, 3, 4, 5]` → Mode = `3` (৩ বার এসেছে)

> 📌 একটি dataset-এ একাধিক Mode-ও থাকতে পারে।

---

### কোন situation-এ কোনটা ব্যবহার করবো?

| Situation | Best Measure | কারণ |
|---|---|---|
| Symmetric data (height, temperature) | Mean | সব মান সমানভাবে contribute করে |
| Skewed data / outlier (income, price) | Median | Outlier-এ প্রভাবিত হয় না |
| Categorical data (color, gender) | Mode | Non-numeric data-তেও কাজ করে |

---

## Part 2 — Variance ও Standard Deviation (ছড়িয়ে পড়ার পরিমাপ)

Mean কোথায় সেটা জানলাম। এখন জানবো data **কতটা ছড়িয়ে** আছে।

---

### Variance কী?

Data points গুলো mean থেকে গড়ে কতটা দূরে আছে তার **average squared deviation**।

**দুই ধরনের Variance:**

| | Population Variance (σ²) | Sample Variance (s²) |
|---|---|---|
| কখন | পুরো population জানা থাকলে | Subset থেকে হিসাব করলে |
| Formula | $\frac{\sum(x_i - \mu)^2}{n}$ | $\frac{\sum(x_i - \bar{x})^2}{n-1}$ |

> **Bessel's Correction:** Sample-এ `n` দিয়ে ভাগ করলে variance একটু underestimate হয়। তাই `n-1` ব্যবহার করা হয় — এটাই Bessel's Correction। এতে result **unbiased** হয়।

**উদাহরণ:** `[70, 80, 90]`, Mean = 80

| Data Point | $(x - \bar{x})^2$ |
|---|---|
| 70 | 100 |
| 80 | 0 |
| 90 | 100 |

$$\sigma^2 \text{ (Population)} = \frac{200}{3} \approx 66.67 \qquad s^2 \text{ (Sample)} = \frac{200}{2} = 100$$

---

### Standard Deviation (SD) কী?

Variance-এর **square root** — data-র মতোই same unit-এ থাকে, তাই বুঝতে সহজ।

$$\sigma = \sqrt{\sigma^2} \qquad s = \sqrt{s^2}$$

**উদাহরণ:** `[70, 80, 90]` → σ = √66.67 ≈ **8.16**
মানে scores গুলো mean থেকে সাধারণত ±8 points দূরে থাকে।

---

### Visual — SD-এর অর্থ

```
Small SD → curve লম্বা ও সরু (data কাছাকাছি)
Large SD → curve চওড়া ও চ্যাপ্টা (data ছড়িয়ে)

Normal Distribution-এ:
  ±1σ → 68% data
  ±2σ → 95% data
  ±3σ → 99.7% data
```

---

### ML-এ Variance ও SD-এর ব্যবহার

**Feature Scaling (Standardization):**
$$z_i = \frac{x_i - \mu}{\sigma}$$
এতে Mean = 0, SD = 1 হয়।

**Model-এ Variance মানে:**

| অবস্থা | সমস্যা |
|---|---|
| High Variance | Model training data মুখস্থ করে → Overfitting |
| Low Variance + High Bias | Model খুব সরলভাবে শেখে → Underfitting |

> 💡 **মূল লক্ষ্য:** Bias ও Variance-এর মধ্যে সঠিক balance খুঁজে বের করাই ভালো ML model তৈরির মূল চ্যালেঞ্জ।

```python
np.var(data, ddof=0)  # Population variance
np.var(data, ddof=1)  # Sample variance
np.std(data, ddof=0)  # Population SD
np.std(data, ddof=1)  # Sample SD
```

## Part 3 — Percentile, Quartile, IQR ও Z-Score

---

### Percentile কী?

একটি নির্দিষ্ট মানের **নিচে মোট data-র কত শতাংশ** আছে তা বলে।

$$\text{Percentile Rank} = \frac{\text{number of values below } x}{\text{total values}} \times 100$$

> **উদাহরণ:** তোমার score 90th percentile-এ থাকলে মানে তুমি 90% মানুষের চেয়ে ভালো করেছো।

---

### Quartiles কী?

Quartile হলো বিশেষ Percentile — data-কে **চারটি সমান ভাগে** ভাগ করে।

| Quartile | Percentile | অর্থ |
|---|---|---|
| Q1 | 25th | Data-র 25% এর নিচে |
| Q2 | 50th | Median |
| Q3 | 75th | Data-র 75% এর নিচে |

---

### IQR (Interquartile Range) কী?

Data-র **মাঝের 50%** কতটা ছড়িয়ে আছে তার পরিমাপ।

$$\text{IQR} = Q3 - Q1$$

**Outlier Detection Rule (IQR Fences):**
$$\text{Lower Bound} = Q1 - 1.5 \times IQR$$
$$\text{Upper Bound} = Q3 + 1.5 \times IQR$$

এই দুই সীমার বাইরে যেকোনো মান → **Outlier**

> ✅ IQR নিজেই outlier দ্বারা প্রভাবিত হয় না — তাই Mean বা SD-এর চেয়ে বেশি **reliable**।

---

### Boxplot — Visual Representation

```
        Outlier
          ●
    ┌─────────────┐
    │  Q1  │  Q3  │   ← Box = IQR
    └─────────────┘
  Whisker         Whisker
●                         ●
Min                      Max (outlier বাদে)
```
- Box-এর ভেতরের লাইন = **Median (Q2)**
- Whisker-এর বাইরের বিন্দু = **Outliers**

---

### Z-Score কী?

একটি মান, mean থেকে **কতটি standard deviation দূরে** আছে তা বলে।

$$z = \frac{x - \mu}{\sigma}$$

| Z-Score | অর্থ |
|---|---|
| z = 0 | মানটি ঠিক mean-এ |
| z = 1 | Mean-এর 1 SD উপরে |
| z = -2 | Mean-এর 2 SD নিচে |
| \|z\| > 3 | Outlier হিসেবে ধরা হয় |

> Z-Score করার পর → Mean = 0, SD = 1

**ML-এ জরুরি যেসব model-এ:**
- KNN (distance-based)
- SVM (margin-based)
- Gradient Descent-based models

## Part 4 — Distribution Shapes (বিতরণের আকৃতি)

---

### Symmetric Distribution

কেন্দ্রের **দুই পাশে একই রকম** — mirror image।

- Mean = Median = Mode (তিনটিই center-এ)
- সবচেয়ে পরিচিত উদাহরণ → **Normal Distribution**

```
         ┌─────────────────────────────────┐
         │          ╭───╮                  │
         │        ╭╯   ╰╮                 │
         │      ╭╯       ╰╮               │
         │  ───╯           ╰───           │
         └─────────────────────────────────┘
              Mean=Median=Mode (center)
```

> ✅ Symmetric data-তে Mean ব্যবহার করা সবচেয়ে নির্ভরযোগ্য।
> Linear Regression, SVM, KNN — এই algorithm গুলো normal distribution assume করে।

---

### Skewed Distribution

Symmetric **নয়** — দুই পাশ সমান নয়।

| ধরন | অন্য নাম | বৈশিষ্ট্য | উদাহরণ |
|---|---|---|---|
| Positively Skewed | Right-skewed | লেজ ডান দিকে লম্বা | Income, house price |
| Negatively Skewed | Left-skewed | লেজ বাম দিকে লম্বা | সহজ পরীক্ষার score |

```
Right-skewed:              Left-skewed:
 ██                               ██
 ████                         ████
 ██████──────                 ──────██████
 Mode < Median < Mean     Mean < Median < Mode
```

> ⚠️ Skewed data-তে Mean use করলে misleading result আসে — **Median ব্যবহার করো।**

**সমাধান — Transformation:**

| Transformation | কখন |
|---|---|
| Log Transformation | সবচেয়ে সহজ ও common |
| Box-Cox | শুধু positive data-তে |
| Yeo-Johnson | Positive ও negative উভয়তে |

---

### Long Tail Distribution

Distribution-এর কিছু extreme value center থেকে **অনেক দূরে লম্বা লেজের মতো** ছড়িয়ে থাকে।

**Real-world উদাহরণ:**
- YouTube views — কিছু video viral, বাকিগুলো কম
- Social media followers — কিছু account-এ লক্ষ লক্ষ, বাকিদের কম

> ⚠️ Long-tailed data সরাসরি model-এ দিলে model frequent cases-এ overfit করে এবং rare কিন্তু গুরুত্বপূর্ণ ঘটনা ignore করে।

**সমাধান:** Log transformation, Resampling, Outlier-aware models

---

### Skewness vs Long Tail — পার্থক্য

| Concept | Focus | উদাহরণ |
|---|---|---|
| Skewness | Data কোন দিকে হেলে আছে | Income distribution |
| Long Tail | Rare/extreme value কতটা বিস্তৃত | YouTube video views |

### Outlier Detection using IQR Fences

```python
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR
outliers = data[(data < lower_fence) | (data > upper_fence)]
```

> 💡 **সব outlier খারাপ নয়!** Fraud detection বা rare disease prediction-এ outlier-ই সবচেয়ে গুরুত্বপূর্ণ case। সেক্ষেত্রে remove না করে আলাদাভাবে handle করো।

## Part 5 — কখন Mean & SD বনাম Median & IQR?

| পরিস্থিতি | ব্যবহার করো |
|---|---|
| Data roughly normal (bell-shaped) | Mean & SD |
| Skewed / outlier-heavy / messy data | Median & IQR |
| sklearn-এ robust scaling দরকার | `RobustScaler` → Median & IQR দিয়ে scale করে |

```python
data = [25, 27, 28, 29, 30, 31, 35, 500]
np.mean(data)    # → 88.1  (outlier-এ টেনে গেছে)
np.median(data)  # → 29.5  (স্থির থেকেছে)
```

> 💡 **Rule of thumb:** সবার আগে একটি `histogram` বা `boxplot` দেখো — তারপর সিদ্ধান্ত নাও।
