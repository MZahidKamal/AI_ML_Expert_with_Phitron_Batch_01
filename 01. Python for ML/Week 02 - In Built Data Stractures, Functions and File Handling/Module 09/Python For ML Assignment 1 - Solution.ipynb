# Question: 2

```python
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Division by zero"


a, b = list(map(float, input().split()))
output = safe_divide(a, b)
print(output)
```

# Question: 3

```python
from abc import ABC, abstractmethod
from math import pi


#--------------------------------------


# Creating a base class
class Shape (ABC):

    @abstractmethod
    def area(self):
        return 0


#--------------------------------------


# Creating a subclass
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height


#--------------------------------------


# Creating a subclass
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return pi * self.radius ** 2


#--------------------------------------


userInput = input().split()
shape = userInput[0]

if shape == "Rectangle":
    shape, width, height = userInput
    rectangle = Rectangle(float(width), float(height))
    print(f"{rectangle.area():.2f}")

elif shape == "Circle":
    shape, radius = userInput
    circle = Circle(float(radius))
    print(f"{circle.area():.2f}")
```

# Question: 5

```python
from math import factorial


def factorialSum(numberList):
    summ = 0
    for digit in numberList:
        summ += factorial(digit)
    return summ


inputInteger = input()
# print(inputInteger)

intList = list(map(int, inputInteger))
# print(intList)

output = "Strong Number" if int(inputInteger) == factorialSum(intList) else "Not Strong Number"
print(output)
```
