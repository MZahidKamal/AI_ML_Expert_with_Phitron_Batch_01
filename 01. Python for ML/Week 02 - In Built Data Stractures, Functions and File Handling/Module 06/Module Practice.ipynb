# *args & **kwargs

*args ব্যবহার করা হয় যখন একটি function এ কতগুলো positional argument আসবে আগে থেকে জানা থাকে না। এটা সব argument কে একটি tuple আকারে নেয়।

**kwargs ব্যবহার করা হয় যখন কতগুলো named argument আসবে জানা থাকে না। এটা সব argument কে key-value pair আকারে একটি dictionary হিসেবে নেয়।

সংক্ষেপে:
*args → extra positional argument (tuple)
**kwargs → extra keyword argument (dictionary)

```python
def square_addition(*args):
    summation = 0
    for i in args:
        i = i ** 2
        summation += i
    return summation


ans = square_addition(10, 20, 30, 40, 50)
print(ans)
```

```python
def student(**kwargs):
    for key, val in kwargs.items():
        print(f"{key} : {val}")


student(name="adil", cls=10, roll=30, marks=30.5)
```

```python
dataSet = [x for x in range(30)]  # dataset


def data_loader(chunk_size, dataSet):
    for i in range(0, len(dataSet), chunk_size):
        yield lst[i:i + chunk_size]


x = data_loader(5, dataSet)

print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
# print(next(x))    # This will raise the 'StopIteration' error
```

Generator হলো Python এ এমন একটি বিশেষ function যেখানে return এর বদলে yield ব্যবহার করা হয়। yield ব্যবহার করলে function একবারে পুরো result না দিয়ে ধাপে ধাপে (one by one) value দেয়।

এখানে data_loader function একটি generator। এটা পুরো list একসাথে memory তে না তুলে, নির্দিষ্ট chunk_size অনুযায়ী ছোট ছোট অংশে data দিচ্ছে।

range(0, len(lst), chunk_size) দিয়ে step আকারে index এগোচ্ছে, আর lst[i:i + chunk_size] দিয়ে list slicing করে ছোট অংশ return করছে।

yield ব্যবহার করার কারণে function একবার value দেওয়ার পর pause হয়ে যায়। আবার next() call করলে সেখান থেকেই execution শুরু হয়।

এভাবে বড় dataset handle করলে memory efficient কাজ করা যায়।

সংক্ষেপে:
yield → ধাপে ধাপে data দেয়,
generator → memory efficient,
next() → পরের chunk আনে,
বড় data processing এর জন্য খুব গুরুত্বপূর্ণ ধারণা.

# Lambda function

lambda function হলো Python এ ছোট, এক লাইনের anonymous function। এখানে def ব্যবহার করা হয় না, বরং lambda keyword দিয়ে সরাসরি expression লেখা হয়।

এর syntax:
lambda parameters: expression

lambda function একটাই expression return করতে পারে, একাধিক statement লেখা যায় না।
এটা সাধারণত ছোট logic দ্রুত লেখার জন্য ব্যবহার করা হয়, বিশেষ করে map(), filter(), sorted(), reduce() এর সাথে।

উদাহরণ:
square = lambda x: x ** 2
এখানে x হলো parameter, আর x ** 2 হলো return হওয়া expression।

সংক্ষেপে:
lambda = এক লাইনের ছোট function
def ছাড়া function বানানোর উপায়
Functional programming এ বেশি ব্যবহার হয়

```python
def square_addition(x, y):
    return x ** 2 + y ** 2


print(square_addition(2, 3))

sq_add = lambda x, y: x ** 2 + y ** 2
print(sq_add)  # It will print the function address in memory
print(type(sq_add))  # it will print <class 'function'>
print(sq_add(2, 3))  # It will print the final result of the function
```

```python
if_even = lambda x: x % 2 == 0
print(if_even(11))
```

# Filter

filter এবং map দুটোই iterable data এর উপর কাজ করে, কিন্তু কাজের ধরন আলাদা।

map ব্যবহার করা হয় প্রতিটি element এর উপর একটি function প্রয়োগ করে নতুন value তৈরি করার জন্য। মানে element পরিবর্তন করে নতুন output দেয়।
filter ব্যবহার করা হয় condition অনুযায়ী কিছু element রেখে বাকিগুলো বাদ দেওয়ার জন্য। মানে element পরিবর্তন করে না, শুধু select করে।

map → transformation করে
filter → selection করে

উদাহরণ ধারণা:
map দিলে সব সংখ্যার square বের হবে।
filter দিলে শুধু even সংখ্যাগুলো রাখা হবে।

সংক্ষেপে:
map = প্রতিটি element বদলায়
filter = condition অনুযায়ী element বেছে নেয়

```python
numbers = [x for x in range(100)]
print(numbers)

# even
even = list(filter(lambda x: x % 2 == 0, numbers))
print(even)

fifty_upper = list(filter(lambda x: x > 50, numbers))
print(fifty_upper)
```

```python
data = [0, 1, '', None, 'Hello', [], [1, 2], None, True, False, None]

cleaned_data = list(filter(None, data))
print(cleaned_data)
```

এখানে filter(None, data) ব্যবহার করা হয়েছে।

filter এর ভিতরে None দিলে Python স্বয়ংক্রিয়ভাবে প্রতিটি element এর truthiness চেক করে।

যেসব element Falsy (যেমন 0, '', None, [], False) — সেগুলো বাদ পড়ে যায়।
যেসব element Truthy (যেমন 'Hello', [1, 2], True) — সেগুলো থেকে যায়।

মানে এখানে কোনো আলাদা function লেখা হয়নি, None নিজেই truth test হিসেবে কাজ করছে।

সংক্ষেপে:
filter(None, data) → সব Falsy value remove করে শুধু Truthy value রাখে।

```python
# string
# string = "HELLO WORLD, WELCOME TO THE AGE OF AI"
string = "hello world, welcome to the age of AI"
vowels = list(filter(lambda x: x.upper() in "AEIOU", string))

print(vowels)
```

```python
fruits = ['aa', 'bbb', 'cccc', 'ddddd']

filtering = list(filter(lambda x: len(x) > 2, fruits))

print(filtering)
```

# Reduce

reduce হলো Python এর একটি function যা functools module থেকে import করতে হয়।

reduce একটি iterable এর সব element একে একে নিয়ে একটি single value এ নামিয়ে আনে। মানে অনেকগুলো value কে combine করে একটি final result তৈরি করে।
এটা একটি function এবং iterable নেয়। প্রথমে প্রথম দুইটি element নিয়ে function apply করে, তারপর সেই result কে পরের element এর সাথে আবার function apply করে — এভাবে শেষ পর্যন্ত একটি final output তৈরি হয়।

উদাহরণ ধারণা:
সংখ্যাগুলোর যোগফল বের করা, গুণফল বের করা, maximum বের করা ইত্যাদি ক্ষেত্রে reduce ব্যবহার হয়।

map প্রতিটি element পরিবর্তন করে নতুন list দেয়।
filter condition অনুযায়ী কিছু element বেছে নেয়।
reduce সব element combine করে একটি মাত্র value দেয়।

সংক্ষেপে:
reduce = multiple value → single value
combine logic প্রয়োগ করে final result তৈরি করে

```python
from functools import reduce

lst = [1, 2, 3, 4, 5, 6]
summation = reduce(lambda x, y: x + y, lst)

print(summation)
```

```python
lst = [1, 2, 10, 4, 5, 6]
max_value = reduce(lambda x, y: x if x > y else y, lst)

print(max_value)
```

```python
lst = ["hello", "world", "welcome", "to", "programming"]
print(lst)

string = reduce(lambda x, y: x + " " + y, lst)
print(string)
```

#### Solution of the CODEFORCES problem: G. Max and MIN
Url: https://codeforces.com/group/MWSDmqGsZm/contest/223205/problem/G

```python
# Using lambda function
N = int(input())
ArrayA = list(map(int, input().split()))

maxA = reduce(lambda x, y: x if x > y else y, ArrayA)
minA = reduce(lambda x, y: x if x < y else y, ArrayA)
print(f"{minA} {maxA}")
```

```python
# Using regular function and built-in function
def get_min_max(arr):
    return min(arr), max(arr)


N = int(input())
ArrayA = list(map(int, input().split()))

minA, maxA = get_min_max(ArrayA)
print(f"{minA} {maxA}")
```

#### Solution of the CODEFORCES problem: J. Average
Url: https://codeforces.com/group/MWSDmqGsZm/contest/223205/problem/J

```python
# Using lambda function
N = int(input())
ArrayA = list(map(float, input().split()))

avg = reduce(lambda x, y: x + y, ArrayA) / len(ArrayA)
print(f"{avg:.7f}")
```

```python
# Using regular function and built-in function
def get_average(arr):
    avg = sum(arr) / len(arr)
    return f"{avg:.7f}"


N = int(input())
ArrayA = list(map(float, input().split()))

average = get_average(ArrayA)
print(average)
```
