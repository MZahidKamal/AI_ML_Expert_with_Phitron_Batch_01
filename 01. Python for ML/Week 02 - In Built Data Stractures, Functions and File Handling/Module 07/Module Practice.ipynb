# File Read

```python
file = open("./Sample Data/Sample Text.txt", "r")

content = file.read()
print(type(content))
print(content)

file.close()
print(file.closed)
```

```python
file = open("./Sample Data/Sample Paragraph.txt", "r")

content = file.readlines()  # .readlines() returns a list of lines, whereas .read() returns the entire file as one single string
print(type(content))
print(content)

file.close()
print(file.closed)
```

Python এ file read বা write করার পর close করা খুবই জরুরি।

কারণ file open করলে system একটি resource allocate করে। close না করলে সেই resource release হয় না। এতে memory leak, file lock বা unexpected error হতে পারে।

বিশেষ করে write করলে, close না করলে data পুরোপুরি save নাও হতে পারে, কারণ buffer এ থাকা data disk এ flush নাও হতে পারে।

তাই সবসময় file.close() ব্যবহার করা উচিত, অথবা সবচেয়ে নিরাপদ পদ্ধতি হলো with statement ব্যবহার করা, যেখানে কাজ শেষ হলে automatic close হয়ে যায়।

```python
with open("./Sample Data/Sample Paragraph.txt", "r") as file:
    content = file.readlines()  # .readlines() returns a list of lines, whereas .read() returns the entire file as one single string
    print(type(content))
    print(content)

print(file.closed)  # As soon as we'll come out of the with statement, the fill will automatically be closed
```

```python
with open("./Sample Data/Sample Paragraph.txt", "r") as file:
    for line in file:
        # print(line, end="")
        line = line.strip()  # .strip() removes leading/trailing whitespace (including the ending '\n') for clean printing
        print(line)

print(file.closed)  # As soon as we'll come out of the with statement, the fill will automatically be closed
```

```python
with open("./Sample Data/Sample Paragraph.txt", "r") as file:
    content = file.readlines()
    print(type(content))  # It's a list of strings
    print(content)  # Each string is a line

    content = list(map(str.strip, content))  # cleaning leading/trailing whitespace (including '\n') from each line
    print(content)

    content = list(filter(lambda x: len(x) > 0, content))   # cleaning out empty lines after stripping (i.e., lines that became "")
    print(content)
```

# Create a File and Write

```python
# Creating a new file and writing to it
with open("Sample Data/Sample Text 1.txt", "w") as file:
    file.write("Hello new test file!")
    file.write("\nThis is a new line, in the new test file!")
    file.writelines(["This is a new line, in the new test file!\n", "This is a new line, in the new test file!\n"])

with open("Sample Data/Sample Text 1.txt", "r") as file:
    print(file.read())
```

```python
# Writing in an existing file, with w will overwrite the existing file content
with open("Sample Data/Sample Text 1.txt", "w") as file:
    file.write("\nThis is a new line, in the existing test file!")

with open("Sample Data/Sample Text 1.txt", "r") as file:
    print(file.read())
```

```python
# Writing in an existing file, with a will not overwrite the existing file content, rather will add new text content
with open("Sample Data/Sample Text 1.txt", "a") as file:
    file.write("\nThis is another new line, in the existing test file!")

with open("Sample Data/Sample Text 1.txt", "r") as file:
    print(file.read())
```

```python
strings = ['hello', 'hi', 'good bye', 'what up']
with open("Sample Data/Sample Text 1.txt", "a") as file:
    file.writelines(strings)

with open("Sample Data/Sample Text 1.txt", "r") as file:
    print(file.read())
```

# Cursor / Pointer

```python
with open("Sample Data/Sample Text 1.txt", "r") as file:
    cursor_position_now = file.tell()
    print(cursor_position_now)

    content = file.readlines()
    print(content)

    cursor_position_after_reading = file.tell()
    print(cursor_position_after_reading)

    content_read_again = file.readline()
    print(content_read_again) # It will not get any content to read, just because currently the cursor position is at the end of the content

with open("Sample Data/Sample Text 1.txt", "r") as file:
    cursor_position_now = file.tell()
    print(cursor_position_now)

    content = file.read(10) # Read 10 characters only
    print(content)

    cursor_position_after_reading = file.tell()
    print(cursor_position_after_reading)

    content_read_again = file.readlines()
    print(content_read_again)

    cursor_position_after_reading = file.tell()
    print(cursor_position_after_reading)
```

# Seek

```python
with open("Sample Data/Sample Text 1.txt", "r") as file:
    cursor_position_now = file.tell()
    print(cursor_position_now)

    content = file.readlines()
    print(content)

    cursor_position_after_reading = file.tell()
    print(cursor_position_after_reading)

    file.seek(0)    # This will move the cursor position to the beginning of the file

    content_read_again = file.readlines()
    print(content_read_again) # It will read the content from the beginning

    file.seek(10)   # Moving the cursor position to the 11th character
    cursor_position_now = file.tell()
    print(cursor_position_now)
```

# Practice 1:

Get the text from sample.txt, and count the following
1. Number of lines,
2. Number of words,
3. Number of characters,
4. and save it to the Counter_of_string.txt file.

```python
with open("Sample Data/Sample Text 2.txt", "r") as file:
    line_list = file.readlines()
    # print(line_list)

    total_lines = len(line_list)  # Total number of lines including the empty lines in between paragraphs
    print(f"Total number of lines including empty lines: {total_lines}")

    total_text_lines = len(list(
        filter(lambda line: len(line) > 1, line_list)))  # Total number of lines that contain text without empty lines
    print(f"Total number of text lines without empty lines: {total_text_lines}")

    # total_words = 0
    # for line in line_list:
    #     wordArray = line.strip().split()
    #     if len(wordArray)>0:
    #         total_words += len(wordArray)
    # print(total_words)

    total_words = sum(list(len(line.strip().split()) for line in line_list))
    print(f"Total number of words: {total_words}")

    # total_character = 0
    # for line in line_list:
    #     wordArray = line.strip().split()
    #     if len(wordArray)>0:
    #         for word in wordArray:
    #             total_character += len(word)
    # print(total_character)

    total_character = sum(len(word) for line in line_list for word in line.strip().split())
    print(f"Total number of characters: {total_character}")

with open("Sample Data/Counter_of_string.txt", "w") as file:
    file.write(f"Total number of lines including empty lines: {total_lines}\n")
    file.write(f"Total number of text lines without empty lines: {total_text_lines}\n")
    file.write(f"Total number of words: {total_words}\n")
    file.write(f"Total number of characters: {total_character}\n")

with open("Sample Data/Counter_of_string.txt", "r") as file:
    print(file.read())
```

# Practice 2:

Write a text into a .txt file and read it back. All in one open operation.

```python
newParagraph = """Ultimately, becoming a pioneer requires cultivating deep curiosity and persistence. You must be comfortable failing, debugging complex architectural issues, and constantly re-evaluating your knowledge as new frameworks emerge. True mastery comes from bridging the gap between theoretical research and practical, impactful applications."""

with open("Sample Data/Sample Text 2.txt", "a+") as file:
    current_cursor_position = file.tell()
    print(current_cursor_position)  # For a+ mode, the cursor will always be at the end of the file

    file.seek(0)    # So let's move the cursor position to the beginning of the file'
    current_cursor_position = file.tell()   # Now we'll be able to read the content from the beginning
    print(current_cursor_position)

    old_content = file.read()
    print(old_content)

    current_cursor_position = file.tell()   # After reading the content, the cursor position will be at the end of the file
    print(current_cursor_position)

    file.writelines('\n\n') # We'll add a new paragraph after the old content, that's why two new lines are added
    file.write(newParagraph)

    current_cursor_position = file.tell()   # After writing the new content, the cursor position will be at the end of the new content
    print(current_cursor_position)

    file.seek(0)    # Again, let's move the cursor position to the beginning of the file'
    current_cursor_position = file.tell()   # Now we'll be able to read the content from the beginning again
    print(current_cursor_position)

    new_content = file.read()
    print(new_content)
```

# Exception Handling

Python এ Exception handling ব্যবহার করা হয় program চলার সময় error হলে যেন program হঠাৎ বন্ধ না হয়ে যায়।
1. এখানে try ব্লকে risky code রাখা হয়, আর error হলে except ব্লক সেটা handle করে।
2. finally ব্লক সবসময় execute হয় (error হোক বা না হোক), সাধারণত cleanup কাজের জন্য ব্যবহার করা হয়।
3. else ব্লক তখনই চলে যখন try ব্লকে কোনো error হয় না।
4. Order সবসময় হবে try → except → else → finally, আর Exception handling program কে safe এবং controlled ভাবে error manage করতে সাহায্য করে।

```python
# a = 10/0    # Error: ZeroDivisionError will raise
# print('Then execute this line 1.')  # Therefore, this line will never execute

try:
    a = 10 / 0
    print(a)
    print('Then execute this line 1.')
except ZeroDivisionError as error:
    print(f"Error: {error}, is the main error. Try again with a different value.")
    print('Then execute this line 2.')
else:
    print('No error occurred. Then execute this line (else block).')
finally:
    print('This line will always execute (finally block).')

print('Then execute this line 3.')
```

```python
try:
    a = b
    print(a)
    print('Then execute this line 1.')
except NameError as error:
    print(f"Error: {error}, is the main error. Try again with a different value.")
    print('Then execute this line 2.')
else:
    print('No error occurred. Then execute this line (else block).')
finally:
    print('This line will always execute (finally block).')

print('Then execute this line 3.')
```

```python
try:
    a = b = c
    print(a)
    print('Then execute this line 1.')
except Exception as error:  # If the type of error is not specified, this Exception will catch all errors
    print(f"Error: {error}, is the main error. Try again with a different value.")
    print('Then execute this line 2.')
else:
    print('No error occurred. Then execute this line (else block).')
finally:
    print('This line will always execute (finally block).')

print('Then execute this line 3.')
```

```python
try:
    with open("Sample Data/Sample Text 3.txt", "r") as file:
        content = file.readlines()
        print(content)
except FileNotFoundError as error:
    print(f"Error: {error}, is the main error. Try again with a different file path.")
    print('Then execute this line 2.')
else:
    print('No error occurred. Then execute this line (else block).')
finally:
    print('This line will always execute (finally block).')

print('Then execute this line 3.')
```
