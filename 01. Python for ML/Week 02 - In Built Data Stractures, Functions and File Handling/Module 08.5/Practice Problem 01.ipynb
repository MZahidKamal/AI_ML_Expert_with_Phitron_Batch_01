```python
from abc import ABC, abstractmethod


# Creating a Vehicle class------------------------------------------------------------------
class Vehicle(ABC):
    # Class Attributes----------------------------------------------------------------------
    __type = str
    __name = str
    __speed = int
    __category = str

    # Class Constructor and Instance Attributes---------------------------------------------
    def __init__(self, type, name, speed):

        if type not in ["Car", "Bike"]:
            raise ValueError("Warning: Invalid vehicle type. Please enter either 'Car' or 'Bike'.")
        if name == "":
            raise ValueError("Warning: Name cannot be empty. Please enter a valid value.")
        if isinstance(speed, float):
            raise TypeError("Warning: Speed must be an integer value. No decimal points allowed.")
        if speed < 0:
            raise ValueError("Warning: Speed cannot be negative value. Please enter a positive value.")

        self.__type = type
        self.__name = name
        self.__speed = speed

        if self.__speed > 100:
            self.__category = "High-speed Vehicle"
        else:
            self.__category = "Normal Vehicle"

    # Class Method--------------------------------------------------------------------------
    def get_type(self):
        return self.__type

    def get_name(self):
        return self.__name

    def get_speed(self):
        return self.__speed

    def classify(self):
        return self.__category

    @abstractmethod
    def show_info(self):
        pass


#-------------------------------------------------------------------------------------------

# Creating a Car class inherited from Vehicle class-----------------------------------------
class Car(Vehicle):
    def __init__(self, type, name, speed):
        super().__init__(type, name, speed)

    # Overriding the show_info() method-----------------------------------------------------
    def show_info(self):
        type = self.get_type()
        name = self.get_name()
        speed = self.get_speed()
        category = self.classify()
        print(f"Type: {type}, Name: {name}, Speed: {speed} km/h, Predicted: {category}")


#------------------------------------------------------------------------------------------

# Creating a Bike class inherited from Vehicle class---------------------------------------
class Bike(Vehicle):
    def __init__(self, type, name, speed):
        super().__init__(type, name, speed)

    # Overriding the show_info() method----------------------------------------------------
    def show_info(self):
        type = self.get_type()
        name = self.get_name()
        speed = self.get_speed()
        category = self.classify()
        print(f"Type: {type}, Name: {name}, Speed: {speed} km/h, Predicted: {category}")


#-------------------------------------------------------------------------------------------

# c1 = Car("Car", "BMW", 100)
# c1.show_info()

# b1 = Bike("Bike", "Honda", 100)
# b1.show_info()

#-------------------------------------------------------------------------------------------


carBikeCollection = []


testCasesN = int(input())
testCasesValues = input().split('\n')


try:
    for line in range(testCasesN):
        type, name, speed = testCasesValues[line].split()
        if type == "Car":
            carBikeCollection.append(Car(type, name, int(speed)))
        else:
            carBikeCollection.append(Bike(type, name, int(speed)))
        print(f"{type} Added: {name}")
except IndexError:
    print(f"Expected {testCasesN} values, got {line}.")


print("\n--- Vehicle Details ---")
for vehicle in carBikeCollection:
    vehicle.show_info()
```
