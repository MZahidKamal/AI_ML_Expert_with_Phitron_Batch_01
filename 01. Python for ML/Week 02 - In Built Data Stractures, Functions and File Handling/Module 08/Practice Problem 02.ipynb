```python
# Creating the class
class MobilePhone:
    def __init__(self, model, battery=100):
        self.model = model
        self.__battery = battery

    def get_battery(self):
        print(f"Battery remaining: {self.__battery}%")

    def use(self, duration):
        self.__battery -= duration * 0.15
        if self.__battery < 0:
            self.__battery = 0
        print(f"Phone used for {duration} minutes. Now battery remaining: {self.__battery}%.")

    def charge(self, duration):
        self.__battery += duration * 2.5
        if self.__battery > 100:
            self.__battery = 100
        print(f"Phone charged for {duration} minutes. Now battery remaining: {self.__battery}%.")


myPhone = MobilePhone('Samsung Galaxy S21')
print(myPhone.model)

# Accessing the private attribute is not possible; therefore, it will raise an error
# print(myPhone.__battery)

# Accessing the private attribute through a method
myPhone.get_battery()

# Using the phone for 60 minutes and checking the battery remaining
myPhone.use(60)

# Charging the phone for 15 minutes and checking the battery remaining
myPhone.charge(5)
```
