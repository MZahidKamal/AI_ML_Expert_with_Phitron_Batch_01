## Python এ Class এবং Object

### 🔹 Class কী?

**Class** হলো একটি নকশা (blueprint) বা template, যার মাধ্যমে আমরা একই ধরনের অনেক **Object** তৈরি করতে পারি।
মানে, Class বলে দেয় একটি জিনিসের কী কী বৈশিষ্ট্য (attribute) এবং কাজ (method) থাকবে।

উদাহরণ চিন্তা করো:
Car হলো একটি **Class** — আর BMW, Toyota, Tesla হলো সেই Class থেকে তৈরি করা আলাদা আলাদা **Object**।

---

### 🔹 Object কী?

**Object** হলো Class থেকে তৈরি বাস্তব instance।
Class হচ্ছে ধারণা, Object হচ্ছে বাস্তব রূপ।

একই Class থেকে অনেক Object তৈরি করা যায়, কিন্তু প্রত্যেক Object এর আলাদা আলাদা data থাকতে পারে।

---

### 🔹 Attribute কী?

**Attribute** হলো Object এর বৈশিষ্ট্য বা data।
যেমন:

* name
* age
* color
* balance

এগুলো Object এর ভিতরে সংরক্ষিত থাকে।

---

### 🔹 Method কী?

**Method** হলো Class এর ভিতরে লেখা function, যা Object এর আচরণ নির্ধারণ করে।
মানে Object কী কাজ করতে পারবে, সেটা Method বলে দেয়।

---

### 🔹 Class এর গঠন

একটি Class সাধারণত থাকে:

* Class নাম
* Constructor (**init**)
* Attribute
* Method

---

### 🔹 গুরুত্বপূর্ণ বিষয়

* Class হচ্ছে blueprint

## Python এ Inheritance

### 🔹 Inheritance কী?

**Inheritance** মানে হলো একটি Class (Child/Subclass) অন্য একটি Class (Parent/Superclass) থেকে feature (attribute + method) উত্তরাধিকার হিসেবে নিয়ে ব্যবহার করা।
এতে একই code বারবার না লিখে reuse করা যায় এবং program structure clean থাকে।

---

### 🔹 কেন Inheritance ব্যবহার করা হয়?

* **Code reuse** (একই জিনিস বারবার লিখতে হয় না)
* **Extend** করা যায় (Parent এর feature রেখে নতুন feature যোগ করা যায়)
* **Maintainability** বাড়ে (এক জায়গায় change করলে সব জায়গায় effect পড়ে)
* Real-world relationship model করা যায় (যেমন Animal → Cat/Dog)

---

## Types of Inheritance (Inheritance এর ধরন)

### 1) Single Inheritance

একটি Child Class শুধু একটি Parent Class থেকে inherit করে।
উদাহরণ ধারণা:
Parent: Vehicle → Child: Car

---

### 2) Multilevel Inheritance

একটা chain তৈরি হয়: Parent → Child → Grandchild
উদাহরণ ধারণা:
Animal → Mammal → Dog

---

### 3) Hierarchical Inheritance

একটি Parent Class থেকে একাধিক Child Class inherit করে।
উদাহরণ ধারণা:
Animal → Dog, Cat, Bird

---

### 4) Multiple Inheritance

একটি Child Class একাধিক Parent Class থেকে inherit করে।
উদাহরণ ধারণা:
Student + Athlete → SportsStudent
(এখানে Method Resolution Order (MRO) গুরুত্বপূর্ণ)

---

### 5) Hybrid Inheritance

Single + Multiple + Multilevel ইত্যাদি mix হয়ে complex structure তৈরি হলে তাকে Hybrid বলে।
উদাহরণ ধারণা:
বিভিন্ন inheritance type combine করে তৈরি করা architecture

---

## 🔹 গুরুত্বপূর্ণ ধারণা

* Child Class Parent এর method override করতে পারে (Method Overriding)
* Parent এর constructor/method call করতে `super()` ব্যবহার করা হয়
* Multiple Inheritance এ conflict avoid করতে `MRO` (Method Resolution Order) follow হয়

---

### ✅ ছোট সারাংশ

Inheritance = Parent থেকে Child এ feature transfer
Types = Single, Multilevel, Hierarchical, Multiple, Hybrid
Goal = code reuse + extend + clean design


# Single Inheritance

````python
class Phone:
    category = 'Electronics'

    # Constructor
    def __init__(self, brand, model, battery_capacity, camera_resolution):
        self.brand = brand
        self.model = model
        self.battery_capacity = battery_capacity
        self.camera_resolution = camera_resolution
        self.remaining_charge = 100

    # Method
    def display_details(self):
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Battery Capacity: {self.battery_capacity}")
        print(f"Camera Resolution: {self.camera_resolution}")

    def charge_battery(self, remaining_charge, minute_of_charge):
        self.remaining_charge = remaining_charge
        print(f"Remaining charge is {self.remaining_charge}%")
        print(f"Battery charged for {minute_of_charge} minutes.")

        if self.remaining_charge == 100:
            print("Battery is fully charged.")
        else:
            self.remaining_charge += minute_of_charge * 2.5

        print(f"Now Remaining charge is {self.remaining_charge}%")


# Inheritance
class Smartphone(Phone):

    # Constructor of Child Class
    def __init__(self, brand, model, battery_capacity, camera_resolution, os, ram):
        super().__init__(brand, model, battery_capacity, camera_resolution) # super() is called to call the parent constructor
        self.os = os
        self.ram = ram

    # New Method of Child Class
    def browse_internet(self):
        print(f"Browsing internet using {self.os}...")

    # Method Overriding from Parent Class
    def display_details(self):
        super().display_details()
        print(f"Operating System: {self.os}")
        print(f"RAM: {self.ram}")

my_phone = Smartphone('Apple', 'iPhone 12', '4000 mAh', '12 MP', 'iOS', '8 GB')
my_phone.display_details()
my_phone.browse_internet()
````

# 📝 সংক্ষিপ্ত ব্যাখ্যা (Single Inheritance)

* এখানে **Smartphone** class, **Phone** class থেকে inherit করেছে।
* `super().__init__()` দিয়ে Parent class এর constructor call করা হয়েছে।
* Child class Parent এর সব attribute ও method ব্যবহার করতে পারছে।
* Child class নতুন method (`browse_internet`) যোগ করেছে।
* Child class `display_details()` override করেছে (Method Overriding)।

### মূল ধারণা:

Single Inheritance = ১টি Child class ← ১টি Parent class
Code reuse + extend করার জন্য ব্যবহৃত হয়।


# Multiple Inheritance

````python
class Phone:
    category = 'Electronics'

    # Constructor
    def __init__(self, brand, model, battery_capacity, camera_resolution):
        self.brand = brand
        self.model = model
        self.battery_capacity = battery_capacity
        self.camera_resolution = camera_resolution
        self.remaining_charge = 100

    # Method
    def display_details(self):
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Battery Capacity: {self.battery_capacity}")
        print(f"Camera Resolution: {self.camera_resolution}")

    def charge_battery(self, remaining_charge, minute_of_charge):
        self.remaining_charge = remaining_charge
        print(f"Remaining charge is {self.remaining_charge}%")
        print(f"Battery charged for {minute_of_charge} minutes.")

        if self.remaining_charge == 100:
            print("Battery is fully charged.")
        else:
            self.remaining_charge += minute_of_charge * 3.5

        print(f"Now Remaining charge is {self.remaining_charge}%")










# Inheritance
class Smartphone(Phone):

    # Constructor of Child Class
    def __init__(self, brand, model, battery_capacity, camera_resolution, os, ram):
        super().__init__(brand, model, battery_capacity, camera_resolution)  # super() is called to call the parent constructor
        self.os = os
        self.ram = ram

    # New Method of Child Class
    def browse_internet(self):
        print(f"Browsing internet using {self.os}...")

    # Method Overriding from Parent Class
    def display_details(self):
        super().display_details()
        print(f"Operating System: {self.os}")
        print(f"RAM: {self.ram}")










# 🔹 New Class (Second Parent)
class CameraDevice:
    def __init__(self, lens_type):
        self.lens_type = lens_type

    # New Method of this second parent class
    def capture_photo(self):
        print(f"Capturing photo using {self.lens_type} lens.")










# 🔹 Multiple Inheritance, Child Class = Parent Class + Second Parent Class
class SmartDevice(Smartphone, CameraDevice):

    # Constructor of Child Class, combining the constructors of Parent Classes
    def __init__(self, brand, model, battery_capacity, camera_resolution, os, ram, lens_type):
        Smartphone.__init__(self, brand, model, battery_capacity, camera_resolution, os, ram)
        CameraDevice.__init__(self, lens_type)

    # Inherited Method of Parent Class
    def display_details(self):
        Smartphone.display_details(self)

    # New Method of Child Class
    def make_call(self, number):
        print(f"Calling {number}...")

    # New Method of Child Class
    def device_info(self):
        print("This is a Multi-Functional Smart Device.")










# Creating an object of the class SmartDevice
device = SmartDevice('Samsung', 'Galaxy Ultra', '5000 mAh', '108 MP', 'Android', '12 GB', 'Wide Angle')

# Calling methods of the SmartDevice object, which are inherited from both Parent Classes
device.display_details()
device.charge_battery(5, 20)
device.make_call("017XXXXXXXX")
device.browse_internet()
device.capture_photo()
device.device_info()
````

# 📝 Multiple Inheritance ব্যাখ্যা

* **SmartDevice** দুইটি Parent class থেকে inherit করেছে → **Smartphone** এবং **CameraDevice**
* অর্থাৎ একটি Child class একাধিক Parent class ব্যবহার করছে
* Parent-1 থেকে এসেছে: call, internet, OS, RAM
* Parent-2 থেকে এসেছে: camera feature
* Constructor দুটো আলাদাভাবে call করা হয়েছে

### মূল ধারণা

Multiple Inheritance = ১টি Child class ← একাধিক Parent class

এভাবে একটি object একসাথে একাধিক class এর feature ব্যবহার করতে পারে।
Code reuse + feature combination এর জন্য এটি ব্যবহার করা হয়।


# Polymorphism

# 🔷 Polymorphism in Python

**Polymorphism** মানে — একই method নাম, কিন্তু ভিন্ন behavior।
অর্থাৎ একই interface ব্যবহার করে আলাদা class আলাদা কাজ করতে পারে।

````python
class Phone:
    def make_sound(self):
        print("Phone is ringing...")


class Smartphone(Phone):
    def make_sound(self):
        print("Smartphone is playing a custom ringtone...")


# Creating objects of the classes Phone and Smartphone
nokia = Phone()
samsung = Smartphone()


# Polymorphism in action
nokia.make_sound()
samsung.make_sound()
````

## 🔹 এখানে কী হলো?
* দুইটা class এ একই method নাম → `make_sound()`
* কিন্তু behavior আলাদা
* সাধারণ Nokia → default ringtone বাজায়
* Samsung Smartphone → custom ringtone বাজায়
* কিন্তু user শুধু `make_sound()` call করে। কোন device সেটা Python নিজে বুঝে নেয়।
* এটাই **Runtime Polymorphism**।

---

# 🧠 গুরুত্বপূর্ণ ধারণা

### ✔ Method Overriding
Child class Parent class এর method কে নতুনভাবে define করলে তাকে **method overriding** বলে।

### ✔ Dynamic Binding
Runtime এ Python ঠিক করে কোন class এর method execute হবে।

---

# 📌 সংক্ষেপে
Polymorphism =
একই method নাম
ভিন্ন class
ভিন্ন behavior
Code flexible এবং scalable করার জন্য Polymorphism খুব গুরুত্বপূর্ণ।

# Encapsulation

**Encapsulation** হলো OOP এর একটি গুরুত্বপূর্ণ ধারণা, যেখানে **data (attribute)** এবং **behavior (method)** একসাথে class এর ভিতরে রাখা হয় এবং কিছু অংশ বাইরে থেকে সরাসরি access করা সীমাবদ্ধ করা হয়। এতে data সুরক্ষিত থাকে এবং controlled access নিশ্চিত করা যায়।

````python
class Phone:
    category = 'Electronics'

    # Constructor
    def __init__(self, brand, model, battery_capacity, camera_resolution):
        self.brand = brand
        self.model = model
        self.battery_capacity = battery_capacity
        self.camera_resolution = camera_resolution
        self.remaining_charge = 100
        self.__imei = 1234567891011121314151617181920

    # Method
    def display_details(self):
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Battery Capacity: {self.battery_capacity}")
        print(f"Camera Resolution: {self.camera_resolution}")

    def get_imei(self):
        print(f"IMEI: {self.__imei}")

    def __change_imei(self, new_imei):
        self.__imei = new_imei

    def update_imei(self, new_imei):
        self.__change_imei(new_imei)



# New object of the class Phone is created
apple = Phone('Apple', 'iPhone 12', '4000 mAh', '12 MP')

# Accessing attributes and methods of the object
apple.display_details()

# Modifying attributes values directly
apple.remaining_charge = 50
print(apple.remaining_charge)

# Accessing an encapsulated private attribute value directly using double underscore
# print(apple.__imei)
# It will raise an error, because this attribute is private

# But we can access it using the getter method
print(apple.get_imei())

# Accessing an encapsulated private method directly using double underscore
# apple.__change_imei(9876543210987654321)
# print(apple.get_imei())
# It will raise an error, because this method is private

# The encapsulated private method can only be accessible inside the class
apple.update_imei(9876543210987654321)
print(apple.get_imei())
````

## 🔹 Attribute Encapsulation

```python
self.__imei = 1234567891011121314151617181920
```

এখানে `__imei` একটি **private attribute**।

👉 বাইরে থেকে সরাসরি:

```
apple.__imei   ❌ (Error)
```

👉 Access করতে হলে:

```
apple.get_imei()   ✅
```

এখানে `get_imei()` হলো একটি **getter method**, যা private data কে controlled way তে দেখায়।

---

## 🔹 Method Encapsulation

```python
def __change_imei(self, new_imei):
```

এটি একটি **private method**।

👉 বাইরে থেকে সরাসরি:

```
apple.__change_imei(...)  ❌ (Error)
```

👉 কিন্তু class এর ভিতরে:

```
update_imei() → __change_imei()
```

এভাবে private method কে indirect ভাবে call করা হয়েছে।

এটাই method encapsulation।

---

## 🔹 কেন Encapsulation ব্যবহার করা হয়?

✔ Sensitive data protect করতে (যেমন: IMEI)
✔ Direct modification আটকাতে
✔ Controlled access দিতে
✔ Code maintainable ও secure করতে

---

## 🔹 গুরুত্বপূর্ণ ধারণা

Python এ `__variable` বা `__method` আসলে **name mangling** ব্যবহার করে কাজ করে।
মানে Python internally নাম পরিবর্তন করে দেয়, তাই বাইরে থেকে সরাসরি access করা যায় না।

# Python এ Abstraction

### 🔹 Abstraction কী?

**Abstraction** মানে হলো — user কে শুধু দরকারি জিনিসটা দেখানো, আর ভিতরের জটিল implementation লুকিয়ে রাখা।
এক কথায়: “কি করবে” দেখাবে, “কিভাবে করবে” সেটা hide করবে।

Real life উদাহরণ:
তুমি গাড়ি চালাও — steering, brake, accelerator ব্যবহার করো। কিন্তু engine কিভাবে কাজ করে সেটা না জেনেও গাড়ি চালাতে পারো। এটিই abstraction।

## 🔹 Python এ Abstraction কিভাবে করা হয়?

Python এ abstraction করার standard উপায় হলো:

✅ `abc` module ব্যবহার করে

* `ABC` (Abstract Base Class)
* `@abstractmethod` decorator

Abstract class এ কিছু method থাকবে যেগুলোর body থাকবে না (বা empty থাকবে)।
Child class কে বাধ্যতামূলকভাবে সেগুলো implement করতে হবে।

````python
from abc import ABC, abstractmethod

class Phone(ABC):
    def __init__(self, brand):
        self.brand = brand

    # abstract method (must be implemented)
    @abstractmethod
    def make_call(self, number):
        pass

    # abstract method (must be implemented)
    @abstractmethod
    def charge_battery(self, minutes):
        pass

    # concrete method (shared for all), can be used directly in child class, or can be overridden in child class
    def show_brand(self):
        print(f"Brand: {self.brand}")





class Smartphone(Phone):
    def __init__(self, brand, os):
        super().__init__(brand)
        self.os = os
        self.charge = 50

    def make_call(self, number):
        print(f"{self.brand} Smartphone calling {number} using VoLTE...")

    def charge_battery(self, minutes):
        self.charge = min(100, self.charge + minutes * 2)
        print(f"Charged to: {self.charge}%")

    def browse_internet(self):
        print(f"Browsing internet on {self.os}...")





class FeaturePhone(Phone):
    def __init__(self, brand):
        super().__init__(brand)
        self.charge = 30

    def make_call(self, number):
        print(f"{self.brand} Feature phone calling {number}...")

    def charge_battery(self, minutes):
        self.charge = min(100, self.charge + minutes * 1)
        print(f"Charged to: {self.charge}%")





phone_01 = Smartphone("Samsung", "Android")
phone_01.show_brand()
phone_01.make_call("017XXXXXXXX")
phone_01.charge_battery(10)
phone_01.browse_internet()
print('-----')
phone_02 = FeaturePhone("Nokia")
phone_02.show_brand()
phone_02.make_call("018XXXXXXXX")
phone_02.charge_battery(10)
````
