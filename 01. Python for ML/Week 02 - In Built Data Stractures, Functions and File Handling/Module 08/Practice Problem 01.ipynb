```python
# Create 1rst class, without any method
class Phone:
    def __init__(self, model, battery = 100):
        self.model = model
        self.battery = battery





# Creating the 2nd class but inheriting from the 1rst class, and also no method
class SmartPhone(Phone):
    def __init__(self, model, os, battery = 100):   # Non-default arguments take place before arguments with default values
        super().__init__(model, battery)
        self.os = os





# Creating the 3rd class but inheriting from the 2nd class, and this time will create a method
class GamingPhone(SmartPhone):
    def __init__(self, model, os, cooling_system, battery = 100):
        super().__init__(model, os, battery)
        self.cooling_system = cooling_system

    def star_game (self, game_name):
        print(f"Playing {game_name} on {self.model}")





myPhone = GamingPhone('Motorola', 'Old OS', 'Fan')
print(myPhone.model)
myPhone.star_game('Sudoku')
```
