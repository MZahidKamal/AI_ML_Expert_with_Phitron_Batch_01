#### Solution of the HACKERRANK problem: Sensor Calibrator — Min-Max Scaling
Url: https://www.hackerrank.com/contests/ai-ml-module-4-practice-day-1/challenges/b3-sensor-calibrator-min-max-scaling

```python
inputs = list(map(float, input().split()))
x = float(inputs[0])

min_v = float(inputs[1])
max_v = float(inputs[2])
norm = (x - min_v) / (max_v - min_v)

print(f"{norm:.2f}")
```

#### Solution of the HACKERRANK problem: Mini Loss Tracker
Url: https://www.hackerrank.com/contests/ai-ml-module-4-practice-day-1/challenges/b4-mini-loss-tracker-average-monitor

```python
testCases = int(input())
lossThreshold = float(0)
totalLoss = float(0)

for i in range(testCases + 1):
    if i == 0:
        lossThreshold = float(input())
    else:
        loss = float(input())
        totalLoss += loss

lossTracker = "PASS" if (totalLoss / testCases) <= lossThreshold else "RETRY"
print(lossTracker)
```

#### Solution of the HACKERRANK problem: Majority Vote Gate
Url: https://www.hackerrank.com/contests/ai-ml-module-4-practice-day-1/challenges/b5-majority-vote-gate-simple-ensemble

```python
testCases = int(input())
yesVoteCount = int(0)
noVoteCount = int(0)

for i in range(testCases):
    vote = input()
    yesVoteCount += 1 if vote == "YES" else 0
    noVoteCount += 1 if vote == "NO" else 0

voteGate = "ACCEPT" if yesVoteCount >= noVoteCount else "REJECT"
print(voteGate)
```

#### Solution of the HACKERRANK problem: AI Mood Detector
Url: https://www.hackerrank.com/contests/ai-ml-module-4-practice-day-1/challenges/ai-mood-detector

```python
happyMoodWords = ["happy", "joy", "smile"]
sadMoodWords = ["sad", "cry", "angry"]
usersMode = "Neutral Mood"

usersMessage = str(input()).lower().split()

for word in usersMessage:
    if word in happyMoodWords:
        usersMode = "Happy Mood"
        break
    elif word in sadMoodWords:
        usersMode = "Sad Mood"
        break

print(usersMode)
```

#### Solution of the HACKERRANK problem: AI Brightness Analyzer
Url: https://www.hackerrank.com/contests/ai-ml-module-4-practice-day-1/challenges/p2-a2-ai-brightness-analyzer

```python
averageBrightness = str()

pixelBrightnessList = list(map(int, input().split()))

averagePixelBrightness = sum(pixelBrightnessList) / len(pixelBrightnessList)

if averagePixelBrightness < 85:
    averageBrightness = "Dark Image"
elif averagePixelBrightness <= 170:
    averageBrightness = "Normal Image"
else:
    averageBrightness = "Bright Image"

print(averageBrightness)
```

#### Solution of the HACKERRANK problem: AI Bias Detector
Url: https://www.hackerrank.com/contests/ai-ml-module-4-practice-day-1/challenges/p2-a3-ai-bias-detector

```python
predictions = list(map(str, input().split()))

countA = predictions.count("A")
countB = predictions.count("B")

presenceOfA = (countA / len(predictions)) * 100
presenceOfB = (countB / len(predictions)) * 100

model = str()

if presenceOfA > 70:
    model = "Biased Model"
elif presenceOfB > 70:
    model = "Biased Model"
else:
    model = "Fair Model"

print(model)
```

#### Solution of the HACKERRANK problem: AI Keyword Counter
Url: https://www.hackerrank.com/contests/ai-ml-module-4-practice-day-1/challenges/p2-a4-ai-keyword-counter

```python
aiRelatedKeywords = ['ai', 'data', 'model', 'learn', 'train', 'neural']
keywordDetectionCount = 0

usersMessage = str(input()).lower().split()

for word in aiRelatedKeywords:
    if word in usersMessage:
        keywordDetectionCount += 1

if keywordDetectionCount >= 2:
    print("AI Detected")
else:
    print("Not AI Related")
```
