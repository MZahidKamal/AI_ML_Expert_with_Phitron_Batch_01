# Tuples

tuple দরকার হয় যখন এমন data রাখতে চাও যেটা পরিবর্তন হবে না (immutable)।
list পরিবর্তনযোগ্য (mutable), কিন্তু tuple পরিবর্তনযোগ্য না — তাই data safe থাকে।

Real life example:
একজন মানুষের জন্মতারিখ (দিন, মাস, বছর) একবার সেট করলে আর বদলানো উচিত না। তাই এটা tuple দিয়ে রাখা ভালো।
উদাহরণ:
birth_date = (15, 12, 1998)
এখানে tuple ব্যবহার করলে ভুল করে value পরিবর্তন হওয়ার ঝুঁকি কমে যায়।

সংক্ষেপে:
Fixed data, constant data, coordinate (x, y), RGB color, database record — এসব ক্ষেত্রে tuple বেশি ব্যবহার হয়।


```python
# declaration
tup = (10, 20, 30, 40)

float_tup = (10.5, 10.2)
mixed_tup = (10, 10.5, "str", True)

lst = [10, 20, 30]
tup1 = tuple(lst)

print(type(tup1), tup1)
```

```python
# Access
tup = (10, 20, 30, 10.5, "str", False)
print(tup)

#index
print(tup[3])

# slicing
new_tup = tup[0:3]
print(new_tup)
```

```python
# mutable
listA = [10, 20, 30, 40]
listA.append(60)
print(listA)

listA[1] = 100
print(listA)

# immutable
tup = (10, 20, 30, 40)
# tup[1] = 100 # Will raise an error, because tuples are immutable
print(tup)
```

```python
# tuple method
tup = (10, 20, 30, 40, 10, 20, 30, 10)

print(tup.count(10))
print(tup.index(40))
```

# Set
set হলো এমন একটি data type যেখানে duplicate value রাখা যায় না।
set এর ভিতরে সব element unique হয় এবং এর কোনো index থাকে না।

এটা সাধারণত ব্যবহার করা হয় duplicate remove করার জন্য বা membership test (in, not in) দ্রুত করার জন্য।
উদাহরণ:
nums = {1, 2, 2, 3}
Result হবে: {1, 2, 3}

সংক্ষেপে:
Unique data store, duplicate remove, এবং fast lookup এর জন্য set ব্যবহার করা হয়।

```python
A = {1, 2, 3}

print(type(A), A)
```

```python
## empty structures
B = []
print(type(B))

C = ()
print(type(C))

S = set()  # Set is basically declared with curly braces {}, but for empty set we can use set() function
print(type(S))
```

```python
#accessing Set
S = {1, 2, 3, 4, 1, 2}

if 3 in S:
    print("3 is present")
else:
    print("not present")

summ = 0
for element in S:
    summ += element

print(summ)
```

```python
#mutable Set
s = {5, 1, 2, 4, 3}

s.add(10)
s.add(0)
print(s)

s.pop()
s.pop()
print(s)

s.remove(5)
print(s)
```

# Set Mathematical Methods

union() দুইটা set একসাথে মিলিয়ে সব unique element দেয়।

intersection() দুইটা set এর common element গুলো দেয়।

isdisjoint() চেক করে দুইটা set এর মধ্যে কোনো common element আছে কি না। থাকলে False, না থাকলে True।

issubset() চেক করে একটি set অন্য set এর ভিতরে পুরোপুরি আছে কি না।

```python
a = {1, 2, 3}
b = {1, 2, 5, 6, 4}

print(a.union(b))

print(a.intersection(b))

print(a.isdisjoint(b))

print(a.issubset(b))
```

# Dictionaries in Python

dictionary হলো Python এর একটি mapping data structure যেখানে data রাখা হয় key-value pair আকারে।
প্রতিটি key unique হতে হবে, কিন্তু value duplicate হতে পারে।
dictionary mutable, মানে পরে নতুন key যোগ করা, value পরিবর্তন করা বা delete করা যায়।
Access করা হয় key দিয়ে, index দিয়ে না।

সংক্ষেপে:
dictionary = key দিয়ে দ্রুত data খোঁজা ও সংরক্ষণ করার উপায়।


```python
# empty
dic = {}
print(type(dic))

dic = {
    "name": "Phitron",
    "age": 20,
    "address": "Dhaka",
    "numbers": [10, 20, 30]
}

print(type(dic), dic)
```

```python
# access
# dic_name[key]

print(dic["numbers"])
print(dic.get("age"))
dic["name"] = "Phitron AI/ML"

dic = {
    "name": "Phitron",
    "age": 20,
    "address": "Dhaka",
    "numbers": [10, 20, 30],
    "age": 30  # Duplicate key will override the previous value
}
print(dic)
```

```python
#printing default values
dic = {
    "name": "Phitron",
    "age": 30,
    "address": "Dhaka",
    "numbers": [10, 20, 30],
}
print(dic)

print(dic.get("math_marks"))
print(dic.get("math_marks", 0))  # Find math_marks key in dictionary, if not present return 0, but don't raise an error
print(dic.get("math_marks",
              "Not Found"))  # Find math_marks key in dictionary, if not present return "Not Found", but don't raise an error
```

```python
# adding a value

dic = {
    "name": "Phitron",
    "age": 30,
    "address": "Dhaka",
    "numbers": [10, 20, 30],
}

print(dic)
dic["math_marks"] = 30
print(dic)
dic.update({"english_marks": 40})

print(dic)
```

```python
# deleting
del dic["math_marks"]
print(dic)
```

```python
dic_2 = dic.copy()
print(dic_2)
```

Dictionary এর key অবশ্যই hashable এবং immutable হতে হবে।

যে ধরনের key ব্যবহার করা যাবে:
int, float, str, bool, tuple (যদি tuple এর ভিতরের সব element immutable হয়)।

যে ধরনের key ব্যবহার করা যাবে না:
list, dict, set — কারণ এগুলো mutable, তাই hashable না।

একই key দুইবার দিলে আগের value overwrite হয়ে যাবে।

সংক্ষেপে:
key = immutable + hashable হতে হবে, mutable data type key হিসেবে ব্যবহার করা যাবে না।

```python
# testing
dic_3 = {
    {"name": "adil", "age": 20}: 20
}  # It's not possible to use a dictionary as a key inside another dictionary
print(dic_3)
```

# Iterating over Dictionary

```python
dic = {
    "name": "Phitron",
    "age": 30,
    "address": "Dhaka",
    "numbers": [10, 20, 30],
}

keys = dic.keys()
print(keys)

dic["math_numbers"] = 30
print(keys)

values = dic.values()
print(values)

dic["math_numbers"] = 40
print(values)

items = dic.items()
print(items)
```

```python
# iteration
for key, value in dic.items():
    print(key, value)
```

# Dictionary Comprehension

```python
square = {x: x ** 2 for x in range(1, 11) if x % 2 == 0}

print(square)
```

```python
Locations = ["dhaka", "chattogram", "sylhet"]
Co_ordinates = [(10, 10.5), (20.5, 192), (101, 102)]  # It's a list of tuples

exact_location = {eachLocation: eachCoordinate for eachLocation, eachCoordinate in zip(Locations, Co_ordinates)}
print(exact_location)
```

### 2. Given a string , print the frequency of the words

```python
string = """In 2026, the AI-ML landscape is rapidly shifting from passive chatbots to Agentic AI. Unlike traditional systems that merely answer queries, these AI agents autonomously execute complex, multi-step workflows by utilizing external tools without constant human oversight. Alongside this, there is a surge in Domain-Specific Models and Small Language Models (SLMs), which offer faster, cost-effective, and highly accurate solutions for specialized tasks compared to massive general-purpose models. This evolution is transforming AI from a mere conversational tool into a proactive digital partner, fundamentally redefining efficiency across industries and paving the way for truly autonomous digital workforce integration."""

words = string.split()
print(len(words))

count = {}
for word in words:
    count[word] = count.get(word, 0) + 1
print(count)

for key, value in count.items():
    print(f"count of {key} is {value}")
```
