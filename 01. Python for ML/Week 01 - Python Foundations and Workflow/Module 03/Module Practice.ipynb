# String Methods

```python
string  = "Welcome to Phitron AI-ML Course"

# Finding length
print(len(string))

# Converting to lowercase
processed_string = string.lower()
print(processed_string)

# Finding the existence of substring
print("phitron" in processed_string)

# Finding the starting index of the substring
starting_index = processed_string.find("phitron")
print(starting_index)

# Finding the starting index of the substring from the end
starting_index_backward = processed_string.rfind("phitron")

# Finding how many times a substring occurs
substring_count = processed_string.count("phitron")
print(substring_count)

## Replacing the substring
processed_string = string.replace("AI-ML", "AI/ML")
print(processed_string)
print(string)   # String is immutable
```

# List as Stack

```python
# Declaration of an empty stack
stack = []  # Stack is a LIFO data structure

# Putting a new element on top of the stack
stack.append(1)
stack.append(2)
stack.append(3)
stack.append(4)
stack.append(5)

# Finding the top element of the stack
print(f"Top element of the stack: {stack[-1]}")

# Removing the top element of the stack
stack.pop()

# Then again, finding the top element of the stack
print(f"Top element of the stack: {stack[-1]}")
```

# List as Queue

```python
# Declaration of an empty queue
queue = []  # Queue is a FIFO data structure

# Inserting a new element at the end of the queue
queue.append(1)
queue.append(2)
queue.append(3)
queue.append(4)
queue.append(5)

# Removing the last element of the queue
print(queue.pop())

# Accessing the front element of the queue
print(f"Front element of the queue: {queue[0]}")

# Removing the front element of the queue
queue.pop(0)

# Then again, accessing the front element of the queue
print(f"Front element of the queue: {queue[0]}")
```

# List Comprehension

```python
even = []
# for i in range(101):
#     if i % 2 == 0:
#         even.append(i)

# Doing the same operation using the list comprehension
even = [i for i in range(101) if i % 2 == 0]
print(even)

# Finding all odd numbers and adding 10 to them
all_odds_plus_ten = [x + 10 for x in range(1, 101) if x % 2 != 0]
print(all_odds_plus_ten)
```

#### Solution of the CODEFORCES problem: M. Replace MinMax
Url: https://codeforces.com/group/MWSDmqGsZm/contest/219774/problem/M

```python
inputN = input()
N = int(inputN)

inputNumbers = input()
Numbers = list(map(int, inputNumbers.split()))

highestNumber = max(Numbers)
highestNumberPosition = Numbers.index(highestNumber)

lowestNumber = min(Numbers)
lowestNumberPosition = Numbers.index(lowestNumber)

# print(highestNumber, highestNumberPosition)
# print(lowestNumber, lowestNumberPosition)

Numbers[highestNumberPosition] = lowestNumber
Numbers[lowestNumberPosition] = highestNumber

for i in Numbers:
    print(i, end=" ")
```

#### Solution of the CODEFORCES problem: K. Sum Digits
Url: https://codeforces.com/group/MWSDmqGsZm/contest/219774/problem/K

```python
inputN = input()
N = int(inputN)

inputNumber = input()
NumbersList = list(inputNumber)

summ = 0
for number in NumbersList:
    summ += int(number)

print(summ)
```

#### Solution of the CODEFORCES problem: H. Good or Bad
Url: https://codeforces.com/group/MWSDmqGsZm/contest/219856/problem/H

```python
Good1 = str('010')
Good2 = str('101')

numberOfTestCases = int(input())

for testCase in range(numberOfTestCases):
    inputString = input()
    if Good1 in inputString or Good2 in inputString:
        print('Good')
    else:
        print('Bad')
```

```python
Good1 = str('010')
Good2 = str('101')

numberOfTestCases = int(input())
inputStringsList = list(input().split(' '))

for inputString in inputStringsList:
    if Good1 in inputString or Good2 in inputString:
        print('Good')
    else:
        print('Bad')
```
