# Operators in Python

```python
x = 10 ; y = 3

summ = x + y # summation
sub = x - y # subtraction
mult = x * y # multiplication
div = round(x / y , 2) # division

rem = x % y # remainder

print(summ , sub , mult, div , rem)
```

```python
x = 10 ; y = 3

greater_than = x > y
greater_than_equal = (10 >= 10)
print(greater_than , greater_than_equal)

less_than = x < y
less_than_equal = (10 <= 10)
print(less_than , less_than_equal)

equal =  x==y
print(equal)

not_equal = x != y
print(not_equal)
```

# Operator Precedence Order (Highest → Lowest)

1. () → Parentheses
2. ** → Exponentiation
3. +x, -x, ~x → Unary plus, unary minus, bitwise NOT
4. *, /, //, % → Multiplication, division, floor division, modulus
5. +, - → Addition, subtraction
6. <<, >> → Bitwise left shift, right shift
7. & → Bitwise AND
8. ^ → Bitwise XOR
9. | → Bitwise OR
10. ==, !=, >, >=, <, <=, is, is not, in, not in → Comparison, identity, membership
11. not → Logical NOT
12. and → Logical AND
13. or → Logical OR

Python এ operator precedence মানে কোন operator আগে evaluate হবে। সবচেয়ে আগে () অর্থাৎ Parentheses execute হয়, তারপর ** (Exponentiation), তারপর unary operator যেমন +x, -x, ~x। এরপর multiplication, division, addition ইত্যাদি arithmetic operator। তারপরে bitwise operator গুলো, তারপর comparison operator (==, !=, <, >, is, in)। Logical operator এর মধ্যে not আগে, তারপর and, আর সবশেষে or execute হয়। তাই complex expression লেখার সময় precedence না বুঝলে ভুল result আসতে পারে।
