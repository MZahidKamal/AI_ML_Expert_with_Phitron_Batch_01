#### Solution of the CODEFORCES problem: K. Max and Min
Url: https://codeforces.com/group/MWSDmqGsZm/contest/219158/problem/K

```python
inp = input()

numbers = inp.split()

x = int(numbers[0])
y = int(numbers[1])
z = int(numbers[2])

minn = x
maxx = x

## minn
if y < minn:
    minn = y
if z < minn:
    minn = z

## maxx
if y > maxx:
    maxx = y
if z > maxx:
    maxx = z

print(minn, maxx)
```

#### Solution of the CODEFORCES problem: C. Even, Odd, Positive and Negative
Url: https://codeforces.com/group/MWSDmqGsZm/contest/219432/problem/C

```python
nInput = int(input())

numbersInput = input()
numbers= numbersInput.split()

positive =0 ; negative =0 ; even =0 ; odd =0


for i in range(nInput):
    x = int(numbers[i])
    # checking positive or negative
    if x > 0 :
        positive+=1
    elif x < 0 :
        negative+=1

    # checking odd or even
    if x % 2 !=0 :
        odd+=1
    else:
        even+=1

print("Even:",even)
print("Odd:",odd)
print("Positive:",positive)
print("Negative:",negative)
```

#### Solution of the CODEFORCES problem: Q. Digits
Url: https://codeforces.com/group/MWSDmqGsZm/contest/219432/problem/Q

```python
inputN = int(input())

for i in range(inputN):
    number = int(input())
    if number == 0:
        print(0)
        continue
    while number > 0:
        print(number % 10, end=" ")  # by default end = \n a new line, but we are changing it to space
        number //= 10  # In python // is floor division, equivalent to -> number = int(number / 10)
    print()
```
