# Practice Problem Solution 05

```python
# Creating a python dictionary of student names and marks with 4x3 dimension.

students_data = {
    'Student Name': ['Arif', 'Sajid', 'Nabila', 'Riya'],
    'Math Marks': [85, 92, 78, 88],
    'Science Marks': [90, 85, 82, 94],
    'English Marks': [75, 80, 90, 78]
}

# students_data


# Now creating the DataFrame from the above dictionary

students_df = pd.DataFrame(students_data)
students_df
```

```python
# Now will drop the last column using .drop() function

students_df.drop('English Marks', axis=1, inplace=True)
students_df
```

```python
# Now will drop the first row using index

students_df.drop(index=0, inplace=True)
students_df
```

# Practice Problem Solution 06

```python
# Creating a python dictionary of employee names, departments and salary.

employees_data = {
    'Name': ['Rahim', 'Karim', 'Salma', 'Jashim', 'Tania', 'Arif', 'Lata', 'Sohan', 'Mitu', 'Sajid'],
    'Department': ['IT', 'HR', 'Finance', 'IT', 'Marketing', 'Admin', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [65000, 45000, 55000, 70000, 48000, 42000, 58000, 72000, 47000, 46000]
}

# employees_data


# Now creating the DataFrame from the above dictionary

employees_df = pd.DataFrame(employees_data)
employees_df
```

```python
# Sorting the DataFrame, according to the Salary in ascending order and according to the Name in descending order

employees_df.sort_values(by=['Salary', 'Name'], ascending=[True, False])
```

# Practice Problem Solution 07

```python
# Creating a python dictionary of students' names and score.

students_data = {
    'Name': ['Sabbir', 'Anika', 'Tanvir', 'Mitu', 'Fahad', 'Jannat', 'Rakib', 'Nodi'],
    'Score': [85, 92, 38, 95, 88, 30, 32, 91]
}

# students_data


# Now creating the DataFrame from the above dictionary

students_df = pd.DataFrame(students_data)
students_df
```

```python
# Now filtering students with Score > 60.

students_df.loc[students_df['Score'] > 60]
```

```python
# Now filtering students with Score < 40.

students_df.loc[students_df['Score'] < 40]
```
