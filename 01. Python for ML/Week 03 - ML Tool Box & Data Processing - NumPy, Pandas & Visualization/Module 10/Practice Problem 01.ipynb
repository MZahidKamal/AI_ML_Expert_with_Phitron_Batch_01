```python
import numpy as np

# Given a list of temperature records in Celsius
temperatureRecordsInC = [-273.15, -78.5, -18.0, 0.0, 15.5, 22.0, 37.0, 100.0, 180.0, 1538.0]


# Creating a NumPy array from the given temperature records
temperaturesInC = np.array(temperatureRecordsInC)
# print(temperatures)


# Converting all the temperatures from Celsius to Fahrenheit
temperaturesInF = temperaturesInC * 1.8 + 32
print(temperaturesInF)
print(temperaturesInF.dtype)


# Removing scientific notation by suppressing the decimal places
np.set_printoptions(suppress=True)


# Finally printing the converted temperatures
print(temperaturesInF)
print(temperaturesInF.dtype)


print(temperaturesInF.shape)
print(temperaturesInF.size)
print(temperaturesInF.ndim)
print(temperaturesInF.itemsize)
```
