```python
import numpy as np

# Lap times of an athlete in seconds
lapTimesInSec = [62.5, 63.8, 64.2, 63.9, 65.1, 66.4, 67.0, 68.2, 67.5, 61.3]

# Converting the array into a NumPy array
lapTimesInSec = np.array(lapTimesInSec)

# Printing the details of the array
print(lapTimesInSec)
print(type(lapTimesInSec))
print(lapTimesInSec.shape)
print(lapTimesInSec.ndim)
print(lapTimesInSec.dtype)
print(lapTimesInSec.size)

# Converting the array elements from seconds to minute
lapTimesInMin = lapTimesInSec / 60
print(lapTimesInMin)

# Setting the decimal precision to 2
np.set_printoptions(precision=2)

# Finally, printing the array elements in minutes
print(lapTimesInMin)
```
