### 0) Starter Code (Run First)

```python
# You must run this code and use that data set for all the questions
import numpy as np
import pandas as pd

np.random.seed(42)

ids = np.arange(1, 11)
ages = np.random.randint(18, 60, 10)
salaries = np.random.randint(30000, 90000, 10)
departments = np.array(["HR", "IT", "Finance", "IT", "HR", "Sales", "Finance", "IT", "Sales", "HR"])

DF = pd.DataFrame({
    "id": ids,
    "age": ages,
    "salary": salaries,
    "dept": departments
})

DF.to_csv("employees.csv", index=False)
print("Sample Data Created and Saved as employees.csv")
```

```python
# Creating NumPy array from the ages data generated above.
ages_array = np.array(ages)


print(f"Original ages data: {ages_array}")
print(f"Type of the ages array: {type(ages_array)}")
print(f"Data type of the ages array: {ages_array.dtype}")
print(f"Dimensions of the ages array: {ages_array.ndim}")
print(f"Shape of the ages array: {ages_array.shape}")
print(f"Size of the ages array: {ages_array.size}")
```

### Answer to the Question 2:

```python
# Creating NumPy array from the salaries data generated above.
salaries_array = np.array(salaries)


# Calculating the highest, lowest, and average salaries
highest_salary = np.max(salaries_array)
lowest_salary = np.min(salaries_array)
average_salary = np.mean(salaries_array)


print(f"Original salaries data: {salaries_array}")
print(f"Highest Salary: {highest_salary}")
print(f"Lowest Salary: {lowest_salary}")
print(f"Average/Mean Salary: {average_salary}")


print(f"Original ages data: {ages_array}")
print(f"Average/Mean Age: {ages_array.mean()}")
```

### Answer to the Question 3:

```python
# Finding all ages greater than 30, using the boolean condition
ages_array_greater_than_30 = ages_array[ages_array > 30]
print(f"All ages greater than 30: {ages_array_greater_than_30}")


# Now printing the count of employees with age older than 30
print(f"Count of employees with age older than 30: {len(ages_array_greater_than_30)}")


#-------------------- OR --------------------


# Finding all ages greater than 30, using the boolean condition
ages_array_greater_than_30 = np.where(ages_array > 30, ages_array, None)
print(f"All ages greater than 30: {ages_array_greater_than_30}")


# Now printing the count of employees with age older than 30
print(f"Count of employees with age older than 30: {np.count_nonzero(ages_array_greater_than_30)}")
```

### Answer to the Question 4:

```python
# Creating a new array with all employee ages increased by 5 years
# increased_ages_array = ages_array + 5
increased_ages_array = np.add(ages_array, 5)


# Now printing the new ages array
print(f"New increased ages array: {increased_ages_array}")
```

### Answer to the Question 5:

```python
# Calculating the total salary expenses of all employees
total_salary_expenses = np.sum(salaries_array)
print(f"Total salary expenses: {total_salary_expenses}")


# Calculating the difference between Maximum and Minimum salary
salary_difference = np.subtract(highest_salary, lowest_salary)
print(f"Difference between Maximum and Minimum salary: {salary_difference}")
```
