Pandas হলো পাইথনের সবচেয়ে জনপ্রিয় এবং শক্তিশালী **Data Analysis** এবং **Manipulation** লাইব্রেরি। এটি মূলত টেবিল বা স্ট্রাকচারড ডাটা নিয়ে কাজ করার জন্য তৈরি।

### Pandas কী?

Pandas একটি ওপেন সোর্স লাইব্রেরি যা **Series** (একমাত্রিক) এবং **DataFrame** (দ্বিমাত্রিক বা টেবিল) এর মাধ্যমে ডাটা হ্যান্ডেল করে। আপনি একে পাইথনের "Excel" বা "SQL" বলতে পারেন।

### কেন এটি ব্যবহার করা হয়?

* **Ease of Use:** খুব কম কোড লিখে জটিল ডাটা ক্যালকুলেশন করা যায়।
* **Speed:** এটি অভ্যন্তরীণভাবে NumPy ব্যবহার করে, তাই বড় ডাটাবেস নিয়ে কাজ করতে এটি অত্যন্ত দ্রুত।
* **Handling Messy Data:** অগোছালো বা মিসিং ডাটা (Missing Data) ক্লিন করার জন্য এটি সেরা টুল।

### Pandas দিয়ে যা যা করা যায়:

1. **Data Loading:** CSV, Excel, SQL, বা JSON ফাইল থেকে সরাসরি ডাটা লোড করা।
2. **Data Cleaning:** ডুপ্লিকেট ভ্যালু বাদ দেওয়া এবং ফাকা ঘর (NaN) পূরণ করা।
3. **Filtering & Sorting:** নির্দিষ্ট শর্ত অনুযায়ী ডাটা খুঁজে বের করা বা সাজানো।
4. **Aggregation:** ডাটার গড় (Mean), যোগফল (Sum), বা গ্রুপ করে (GroupBy) বিশ্লেষণ করা।
5. **Merging:** একাধিক টেবিল বা ডাটাসেটকে একত্রে জোড়া লাগানো।

### ১. Pandas যেসব ফরম্যাট সাপোর্ট করে:

Pandas-এ ডাটা রিড করার জন্য মূলত `read_` প্রিফিক্স দিয়ে ফাংশনগুলো ব্যবহার করা হয়:

* **Text/Flat Files:** CSV, TSV, Fixed-width (fwf)
* **Spreadsheets:** Excel (.xlsx, .xls)
* **Structured Data:** JSON, XML, HTML
* **Binary/Big Data:** Parquet, HDF5, Feather, Pickle
* **Databases:** SQL (MySQL, PostgreSQL, SQLite) থেকে সরাসরি ক্যুয়েরি করে ডাটা আনা যায়।
* **Clipboard:** কপি করা ডাটা সরাসরি `read_clipboard()` দিয়ে আনা সম্ভব।

# Accessing a row of a dataframe

Pandas-এ একটি **DataFrame** থেকে নির্দিষ্ট রো (Row) অ্যাক্সেস করার জন্য প্রধানত দুটি পদ্ধতি ব্যবহার করা হয়। নিচে তা সংক্ষেপে আলোচনা করা হলো:

### ১. `.iloc[]` (Integer-based Indexing)

যখন আপনি ইনডেক্সের **পজিশন বা সংখ্যা** (যেমন: ০, ১, ২...) ব্যবহার করে রো অ্যাক্সেস করতে চান, তখন `iloc` ব্যবহার করা হয়।

* **ব্যবহার:** `df.iloc[index_number]`
* **উদাহরণ:** ৩ নম্বর রো অ্যাক্সেস করতে চাইলে: `df.iloc[2]` (যেহেতু ইনডেক্স ০ থেকে শুরু হয়)।

### ২. `.loc[]` (Label-based Indexing)

যখন আপনি ইনডেক্সের **নাম বা লেবেল** (যেমন: 'row1', 'row2' বা নির্দিষ্ট আইডি) ব্যবহার করে রো অ্যাক্সেস করতে চান, তখন `loc` ব্যবহার করা হয়।

* **ব্যবহার:** `df.loc['row_label']`
* **উদাহরণ:** যদি ইনডেক্সের নাম হয় 'id_101', তবে: `df.loc['id_101']`।

### ৩. একাধিক রো একসাথে দেখা (Slicing)

আপনি যদি একটি রেঞ্জ বা নির্দিষ্ট কয়েকটি রো একসাথে দেখতে চান:

* **iloc দিয়ে:** `df.iloc[0:5]` (প্রথম ৫টি রো)।
* **loc দিয়ে:** `df.loc['start_label':'end_label']`।

# Solution of the LEETCODE problem: 595. Big Countries
Url: https://leetcode.com/problems/big-countries/description/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata

```python
# Creating the array with an 'object' dtype and with the exact given data in the problem description
world_array = np.array([
    ['Afghanistan', 'Asia', 652230, 25500100, 20343000000],
    ['Albania', 'Europe', 28748, 2831741, 12960000000],
    ['Algeria', 'Africa', 2381741, 37100000, 188681000000],
    ['Andorra', 'Europe', 468, 78115, 3712000000],
    ['Angola', 'Africa', 1246700, 20609294, 100990000000]
], dtype=object)


# world_array
# type(world_array)


# Now converting the array into a dataframe
world = pd.DataFrame(world_array, columns=['name', 'continent', 'area', 'population', 'gdp'])
# world
# type(world)


# The given function in the problem description
def big_countries(world: pd.DataFrame) -> pd.DataFrame:

    # Filter the dataframe based on the given conditions
    big_country = world.loc[(world['area'] >= 3000000) | (world['population'] >= 25000000)]

    # Select only the name, population and area columns as per the requirement
    result = big_country[['name', 'population', 'area']]

    # Return the result dataframe
    return result


big_countries(world)
```

# Solution of the LEETCODE problem: 1757. Recyclable and Low Fat Products
Url: https://leetcode.com/problems/recyclable-and-low-fat-products/description/

```python
# Creating the array with an 'object' dtype and with the exact given data in the problem description
products_array = np.array([
    [0, 'Y', 'N'],
    [1, 'Y', 'Y'],
    [2, 'N', 'Y'],
    [3, 'Y', 'Y'],
    [4, 'N', 'N']
], dtype=object)


# products_array
# type(world_array)


# Now converting the array into a dataframe
products = pd.DataFrame(products_array, columns=['product_id', 'low_fats', 'recyclable'])
# products
# type(products)


# The given function in the problem description
def find_products(products: pd.DataFrame) -> pd.DataFrame:

    # Filter the dataframe based on the given conditions and select only the product_id column as per the requirement
    low_fat_and_recyclable = products.loc[(products['low_fats'] == 'Y') & (products['recyclable'] == 'Y'), ['product_id']]

    # Return the result dataframe
    return low_fat_and_recyclable


find_products(products)
```
