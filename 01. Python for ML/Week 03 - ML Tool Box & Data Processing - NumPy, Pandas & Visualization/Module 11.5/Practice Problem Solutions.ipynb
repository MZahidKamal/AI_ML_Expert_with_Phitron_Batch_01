# Practice Problem Solution 03

```python
# Creating an array of even numbers between 10 and 50, using arrange function

array_even_01 = np.arange(10, 51, 2)
print(array_even_01)
print(array_even_01.shape)
print(array_even_01.size)
print(array_even_01.ndim)
print(array_even_01.dtype)
```

```python
# Creating an array of even numbers between 10 and 50, using linspace function

array_even_02 = np.linspace(10, 50, 21)
print(array_even_02)
print(array_even_02.shape)
print(array_even_02.size)
print(array_even_02.ndim)
print(array_even_02.dtype)
```

```python
# Creating an array of even numbers between 10 and 50, using the logspace function

array_even_03 = np.logspace(10, 50, 51)             # not possible to get even numbers using logspace
print(array_even_03)
```

# Practice Problem Solution 04

```python
# Creating a 4x4 matrix from 0-15.
array_4x4 = np.arange(16).reshape(4, 4)
print(array_4x4)
print(array_4x4.shape)          # Shape represents the dimensions of the array. The first value is the number of rows, the second value is the number of columns.
print(array_4x4.size)
print(array_4x4.ndim)
print(array_4x4.dtype)
```

```python
row = array_4x4.shape[0]
col = array_4x4.shape[1]


# Extracting the center matrix from the 4x4 matrix, removing the top and bottom rows and the left and right columns. This is called slicing.
# extracted_center_matrix = array_4x4[1:3, 1:3]
extracted_center_matrix = array_4x4[1:(row-1), 1:(col-1)]
print(extracted_center_matrix)
```

# Practice Problem Solution 06

```python
# Creating a random array from 1 to 10
arrayA = np.arange(1, 11)
print(arrayA)


# Creating another array from 11 to 20
arrayB = np.arange(11, 21)
print(arrayB)


# Calculating the average of the two arrays
arrayC = np.broadcast_arrays((arrayA + arrayB)/2)
print(arrayC)


# Calculating the addition of squares of both arrays
arrayD = np.broadcast_arrays((arrayA ** 2 + arrayB ** 2))
print(arrayD)
```
