# Solution of the DEEP-ML problem: Calculate Mean by Row or Column
Url: https://www.deep-ml.com/problems/4

```python
# This needs to be imported at the top of the file, inside the Deep-ML code writing section.
import numpy as np


def calculate_matrix_mean(matrix: list[list[float]], mode: str) -> list[float]:

    # Initially converting the python list of lists into a numpy array. And at the same time changing the data type of the array into float.
    matrix = np.array(matrix).astype(float)

    # Calculating the mean of the matrix based on the given mode.
    if mode == 'row':
        means = np.mean(matrix, axis=1)
    else:
        means = np.mean(matrix, axis=0)

    # Now converting the numpy array into a list of floats.
    means = means.tolist()

    # Returning the list of means.
    return means


print(calculate_matrix_mean([[1, 2, 3], [4, 5, 6], [7, 8, 9]], 'column'))
```

# Solution of the DEEP-ML problem: Matrix-Vector Dot Product
Url: https://www.deep-ml.com/problems/1

```python
# This needs to be imported at the top of the file, inside the Deep-ML code writing section.
import numpy as np


def matrix_dot_vector(a: list[list[int | float]], b: list[int | float]) -> list[int | float]:
    # Return a list where each element is the dot product of a row of 'a' with 'b'.
    # If the number of columns in 'a' does not match the length of 'b', return -1.

    # Initially converting the python list of lists into a numpy array. But we'll not change the data type at this point.
    a = np.array(a)
    b = np.array(b)

    # Checking if the number of columns in 'a' matches the length of 'b'.
    if np.shape(a)[1] != len(b):
        # If not, returning -1.
        return -1
    else:
        # If yes, calculating the dot product of each row of 'a' with 'b'.
        ab_dot_product = np.dot(a, b)

        # Changing the data type of 'ab_dot_product' into int64 only if the data type of 'ab_dot_product' is int64.
        if ab_dot_product.dtype == 'int64':
            ab_dot_product = ab_dot_product.astype(int)

        # Converting the numpy array 'a' into a python list.
        ab_dot_product = ab_dot_product.tolist()

        # Returning the dot product of each row of 'a' with 'b'.
        return ab_dot_product




print(matrix_dot_vector([[1, 2, 3], [2, 4, 5], [6, 8, 9]], [1, 2, 3]))
```

# Solution of the DEEP-ML problem: Transpose of a Matrix
Url: https://www.deep-ml.com/problems/2

```python
# This needs to be imported at the top of the file, inside the Deep-ML code writing section.
import numpy as np


def transpose_matrix(a: list[list[int | float]]) -> list[list[int | float]]:
    """
    Transpose a 2D matrix by swapping rows and columns.
    Args:
        a: A 2D matrix of shape (m, n)
    Returns:
        The transposed matrix of shape (n, m)
    """

    # First, converting the python list of lists into a numpy array.
    a = np.array(a)

    # Then transposing the matrix using the numpy function 'transpose'.
    a_transposed = np.transpose(a)

    # Finally, converting the transposed numpy array into a python list of lists.
    a_transposed = a_transposed.tolist()

    # Finally, returning the transposed matrix.
    return a_transposed


print(transpose_matrix([[1, 2], [3, 4], [5, 6]]))
```
