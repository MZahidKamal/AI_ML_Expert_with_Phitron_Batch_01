**Matplotlib** হলো পাইথনের সবচেয়ে জনপ্রিয় এবং বহুল ব্যবহৃত একটি **Data Visualization** লাইব্রেরি। এটি মূলত ডেটাকে গ্রাফ বা চার্টের মাধ্যমে সুন্দরভাবে উপস্থাপন করার জন্য ব্যবহৃত হয়।

* **Pandas Integration:** এটি **Pandas DataFrame** এর সাথে খুব সহজে কাজ করতে পারে।

# 2D Line Plot


**2D Line Plot** হলো **Matplotlib** লাইব্রেরির সবচেয়ে সাধারণ গ্রাফ, যা মূলত সময়ের সাথে কোনো ডেটার পরিবর্তন বা দুটি ভেরিয়েবলের মধ্যে সম্পর্ক দেখাতে ব্যবহৃত হয়।

### **2D Line Plot** এর মূল বৈশিষ্ট্যসমূহ:

* **Functionality:** এটি কয়েকটি পয়েন্টকে (**Points**) সরলরেখা দিয়ে যুক্ত করে একটি ট্রেন্ড তৈরি করে।
* **Primary Function:** এটি তৈরি করতে `plt.plot()` ফাংশন ব্যবহার করা হয়।
* **Axes:** এতে দুটি এক্সিস থাকে— **X-axis** (অণুভূমিক) এবং **Y-axis** (উল্লম্ব)।

### কাস্টমাইজ করার প্রয়োজনীয় কমান্ডসমূহ:

| কমান্ড | কাজ |
| --- | --- |
| **`color`** | লাইনের রঙ পরিবর্তন করতে (যেমন: `color='red'`)। |
| **`marker`** | ডেটা পয়েন্টগুলো হাইলাইট করতে (যেমন: `marker='o'`)। |
| **`linestyle`** | লাইনের স্টাইল পরিবর্তন করতে (যেমন: `linestyle='--'`)। |
| **`label`** | গ্রাফের নাম বা **Legend** সেট করতে। |

**মনে রাখবেন:** গ্রাফের অর্থ পরিষ্কার করার জন্য সবসময় `xlabel`, `ylabel` এবং `title` ব্যবহার করা উচিত।

```python
# We will generate some data and will plot it using Matplotlib.

hours = [2, 3, 4, 1, 5, 7, 3]
days = [1, 2, 3, 4, 5, 6, 7]

plt.plot(days, hours)   # The first argument is for X axis, and the second argument is for Y axis.

# In general, the continuous time variable comes on the X axis. In this case the continuous time variable is 'day' and therefore it's on the X axis.
```

```python
# Now we will add labels and a title to our plot.

plt.xlabel("Day")
plt.ylabel("Study Hours")
plt.title("Trend of Study Hours over the Week")

plt.plot(days, hours)
```

```python
# Now we'll generate data for 2 students and will plot it.

hours_1 = [2, 3, 4, 1, 5, 7, 3]
hours_2 = [1, 4, 3, 5, 7, 2, 1]
days = [1, 2, 3, 4, 5, 6, 7]

plt.xlabel("Day")
plt.ylabel("Study Hours")
plt.title("Trend of Study Hours over the Week")

# Simple way of doing it
# plt.plot(days, hours_1)
# plt.plot(days, hours_2)

# Advanced way of doing it
plt.plot(days, hours_1, label="Student 1", color="red", linestyle="--", marker="o", alpha=0.5, linewidth=2)
plt.plot(days, hours_2, label="Student 2", color="blue", linestyle="-.", marker="o", alpha=0.5, linewidth=2)

# It adds labels inside the graph area.
plt.legend()
```

* **মূল পর্যবেক্ষণ:** পুরো সপ্তাহে উভয় শিক্ষার্থীর পড়াশোনার রুটিন বেশ অনিয়মিত ছিল, যেখানে **Student 1** ষষ্ঠ দিনে সর্বোচ্চ (৭ ঘণ্টা) এবং **Student 2** পঞ্চম দিনে সর্বোচ্চ (৭ ঘণ্টা) পড়াশোনা করেছেন।
* **বিপরীত চিত্র:** চতুর্থ দিনে যখন **Student 1** সবচেয়ে কম (মাত্র ১ ঘণ্টা) পড়াশোনা করেছেন, ঠিক সেই সময়েই **Student 2** তার পড়াশোনার সময় বাড়িয়ে ৫ ঘণ্টায় নিয়ে গেছেন।

# Scatter Plot

**Scatter Plot** হলো এমন এক ধরণের গ্রাফ যা ডেটাসেটের প্রতিটি ভ্যালুকে একটি বিন্দু বা **Dot** হিসেবে প্রদর্শন করে। এটি মূলত দুটি ভেরিয়েবলের মধ্যে সম্পর্ক বা **Correlation** বোঝার জন্য ব্যবহৃত হয়।

### **Scatter Plot** এর মূল বৈশিষ্ট্যসমূহ:

* **Functionality:** এটি ডেটা পয়েন্টগুলোকে যুক্ত না করে আলাদা আলাদা বিন্দু হিসেবে দেখায়।
* **Primary Function:** এটি তৈরি করতে `plt.scatter()` ফাংশন ব্যবহার করা হয়।
* **Use Case:** এটি মূলত **Outliers** (অস্বাভাবিক ডেটা) খুঁজে বের করতে এবং ডেটার ঘনত্ব বা **Cluster** বুঝতে ব্যবহৃত হয়।

**মনে রাখবেন:** যদি বিন্দুগুলো একটি নির্দিষ্ট দিকে লাইন ধরে আগায়, তবে বুঝতে হবে ভেরিয়েবল দুটির মধ্যে শক্তিশালী সম্পর্ক আছে। আর বিন্দুগুলো এলোমেলো থাকলে বুঝতে হবে তাদের মধ্যে কোনো সম্পর্ক নেই।

```python
# Now we'll read the data from a CSV file, and we'll convert it to a pandas DataFrame'.

student_iq_df = pd.read_csv('student_IQdata.csv')
student_iq_df
```

```python
# Now we'll create a scatter plot from the given DataFrame. Mainly will be used to visualize the relationship between Study hour and IQ Score.

plt.scatter(student_iq_df['Study_Hour'], student_iq_df['IQ_Score'])

plt.xlabel('Daily Study Hour')
plt.ylabel('IQ Level')
plt.title('IQ score aggregation over study time')

plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.5)

plt.show()

# We can see a kind of positive correlation with two outliers.
```

* **মূল পর্যবেক্ষণ:** চার্টটি থেকে দেখা যায় যে পড়াশোনার সময় বাড়ার সাথে সাথে **IQ Level** বাড়ার কোনো শক্তিশালী বা স্পষ্ট ট্রেন্ড নেই; ডেটা পয়েন্টগুলো বেশ ছড়িয়ে ছিটিয়ে আছে।
* **বিশেষ পয়েন্ট:** গ্রাফের উপরের বাম দিকে একটি **Outlier** দেখা যাচ্ছে, যেখানে অনেক কম পড়াশোনা (প্রায় ২ ঘণ্টা) করেও একজনের **IQ Level** অনেক বেশি (প্রায় ১৪৫)।

```python
# Now we'll create a scatter plot from the given DataFrame. But this time will visualize the relationship between Shoe Size and IQ Score.

plt.scatter(student_iq_df['Shoe_Size'], student_iq_df['IQ_Score'])

plt.xlabel('Daily Study Hour')
plt.ylabel('Shoe Size')
plt.title('IQ score aggregation over shoe size')

plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.5)

plt.show()

# We can see a kind of no correlation with two outliers.
```

* **মূল পর্যবেক্ষণ:** পড়াশোনার সময়ের সাথে জুতার সাইজের কোনো যৌক্তিক বা সরাসরি সম্পর্ক নেই; ডেটা পয়েন্টগুলো চার্টে সম্পূর্ণ এলোমেলোভাবে ছড়িয়ে আছে।
* **বিশেষ লক্ষণ:** এখানে পড়াশোনার হার কম বা বেশি হওয়ার সাথে জুতার সাইজ বাড়া বা কমার কোনো নির্দিষ্ট প্যাটার্ন খুঁজে পাওয়া যায় না।

```python
# Now we'll create a scatter plot from the given DataFrame. But this time will visualize the relationship between Chilling Hours and IQ Score.

plt.scatter(student_iq_df['Chilling_Hours'], student_iq_df['IQ_Score'], color='red', marker='o', alpha=1, edgecolor='blue')

plt.xlabel('Daily Study Hour')
plt.ylabel('Chilling Hours')
plt.title('IQ score aggregation over chilling hours')

plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.5)

plt.show()

# We can see a kind of negative correlation with one outlier.
```

* **মূল পর্যবেক্ষণ:** গ্রাফটি থেকে দেখা যায় যে পড়াশোনার সময় এবং বিনোদনের সময়ের মধ্যে কোনো সুনির্দিষ্ট বা শক্তিশালী রৈখিক সম্পর্ক নেই; ডেটা পয়েন্টগুলো বেশ ছড়িয়ে-ছিটিয়ে রয়েছে।
* **ব্যতিক্রম (Outlier):** গ্রাফের একদম ডানদিকে ৩৫ ঘণ্টা পড়াশোনা এবং প্রায় ১৪৫ ঘণ্টা বিনোদনের একটি ডেটা পয়েন্ট দেখা যাচ্ছে, যা একটি অস্বাভাবিক বা **Outlier** মান হিসেবে প্রতীয়মান হয়।

# Histogram

**Histogram** হলো এমন এক ধরণের গ্রাফ যা কোনো ডেটাসেটের ডিস্ট্রিবিউশন বা বিস্তৃতি বোঝার জন্য ব্যবহৃত হয়। এটি মূলত দেখায় যে একটি নির্দিষ্ট সীমার মধ্যে কতবার ডেটাগুলো অবস্থান করছে।

### **Histogram** এর মূল বৈশিষ্ট্যসমূহ:

* **Functionality:** এটি ডেটাকে কতগুলো ছোট ছোট ভাগে বা **Bins** এ ভাগ করে এবং প্রতি ভাগে কতগুলো ডেটা আছে তার ফ্রিকুয়েন্সি দেখায়।
* **Primary Function:** এটি তৈরি করতে `plt.hist()` ফাঞ্চন ব্যবহার করা হয়।
* **Difference from Bar Chart:** **Bar Chart** বিভিন্ন ক্যাটাগরির তুলনা করে, কিন্তু **Histogram** একটি কন্টিনিউয়াস ডেটাসেটের ডিস্ট্রিবিউশন দেখায়।

**মনে রাখবেন:** **Bins** এর সংখ্যা খুব বেশি বা খুব কম হলে গ্রাফের অর্থ বোঝা কঠিন হয়ে যায়, তাই একটি ব্যালেন্সড সংখ্যা ব্যবহার করা জরুরি।

```python
student_iq_df
```

* **প্রবণতা:** স্কোর যত বাড়ছে, শিক্ষার্থীর সংখ্যা তত কমছে; তবে **১৪০-১৪৫** রেঞ্জে একজন অত্যন্ত উচ্চ মেধাবী শিক্ষার্থীর উপস্থিতি (Outlier) লক্ষ্য করা যাচ্ছে।

# Bar Chart

**Bar Chart** হলো এমন এক ধরণের গ্রাফ যা আয়তকার বারের (**Bars**) মাধ্যমে বিভিন্ন ক্যাটাগরির ডেটা তুলনা করতে ব্যবহৃত হয়। এটি মূলত ডিসক্রিট (**Discrete**) বা আলাদা আলাদা শ্রেণির তথ্য প্রদর্শনের জন্য সেরা।

### **Bar Chart** এর মূল বৈশিষ্ট্যসমূহ:

* **Functionality:** এটি বিভিন্ন গ্রুপের মধ্যে তুলনা করার জন্য ব্যবহৃত হয়।
* **Primary Function:** এটি তৈরি করতে `plt.bar()` (ভার্টিক্যাল) অথবা `plt.barh()` (হরাইজন্টাল) ফাংশন ব্যবহার করা হয়।
* **Components:** এর এক অক্ষে থাকে ক্যাটাগরি এবং অন্য অক্ষে থাকে সেই ক্যাটাগরির মান বা ফ্রিকুয়েন্সি।

**মনে রাখবেন:** যখন ক্যাটাগরির নামগুলো অনেক বড় হয়, তখন **Horizontal Bar Chart** (`plt.barh()`) ব্যবহার করা বেশি সুবিধাজনক।

```python
# We will generate some data and will plot it using Matplotlib.

class_student_data = {
    'Student': ['B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'StudyHours': [2, 5, 3, 7, 4, 8, 1],
    'Attendance': [60, 88, 70, 95, 80, 98, 55],
    'Grade': ['Fail', 'Pass', 'Fail', 'Pass', 'Pass', 'Pass', 'Fail']
}

class_student_df = pd.DataFrame(class_student_data)
class_student_df
```

# Pie Chart

**Pie Chart** হলো একটি বৃত্তাকার গ্রাফ যা কোনো একটি সম্পূর্ণ অংশের (**Whole**) বিভিন্ন ভাগ বা শতাংশ (**Percentage**) দেখাতে ব্যবহৃত হয়। এটি মূলত একটি নির্দিষ্ট ডেটাসেটের উপাদানগুলো একে অপরের তুলনায় কত বড় তা বুঝতে সাহায্য করে।

### **Pie Chart** এর মূল বৈশিষ্ট্যসমূহ:

* **Functionality:** এটি ডেটাকে বিভিন্ন স্লাইস বা টুকরো হিসেবে প্রদর্শন করে।
* **Primary Function:** এটি তৈরি করতে `plt.pie()` ফাংশন ব্যবহার করা হয়।
* **Total Value:** একটি **Pie Chart** এর সবগুলোর স্লাইসের সমষ্টি সবসময় ১০০% বা একটি পূর্ণ বৃত্ত বোঝায়।

| প্যারামিটার | কাজ |
| --- | --- |
| **`labels`** | প্রতিটি স্লাইসের নাম দেওয়ার জন্য। |
| **`autopct`** | স্লাইসের ভেতরে শতাংশ (**Percentage**) দেখানোর জন্য (যেমন: `'%1.1f%%'`)। |
| **`explode`** | নির্দিষ্ট কোনো স্লাইসকে বৃত্ত থেকে একটু আলাদা করে দেখাতে। |
| **`shadow`** | চার্টে মডিশ লুক বা ছায়া যোগ করার জন্য। |
| **`startangle`** | চার্টটি কোন ডিগ্রি থেকে শুরু হবে তা ঠিক করতে। |

**মনে রাখবেন:** যখন ক্যাটাগরির সংখ্যা খুব বেশি (যেমন ১০টির বেশি) হয়ে যায়, তখন **Pie Chart** এর বদলে **Bar Chart** ব্যবহার করা বেশি পরিষ্কার এবং অর্থবহ হয়।
