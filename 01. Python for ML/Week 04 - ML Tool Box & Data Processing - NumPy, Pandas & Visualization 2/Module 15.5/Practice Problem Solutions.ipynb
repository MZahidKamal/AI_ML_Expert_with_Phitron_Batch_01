```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Dataset
* **URL:** https://github.com/phitronio/Python-for-ML/blob/main/Practice%20Day%20(1).csv

# Practice Problem Solution 01

```python
dataset_url = "https://raw.githubusercontent.com/phitronio/Python-for-ML/refs/heads/main/Practice%20Day%20(1).csv"
employee_df = pd.read_csv(dataset_url)
employee_df
```

```python
# Now we'll add a new column 'Discounted_Price,' which will contain the 10% discounted price comparing to the price from the column 'Price'.
employee_df['Discounted_Price'] = employee_df['Price'] * 0.9
employee_df
```

```python
# Now we'll save the modified DataFrame as 'discounted_products.csv'.
employee_df.to_csv('discounted_products.csv')

# Now we'll read the modified DataFrame from 'discounted_products.csv', just to confirm that the file was saved correctly.
employee_df2 = pd.read_csv('discounted_products.csv')
employee_df2
```

# Practice Problem Solution 02

```python
# Let's see the DataFrame again.

employee_df2
```

```python
# Now we'll count how many duplicate 'EmployeeID' we have in the DataFrame.

# employee_df2['EmployeeID'].duplicated().sum()
print(employee_df2['EmployeeID'].duplicated().sum())
```

```python
# We have got 2 duplicate entries in the column 'EmployeeID' of the DataFrame. Now we'll remove these duplicate rows completely. And then we'll display the cleaned DataFrame again.

employee_df2.drop_duplicates(subset='EmployeeID', inplace=True)
employee_df2
```

# Practice Problem Solution 03

```python
# Let's see the DataFrame again.

employee_df2
```

```python
# Now we'll find missing values in the columns 'Age' and 'Salary'. And then count them.

employee_df2[['Age', 'Salary']].isnull().sum()
```

```python
# Now we'll fill the missing values in the 'Age' column with the mean value of the full column.

# employee_df2['Age'].fillna(employee_df2['Age'].mean(), inplace=True)    # It may raise the copy_on_write error. Rather than using inplace=True, we can have another approach.

employee_df2['Age'] = employee_df2['Age'].fillna(employee_df2['Age'].mean())
employee_df2
```

```python
# Now we'll drop the row(s), where the value is missing in the 'Salary' column.

employee_df2.dropna(subset=['Salary'], inplace=True)
employee_df2
```

# Practice Problem Solution 04

```python
# Let's see the DataFrame again.

employee_df2
```

```python
# Now we'll calculate mean and max for the columns 'Marks1', 'Marks2' and 'Marks3'.


# employee_df2[['Marks1', 'Marks2', 'Marks3']].mean()
print("Mean values of Marks1, Marks2 and Marks3:")
print(employee_df2[['Marks1', 'Marks2', 'Marks3']].mean(), '\n')


# employee_df2[['Marks1', 'Marks2', 'Marks3']].max()
print("Max values of Marks1, Marks2 and Marks3:")
print(employee_df2[['Marks1', 'Marks2', 'Marks3']].max())
```

```python
# Now we'll calculate the (Marks1 + Marks2 + Marks3) total marks of each student.

employee_df2['Total_Marks'] = employee_df2['Marks1'] + employee_df2['Marks2'] + employee_df2['Marks3']
employee_df2
```

```python
# Now we'll find the standard deviation of the column 'Total_Marks'.

# employee_df2['Total_Marks'].std()
print(employee_df2['Total_Marks'].std())
```

# Practice Problem Solution 05

```python
# Let's see the DataFrame again.

employee_df2
```

```python
# Now we'll Histogram of the column 'Age' of the DataFrame.

plt.hist(employee_df2['Age'], bins=10, color='#44FF99', edgecolor='black', linewidth=0.5)

plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age of Employees')

plt.grid(color='gray', linestyle='dotted', linewidth=0.5, alpha=1)
plt.xticks(range(20, 50, 2), rotation=0, fontsize=9, ha='right', va='top', color='#44FF99')
plt.yticks(range(0, 4, 1), rotation=0, fontsize=9, ha='right', va='top', color='#44FF99')

plt.show()
```

```python
# Now we'll create a Scatter Plot of the column 'Age' vs. 'Salary' of the DataFrame.

plt.scatter(employee_df2['Age'], employee_df2['Salary'], color='#44FF99', edgecolor='black', linewidth=0.5, marker='o', alpha=1)

plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Scatter Plot of Age vs Salary of Employees')

plt.xticks(range(20, 50, 2), rotation=0, fontsize=9, ha='right', va='top', color='#44FF99')
plt.yticks(range(40000, 100000, 10000), rotation=0, fontsize=9, ha='right', va='top', color='#44FF99')

plt.grid(color='gray', linestyle='dotted', linewidth=0.5, alpha=1)

plt.show()
```

```python
# Now we'll create a Scatter Plot of the column 'Price' vs. 'Discounted_Price' of the DataFrame.

plt.scatter(employee_df2['Price'], employee_df2['Discounted_Price'], color='#44FF99', edgecolor='black', linewidth=0.5, marker='o', alpha=1)

plt.xlabel('Price')
plt.ylabel('Discounted Price')
plt.title('Scatter Plot of Price vs Discounted Price of Products')

plt.xticks(range(1000, 3000, 200), rotation=0, fontsize=9, ha='right', va='top', color='#44FF99')
plt.yticks(range(800, 2500, 200), rotation=0, fontsize=9, ha='right', va='top', color='#44FF99')

plt.grid(color='gray', linestyle='dotted', linewidth=0.5, alpha=1)

plt.show()
```
