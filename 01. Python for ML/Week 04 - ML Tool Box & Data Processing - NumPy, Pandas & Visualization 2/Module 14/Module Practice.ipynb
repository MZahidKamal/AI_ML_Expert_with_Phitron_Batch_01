### Pandas-এ ফাংশন অ্যাপ্লাই
Pandas-এ DataFrame-এর ওপর কাস্টম ফাংশন অ্যাপ্লাই করা যায় এবং এটি ডাটা ম্যানিপুলেশনের জন্য অত্যন্ত শক্তিশালী একটি টুল। Pandas-এ কোনো নির্দিষ্ট রো (Row) বা কলামের (Column) ওপর নিজের তৈরি করা কোনো লজিক চালানোর জন্য .apply() মেথড ব্যবহার করা হয়। এটি লুপ (For Loop) চালানোর চেয়ে অনেক দ্রুত এবং ক্লিন।

### কেন .apply() ব্যবহার করবেন?
##### (1) যখন Pandas-এর বিল্ট-ইন কোনো মেথড (যেমন: mean, sum) দিয়ে আপনার কাজ হচ্ছে না।
##### (2) যখন ডাটার ওপর জটিল কোনো কন্ডিশন অ্যাপ্লাই করতে হচ্ছে।


### আপনি দুইভাবে ফাংশন ব্যবহার করতে পারেন:
##### (1) Normal Function: বড় লজিকের জন্য।
##### (2) Lambda Function: ছোট এক লাইনের লজিকের জন্য।

### Scaling করা কি কোনো ফাংশন?
প্রযুক্তিগতভাবে বলতে গেলে, Scaling একটি গাণিতিক প্রক্রিয়া বা Transformation, কোনো বিল্ট-ইন ফাংশন নয়। তবে এটি করার জন্য আমরা ফাংশন ব্যবহার করি।

### কেন করা হয়? (Why is Scaling important?)
##### (1) Algorithm Performance: অনেক Machine Learning অ্যালগরিদম (যেমন: KNN, SVM, K-Means) দূরত্বের ওপর ভিত্তি করে কাজ করে। বড় ভ্যালু থাকলে অ্যালগরিদম মনে করে সেই কলামটি বেশি গুরুত্বপূর্ণ, যা ভুল রেজাল্ট দিতে পারে।
##### (2) Fast Convergence: Gradient Descent ভিত্তিক অ্যালগরিদমগুলো Scaling করা থাকলে খুব দ্রুত রেজাল্ট দিতে পারে।
##### (3) Fair Comparison: বিভিন্ন এককের ডাটাকে (কেজি, মিটার, টাকা) সমান গুরুত্ব দেওয়ার জন্য এটি প্রয়োজন।

## 📐 Data Scaling Methods: বিস্তারিত নোট

যখন আপনার ডাটাসেটের বিভিন্ন **Features** বা কলামের ভ্যালু রেঞ্জ আলাদা হয়, তখন সেগুলোকে একটি নির্দিষ্ট মাপে নিয়ে আসাকেই **Scaling** বলে। এর প্রধান পদ্ধতিগুলো নিচে আলোচনা করা হলো:

### ১. Normalization (Min-Max Scaling)

এই পদ্ধতিতে ডাটার সর্বনিম্ন মানকে **0** এবং সর্বোচ্চ মানকে **1** ধরে বাকি সব ভ্যালুকে এই সীমার মধ্যে নিয়ে আসা হয়।

* **কখন ব্যবহার করবেন:** যখন আপনি জানেন যে আপনার ডাটা কোনো নির্দিষ্ট **Distribution** (যেমন: Gaussian distribution) মেনে চলে না। ইমেজ প্রসেসিং বা এমন অ্যালগরিদম যেখানে দূরত্বের হিসাব আছে (যেমন: KNN) সেখানে এটি খুব কার্যকর।
* **সীমাবদ্ধতা:** এটি **Outliers** (অস্বাভাবিক মান) দ্বারা খুব বেশি প্রভাবিত হয়।

### ২. Standardization (Z-score Normalization)

এই পদ্ধতিতে ডাটার **Mean** (গড়) করা হয় **0** এবং **Standard Deviation** করা হয় **1**। এর ফলে ডাটা একটি স্ট্যান্ডার্ড নরমাল ডিস্ট্রিবিউশনে রূপান্তরিত হয়।

* **কখন ব্যবহার করবেন:** যখন আপনার মেশিন লার্নিং মডেল (যেমন: Logistic Regression, SVM বা Linear Regression) ডাটা নরমাল ডিস্ট্রিবিউশনে আছে বলে ধরে নেয়।
* **সুবিধা:** এটি **Outliers** থাকলেও ডাটার মূল বৈশিষ্ট্য খুব একটা নষ্ট করে না।

### ৩. Robust Scaling

যদি আপনার ডাটাসেটে প্রচুর পরিমাণে **Outliers** থাকে, তবে **Min-Max** বা **Standard Scaling** ভালো কাজ করে না। সেক্ষেত্রে **Robust Scaling** ব্যবহার করা হয়। এটি **Mean** বা **Max** ভ্যালুর বদলে **Median** এবং **Interquartile Range (IQR)** ব্যবহার করে ডাটা স্কেল করে।

* **সুবিধা:** এটি আউটলায়ারের প্রতি অনেক বেশি সহনশীল।

### ৪. Power Transformer / Log Transformation

অনেক সময় ডাটা খুব বেশি একপেশে (Skewed) থাকে। এই পদ্ধতিগুলো ব্যবহার করে ডাটার **Skewness** কমানো হয় এবং ভ্যারিয়েন্সকে স্থিতিশীল করা হয়।


### 📌 মূল পার্থক্য একনজরে:

| বৈশিষ্ট্য | Normalization | Standardization | Robust Scaling |
| --- | --- | --- | --- |
| **Range** | সাধারণত 0 থেকে 1 | কোনো নির্দিষ্ট সীমা নেই | কোনো নির্দিষ্ট সীমা নেই |
| **Mean** | নির্ধারিত নয় | সবসময় 0 হয় | নির্ধারিত নয় |
| **Outliers** | খুব বেশি প্রভাবিত হয় | কিছুটা প্রভাবিত হয় | সবচেয়ে কম প্রভাবিত হয় |

---

**মনে রাখবেন:** মডেল ট্রেনিং করার আগে সবসময় ডাটা স্কেলিং করা উচিত, বিশেষ করে যদি আপনি **Distance-based** বা **Gradient Descent** ভিত্তিক অ্যালগরিদম ব্যবহার করেন।

```python
# Now we'll do the MinMax Scaling on the Total Marks of the DataFrame.


# First, we'll find the minimum and maximum values of the 'Total Marks' column.
mx = student_data_df['Total Marks'].max()
mn = student_data_df['Total Marks'].min()


# Then, this is one way of scaling the 'Total Marks' column:
# student_data_df['Scaled Total Marks'] = (student_data_df['Total Marks'] - mn) / (mx - mn)


# Now this is another way of scaling the 'Total Marks' column:
customBuildFunction = lambda x: (x - mn) / (mx - mn)
student_data_df['Scaled Total Marks'] = student_data_df['Total Marks'].apply(customBuildFunction)


# We can remove the 'Scaled Total Marks' column from the DataFrame and try it again.
# student_data_df.drop(columns=['Scaled Total Marks'], inplace=True)


# Let's see the DataFrame again.
student_data_df
```

```python
# Now we'll build and use another custom function to grade the students based on their Total Marks.

def grading_system(total_mark):
    if total_mark >= 260:
        return 'A+'
    elif total_mark >= 250:
        return 'A'
    elif total_mark >= 240:
        return 'A-'
    else:
        return 'B'

student_data_df['Grade'] = student_data_df['Total Marks'].apply(grading_system)
student_data_df
```

```python
# Now we'll build and use another custom function on the complete DataFrame.


def marking_system(data_frame):
    a = data_frame['Data Structure Marks'] * 2
    b = data_frame['Python Marks'] * 3
    c = data_frame['Algorithm Marks'] * 4
    return a + b + c


student_data_df['Multiplied_by_2_3_4 Marks'] = student_data_df.apply(marking_system, axis=1)    # As we are doing the summation rowwise, we'll use axis = 1.
student_data_df
```

### ১. Datetime কী? (What is Datetime?)

**Datetime** হলো একটি নির্দিষ্ট মুহূর্তের তারিখ এবং সময়ের প্রকাশ। এটি মূলত `Year-Month-Day Hour:Minute:Second` ফরম্যাটে থাকে।

* **ব্যবহার:** কোনো ট্রানজ্যাকশন কখন হয়েছে বা কোনো ইভেন্টের নির্দিষ্ট সময় রেকর্ড করতে এটি ব্যবহৃত হয়।
* **Pandas Command:** স্ট্রিং থেকে তারিখ বানাতে `pd.to_datetime()` ব্যবহার করা হয়।

**Datetime-এর প্রধান কাজসমূহ:**

* তারিখ থেকে দিন (`.dt.day`), মাস (`.dt.month`), বছর (`.dt.year`) আলাদা করা।
* সপ্তাহের নাম (`.dt.day_name()`) বা মাসের নাম বের করা।
* দিনের নাম বা ছুটির দিন কি না তা চেক করা।

---

### ২. Timedelta কী? (What is Timedelta?)

**Timedelta** হলো সময়ের একটি ব্যবধান বা ডিউরেশন (Duration)। এটি কোনো নির্দিষ্ট তারিখ নয়, বরং দুটি তারিখের মধ্যবর্তী সময়ের পার্থক্য।

* **ব্যবহার:** বয়স হিসাব করা, একটি কাজ শেষ হতে কত সময় লাগলো তা বের করা, অথবা আগামী ১০ দিন পর কী তারিখ হবে তা বের করা।
* **Pandas Command:** `pd.to_timedelta()` অথবা দুটি **Datetime** বিয়োগ করে এটি পাওয়া যায়।

**Timedelta-এর প্রধান কাজসমূহ:**

* তারিখের সাথে সময় যোগ বা বিয়োগ করা (যেমন: আজকের তারিখ + ৫ দিন)।
* মোট কত সেকেন্ড বা কত ঘণ্টা পার হয়েছে তা বের করা।

### ৩. মূল পার্থক্য একনজরে (Key Differences)

| বৈশিষ্ট্য | Datetime | Timedelta |
| --- | --- | --- |
| **মূল ধারণা** | ক্যালেন্ডারের একটি নির্দিষ্ট দিন/সময়। | সময়ের একটি পরিমাপ বা গ্যাপ। |
| **উদাহরণ** | ২০ মে, ২০২৪। | ৫ দিন ১০ ঘণ্টা। |
| **গাণিতিক কাজ** | `Date1 - Date2 = Timedelta` | `Date1 + Timedelta = New Date` |

**প্রো-টিপ:** যখনই কোনো CSV ফাইল রিড করবেন, তখন `parse_dates=['column_name']` প্যারামিটারটি ব্যবহার করলে Pandas অটোমেটিকালি ওই কলামকে **Datetime**-এ রূপান্তর করে নেয়।

# Solution of the LEETCODE problem: 196. Delete Duplicate Emails
Url: https://leetcode.com/problems/delete-duplicate-emails/description/

```python
# We have a given list of emails.

# email_data = [
#     [1, 'john@example.com'],
#     [2, 'bob@example.com'],
#     [3, 'john@example.com']
# ]


email_data = [
    [1, 'abc@efg.com'],
    [1, 'abc@efg.com'],
]



print(email_data)
print(type(email_data))  # The output is: list
```

```python
# Now we'll convert the list into a DataFrame. And I want to remove the index column.

person_df = pd.DataFrame(email_data, columns=['id', 'email']).astype({'id':'int64', 'email':'object'})
person_df
```

```python
def delete_duplicate_emails(person: pd.DataFrame) -> None:

    # First, I want to sort the rows according to the 'id' column.
    person.sort_values(by='id', ascending=True, inplace=True)

    # Then, I want to remove the duplicate emails and keep only the first occurrence of each email.
    person.drop_duplicates(subset=['email'], keep='first', inplace=True)

    # Now, I want to print the DataFrame.
    print(person)


delete_duplicate_emails(person_df)
```

# Solution of the LEETCODE problem: 1907. Count Salary Categories
Url: https://leetcode.com/problems/count-salary-categories/description/

```python
# We have a given list of account IDs and their corresponding incomes.

account_data = [
    [3, 108939],
    [2, 12747],
    [8, 87709],
    [6, 91796]
]

print(account_data)
print(type(account_data))
```

```python
# Now we'll convert the list into a DataFrame. And I want to remove the index column.

accounts_df = pd.DataFrame(account_data, columns=['account_id', 'income']).astype({'account_id': 'Int64', 'income': 'Int64'})
accounts_df
```

```python
def count_salary_categories(accounts: pd.DataFrame) -> pd.DataFrame:
    # First, I will count the income under each category.

    # high_salary_count = (accounts['income'] > 50000).sum()
    # average_salary_count = ((accounts['income'] >= 20000) & (accounts['income'] <= 50000)).sum()
    # low_salary_count = (accounts['income'] < 20000).sum()
    # OR
    high_salary_count = accounts['income'].gt(50000).sum()
    average_salary_count = accounts['income'].between(20000, 50000, inclusive="both").sum()
    low_salary_count = accounts['income'].lt(20000).sum()

    print(high_salary_count)
    print(average_salary_count)
    print(low_salary_count)

    # Now with these data, I will create a new DataFrame, with columns 'category' and 'accounts_count'.
    new_accounts_df = pd.DataFrame(
        {
            'category': [
                'Low Salary',
                'Average Salary',
                'High Salary'

            ],
            'accounts_count': [
                low_salary_count,
                average_salary_count,
                high_salary_count
            ]
        }
    )

    # Now will return the new DataFrame.
    return new_accounts_df


count_salary_categories(accounts_df)
```

# Solution of the CODECHEF problem: Filling Missing Data in Pandas
Url: https://www.codechef.com/learn/course/pandas/LPDP06/problems/LPDPR29

```python
import pandas as pd
import numpy as np

# Here is the given DataFrame
df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [5, np.nan, np.nan, None, 10],
    'C': ['a', 'b', 'c', None, 'e']
})


# df
# df.dtypes
# print(type(df))
print(df)


# Replace null values based on given conditions
df_filled1 = df.fillna({'A': 0,'B': 99,'C': 'unknown',})
print(df_filled1)


# Forward fill column B
df_filled2 = df.copy()
df_filled2['B'] = df_filled2['B'].ffill()
print(df_filled2)


# Fills missing values in numeric columns with the mean of the column and in string columns with 'Unknown'.
df_filled3 = df.copy()

for column in df_filled3.columns:
    if df_filled3[column].dtype == 'float64':
        df_filled3[column] = df_filled3[column].fillna(df_filled3[column].mean())
    else:
        df_filled3[column] = df_filled3[column].fillna('Unknown')

print(df_filled3)
```

# Solution of the CODECHEF problem: Dropping missing data
Url: https://www.codechef.com/learn/course/pandas/LPDP06/problems/LPDPR28

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [5, np.nan, np.nan, None, 10],
    'C': ['a', 'b', 'c', None, 'e']
})


print(df)


# Drop rows where column A has missing values
df_dropped1 = df.copy()
df_dropped1.dropna(subset=['A'], inplace=True)
print(df_dropped1)


# Drop rows with less than 2 non-null values
df_dropped2 = df.copy()
df_dropped2.dropna(thresh=2, inplace=True)
print(df_dropped2)


# Drop rows where both Column A and Column B has missing values
df_dropped3 = df.copy()
df_dropped3.dropna(subset=['A', 'B'], how="all", inplace=True)
print(df_dropped3)
```

# Solution of the CODECHEF problem: GroupBy Operation
Url: https://www.codechef.com/learn/course/pandas/LPD10/problems/LPDPR46

```python
import pandas as pd
import numpy as np

# Create a sample DataFrame
df = pd.DataFrame({
    'Category': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Value1': [10, 20, 30, 40, 50, 60],
    'Value2': [100, 200, 300, 400, 500, 600]
})

print(df)
```

```python
# Basic groupby and aggregation
print("Mean of Value1 for each Category:")
print(df.groupby('Category')['Value1'].mean())
```

```python
print("Max of Value2 for each Category:")
print(df.groupby('Category')['Value1'].max())
```

```python
print("Sum of Value1 for each Category:")
print(df.groupby('Category')['Value1'].sum())
```
