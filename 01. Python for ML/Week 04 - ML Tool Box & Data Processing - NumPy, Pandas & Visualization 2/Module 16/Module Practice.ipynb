# 16.1 Lineplot using Seaborn

**Line Plot** হলো একটি অন্যতম সাধারণ চার্ট যা ডেটা পয়েন্টগুলোকে সরলরেখা বা **Line** দিয়ে যুক্ত করে প্রদর্শিত করে। এটি মূলত সময়ের সাথে কোনো বিষয়ের পরিবর্তন বা **Trend** বোঝার জন্য সবচেয়ে কার্যকর।
**Seaborn** দিয়ে **Line Plot** তৈরি করা অনেক বেশি সহজ এবং এটি ডিফল্টভাবেই দেখতে বেশ প্রফেশনাল। **Seaborn** মূলত **Pandas DataFrame**-এর সাথে কাজ করার জন্য অপ্টিমাইজড।

### **Seaborn Line Plot** এর মূল বৈশিষ্ট্যসমূহ:

* **Functionality:** এটি সময়ের সাথে ডেটার পরিবর্তন বা **Trend** দেখানোর জন্য ব্যবহৃত হয়।
* **Primary Function:** এটি তৈরি করতে `sns.lineplot()` ফাংশন ব্যবহার করা হয়।
* **Semantic Mapping:** **Seaborn**-এ কলামের নাম ব্যবহার করে খুব সহজে রঙ (**Hue**) বা স্টাইল পরিবর্তন করা যায়।

**মনে রাখবেন:** **Seaborn**-এর গ্রাফ কাস্টমাইজ করার পর সেটি দেখানোর জন্য বা টাইটেল দেওয়ার জন্য আপনাকে এখনো **Matplotlib**-এর `plt.show()` বা `plt.title()` ব্যবহার করতে হবে।

````python
# Now we'll get the dataset from the directory and will convert it into a DataFrame.

students_df = pd.read_csv('sns_data.csv')
students_df
````

````python
# Now this time we'll create the Lineplot using Seaborn.

sns.lineplot(students_df, x='week', y='attendance_rate')
````

````python
# Now we'll try more parameters to improve the visualization of the Lineplot using Seaborn.

sns.lineplot(students_df, x='week', y='attendance_rate', errorbar=None, color='red', marker='o', linewidth=1, markersize=6, alpha=1)
````

````python
# Now we'll try 'hue' and 'palette' parameters to improve the visualization of the Lineplot using Seaborn.

sns.lineplot(students_df, x='week', y='attendance_rate', hue='gender', errorbar=None, palette=['red', 'blue'], marker='o', linewidth=1, markersize=6, alpha=1)
````

**Seaborn**-এ `hue` প্যারামিটারটি অত্যন্ত শক্তিশালী একটি টুল। এটি মূলত আপনার ডেটাকে একটি নির্দিষ্ট ক্যাটাগরির ওপর ভিত্তি করে আলাদা করে দেখায়।

এখানে `hue='gender'` যা করছে:
* **Data Grouping:** এটি আপনার পুরো ডেটাসেটকে `gender` কলামের মান অনুযায়ী (যেমন: Male এবং Female) দুই বা ততোধিক গ্রুপে ভাগ করে ফেলে।
* **Color Differentiation:** প্রতিটি গ্রুপের জন্য আলাদা আলাদা রঙের লাইন তৈরি করে। অর্থাৎ, আপনি একটি গ্রাফের ভেতরেই ছেলেদের এবং মেয়েদের `attendance_rate` আলাদা রঙের লাইনে দেখতে পাবেন।
* **Automatic Legend:** `hue` ব্যবহার করলে **Seaborn** অটোমেটিক একটি **Legend** (নির্দেশিকা) তৈরি করে দেয়, যা দেখে বোঝা যায় কোন রঙের লাইনটি কোন জেন্ডারকে নির্দেশ করছে।

# Axes-level Functions & Figure-level Functions

**Seaborn**-এ গ্রাফ তৈরি করার সময় **Axes-level** এবং **Figure-level** ফাংশনের পার্থক্য বোঝা খুবই জরুরি। এর ওপর ভিত্তি করেই আপনার গ্রাফের কন্ট্রোল এবং লেআউট নির্ভর করে।

### **১. Axes-level Functions**

এই ফাংশনগুলো সরাসরি একটি নির্দিষ্ট **Axes** (অর্থাৎ একটি একক গ্রাফ উইন্ডো) এর ওপর কাজ করে। এগুলো অনেকটা **Matplotlib**-এর মতো আচরণ করে।

* **উদাহরণ:** `sns.lineplot()`, `sns.scatterplot()`, `sns.barplot()`।
* **কখন কাজে লাগে:** যখন আপনি একটি বড় ফিগারের ভেতর ছোট ছোট অনেকগুলো সাবপ্লট (**Subplots**) নিজের মতো করে সাজাতে চান।
* **সুবিধা:** আপনি আগে থেকে `plt.subplots()` তৈরি করে তার ভেতর নির্দিষ্ট জায়গায় এই গ্রাফগুলো বসাতে পারেন।

### **২. Figure-level Functions**

এই ফাংশনগুলো পুরো **Figure**-কে নিয়ন্ত্রণ করে। এটি নিজে থেকেই একটি নতুন উইন্ডো বা লেআউট তৈরি করে নেয় এবং সাধারণত একসাথে একাধিক সাবপ্লট তৈরি করার জন্য ডিজাইন করা।

* **উদাহরণ:** `sns.relplot()`, `sns.catplot()`, `sns.displot()`।
* **কখন কাজে লাগে:** যখন আপনার ডেটাকে বিভিন্ন ক্যাটাগরি অনুযায়ী আলাদা আলাদা গ্রাফে (**Faceting**) ভাগ করে দেখাতে হয় (যেমন: `col='gender'` ব্যবহার করে ছেলে ও মেয়েদের জন্য আলাদা গ্রাফ)।
* **সুবিধা:** এটি অটোমেটিক লেজেন্ড এবং সাবপ্লটগুলোর স্পেসিং ঠিক করে দেয়।

### **পার্থক্য একনজরে:**

| বিষয় | Axes-level | Figure-level |
| --- | --- | --- |
| **মেইন ফাংশন** | `sns.lineplot()`, `sns.histplot()` | `sns.relplot()`, `sns.displot()` |
| **রিটার্ন টাইপ** | `matplotlib.axes.Axes` | `FacetGrid` |
| **সাবপ্লট (Faceting)** | ম্যানুয়ালি করতে হয়। | `col` বা `row` প্যারামিটার দিয়ে সহজে করা যায়। |
| **আকার পরিবর্তন** | `plt.figure(figsize=...)` দিয়ে। | `height` এবং `aspect` প্যারামিটার দিয়ে। |

**সহজ টিপস:** যদি আপনার একটি সিম্পল গ্রাফ লাগে, তবে **Axes-level** (`lineplot`) ব্যবহার করুন। আর যদি ডাটাকে কয়েক ভাগে ভাগ করে (যেমন: ৩টি আলাদা দেশ বা বছরের জন্য ৩টি গ্রাফ) দেখাতে চান, তবে **Figure-level** (`relplot`) ব্যবহার করুন।

````python
# Now we'll do the same plotting again, but this time using the Figure-level function.

sns.relplot(kind="line", data=students_df, x='week', y='attendance_rate', hue='gender', errorbar=None, palette=['red', 'blue'], marker='o', linewidth=1, markersize=6, alpha=1)
````

````python
# Now we'll create a Line Plot, using Figure-level function, comparing the columns 'week' and 'test_score'.

sns.relplot(kind="line", data=students_df, x='week', y='test_score', hue='gender', errorbar=None, palette=['red', 'blue'], marker='o', linewidth=1, markersize=6, alpha=1)
````

````python
# Now we'll create the same Line Plot again, using Figure-level function, comparing the columns 'week' and 'test_score', with grouping by the column 'class_level'.
# Let's remove the 'palette' parameter and let the Seaborn choose the colors automatically.

sns.relplot(kind="line", data=students_df, x='week', y='test_score', hue='class_level', errorbar=None, marker='o', linewidth=1, markersize=6, alpha=1)
````

````python
# Now we'll create the same Line Plot again, using Figure-level function, comparing the columns 'week' and 'attendance_rate', with grouping by the column 'class_level'.

sns.relplot(kind="line", data=students_df, x='week', y='attendance_rate', hue='class_level', errorbar=None, marker='o', linewidth=1, markersize=6, alpha=1)
````

# 16.2 Scatterplot using Seaborn

````python
# Let's' see the DataFrame once again.

students_df.head()
````

````python
# Now we'll create Scatter Plot using Seaborn.

sns.scatterplot(data=students_df, x='study_hours', y='test_score')
````

**Seaborn**-এ `style` প্যারামিটারটি `hue`-এর মতোই ডেটাকে গ্রুপ করতে ব্যবহৃত হয়, তবে এটি রঙের বদলে লাইনের **আকৃতি বা ধরন** (যেমন: ড্যাশড লাইন বনাম সলিড লাইন) পরিবর্তন করে।

### **Style Parameter** কী?

* **Functionality:** এটি একটি ক্যাটাগরি অনুযায়ী লাইনের ভিজ্যুয়াল স্টাইল (যেমন: `solid`, `dashed`, `dotted`) আলাদা করে দেয়।
* **Visual Aid:** অনেক সময় প্রিন্ট কপি বা কালার-ব্লাইন্ড মানুষের জন্য শুধু রঙ (`hue`) দিয়ে পার্থক্য বোঝা কঠিন হয়। তখন `style` ব্যবহার করলে লাইন দেখে সহজেই গ্রুপগুলো চেনা যায়।

### **Hue** এবং **Style**-এর মধ্যে পার্থক্য:

* **`hue`**: আলাদা রঙের লাইন তৈরি করে।
* **`style`**: আলাদা ধরনের (Dashed/Solid) লাইন তৈরি করে।

আপনি চাইলে এই দুটি একসাথেও ব্যবহার করতে পারেন যাতে লাইনগুলো রঙ এবং ধরন—উভয় দিক থেকেই আলাদা হয়।

### **Style**-এর প্রধান সুবিধা:

১. **Clarity:** যখন গ্রাফে অনেকগুলো লাইন থাকে, তখন শুধু রঙ দিয়ে সব গুলিয়ে যেতে পারে। `style` যোগ করলে গ্রাফটি আরও পরিষ্কার হয়।
২. **Black & White Compatibility:** যদি আপনার গ্রাফটি সাদা-কালো প্রিন্ট করা হয়, তবে শুধু `style` দেখেই আপনি ডেটা আলাদা করতে পারবেন।

````python
# Now we'll create the same Scatter Plot again using Seaborn, but this time we'll add more parameters to improve the visualization of the Scatter Plot.

sns.scatterplot(data=students_df,
                x='study_hours',
                y='test_score',
                hue='gender',
                style='class_level',
                alpha=0.5)
````

````python
# Now we'll create the same Scatter Plot again using Seaborn, but this time we'll add 'subject' in the stype parameter.

sns.scatterplot(data=students_df,
                x='study_hours',
                y='test_score',
                hue='gender',
                style='subject',
                alpha=0.5)
````

**Seaborn**-এ `size` প্যারামিটারটি ব্যবহার করা হয় কোনো একটি ক্যাটাগরি বা ভেরিয়েবলের ওপর ভিত্তি করে লাইনের **পুরুত্ব (Thickness)** বা মার্কারের আকার পরিবর্তন করার জন্য।

### **Size Parameter** কী?

* **Functionality:** এটি ডেটার কোনো একটি নির্দিষ্ট কলামের মানের ওপর ভিত্তি করে লাইনের চওড়া বা সরু ভাব নিয়ন্ত্রণ করে।
* **Visual Weight:** সাধারণত সংখ্যাবাচক (**Numerical**) ডেটার ক্ষেত্রে এটি বেশি ব্যবহৃত হয়। যেমন: যে বছরের সেলস বেশি, সেই বছরের লাইনটি বেশি মোটা দেখাবে।

### **Hue**, **Style**, এবং **Size** এর মধ্যে পার্থক্য:

* **`hue`**: রঙ পরিবর্তন করে (Color)।
* **`style`**: লাইনের ধরন পরিবর্তন করে (Dashes)।
* **`size`**: লাইনের গুরুত্ব বা পুরুত্ব পরিবর্তন করে (Thickness)।

### কেন এটি ব্যবহার করবেন?

১. **Emphasis:** গুরুত্বপূর্ণ ডেটা পয়েন্ট বা ট্রেন্ডকে আলাদাভাবে ফোকাস করার জন্য এটি দারুণ কার্যকর।
২. **Extra Dimension:** একই গ্রাফে `x`, `y` এবং `hue` ছাড়াও যদি চতুর্থ কোনো তথ্য (যেমন: বাজেটের পরিমাণ বা জনসংখ্যার আকার) দেখাতে চান, তবে `size` ব্যবহার করা হয়।

**টিপস:** যদি আপনি চান লাইনগুলো খুব বেশি মোটা না হোক, তবে `sizes` (বহুবচন) প্যারামিটার ব্যবহার করে একটি নির্দিষ্ট রেঞ্জ (যেমন: `sizes=(1, 5)`) সেট করে দিতে পারেন।

````python
# Now we'll create the same Scatter Plot again using Seaborn, but this time we'll add 'Tshirt_size' in the size parameter.

sns.scatterplot(data=students_df,
                x='study_hours',
                y='test_score',
                hue='gender',
                style='subject',
                size='Tshirt_size',
                alpha=0.5)
````

````python
# Now we'll create the same Line Plot again, using the Figure-level function.

sns.relplot(data=students_df,
                x='study_hours',
                y='test_score',
                hue='gender',
                style='subject',
                size='Tshirt_size',
                alpha=0.5)
````

### **Facet Plot** এর মূল বৈশিষ্ট্যসমূহ:

* **Functionality:** এটি একই গ্রাফে অনেক ডেটা ঠাসাঠাসি না করে আলাদা আলাদা গ্রিডে বা বক্সে ডেটা প্রদর্শন করে।

| প্যারামিটার | কাজ |
| --- | --- |
| **`col`** | কলাম অনুযায়ী গ্রাফগুলোকে ভাগ করতে (পাশাপাশি)। |
| **`row`** | রো অনুযায়ী গ্রাফগুলোকে ভাগ করতে (ওপর-নিচ)। |

| **`kind`** | গ্রাফের ধরন ঠিক করতে (যেমন: `'line'` বা `'scatter'`)। |

### কেন এটি ব্যবহার করবেন?

১. **Clarity:** যখন আপনার গ্রাফে অনেক বেশি ক্যাটাগরি থাকে, তখন সব এক জায়গায় না দেখিয়ে আলাদা বক্সে দেখালে তা অনেক বেশি পরিষ্কার হয়।
২. **Patterns:** বিভিন্ন সাবপ্লটের মধ্যে প্যাটার্ন বা ট্রেন্ডের পার্থক্য খুব সহজে চোখে পড়ে।

# KDE Plot

**KDE Plot** (Kernel Density Estimate) হলো এমন একটি গ্রাফ যা কোনো ডেটাসেটের ডিস্ট্রিবিউশন বা বিস্তৃতিকে একটি মসৃণ লাইনের মাধ্যমে প্রকাশ করে। এটি মূলত **Histogram**-এর একটি উন্নত এবং কন্টিনিউয়াস (**Continuous**) সংস্করণ।

### **KDE Plot** এর মূল বৈশিষ্ট্যসমূহ:

* **Functionality:** এটি ডেটার ঘনত্ব (**Density**) নির্দেশ করে। গ্রাফের লাইনটি যেখানে বেশি উঁচু, সেখানে ডেটার পরিমাণ বেশি।
* **Primary Function:** **Seaborn**-এ এটি তৈরি করতে `sns.kdeplot()` ফাংশন ব্যবহার করা হয়।
* **Smoothness:** এটি **Histogram**-এর মতো চারকোনা বক্স বা **Bins** ব্যবহার না করে একটি মসৃণ কার্ভ তৈরি করে, যা ডেটার ট্রেন্ড বুঝতে সহজ করে।

### কাস্টমাইজ করার প্যারামিটারসমূহ:

| প্যারামিটার | কাজ |
| --- | --- |
| **`fill`** | লাইনের নিচের অংশটি রঙ দিয়ে ভরাট করার জন্য (আগে এটি `shade` ছিল)। |
| **`hue`** | নির্দিষ্ট ক্যাটাগরি অনুযায়ী আলাদা আলাদা রঙের কার্ভ তৈরি করতে। |
| **`bw_adjust`** | লাইনের মসৃণতা (**Smoothness**) নিয়ন্ত্রণ করতে। |
| **`multiple`** | একাধিক ক্যাটাগরির ডেটা কীভাবে দেখাবে তা ঠিক করতে (যেমন: `'stack'` বা `'fill'`)। |

### কেন এটি ব্যবহার করবেন?

* **Outlier Detection:** ডেটার মধ্যে কোনো অস্বাভাবিক মান বা **Outlier** আছে কি না তা সহজেই বোঝা যায়।
* **Overlap Comparison:** `hue` ব্যবহার করে একই গ্রাফে একাধিক গ্রুপের ডিস্ট্রিবিউশন তুলনা করা খুব সহজ হয়।

````python
# Now we'll see the DataFrame again.

students_complete_df.head()
````

````python
# Now we'll create a KDE plot, from the DataFrame, where x will represent 'attendance_rate' and x will represent the student count.

sns.kdeplot(data=students_complete_df, x='attendance_rate')
````

# 16.6 Count Plot and Bar Plot using Seaborn

**Count Plot** এবং **Bar Plot** হলো **Seaborn**-এর দুটি অত্যন্ত প্রয়োজনীয় ক্যাটাগরিকাল গ্রাফ। দেখতে অনেকটা একরকম হলেও এদের কাজের উদ্দেশ্য ভিন্ন।

### **১. Count Plot**

এটি মূলত একটি কলামে প্রতিটি ক্যাটাগরি কতবার আছে তা গুনে দেখায়।

* **Functionality:** এটি শুধু একটি ভেরিয়েবল বা কলামের ফ্রিকোয়েন্সি (**Frequency**) কাউন্ট করে।
* **Primary Function:** `sns.countplot()`।
* **Key Point:** এখানে আপনাকে আলাদা করে **y-axis** দিতে হয় না, কারণ এটি নিজে থেকেই কাউন্ট করে নেয়।

### **২. Bar Plot**

এটি একটি ক্যাটাগরিকাল ভেরিয়েবলের সাথে একটি সংখ্যাবাচক (**Numerical**) ভেরিয়েবলের সম্পর্ক দেখায়।

* **Functionality:** এটি সাধারণত প্রতিটি ক্যাটাগরির গড় মান (**Mean**) প্রদর্শন করে।
* **Primary Function:** `sns.barplot()`।
* **Key Point:** এতে **x** (Category) এবং **y** (Numeric value) —উভয় প্যারামিটারই দিতে হয়।

### **পার্থক্য একনজরে:**

| ফিচার | Count Plot | Bar Plot |
| --- | --- | --- |
| **মূল কাজ** | ডেটা গণনা করা (How many?)। | গড় বা অন্য কোনো এস্টিমেশন দেখানো। |
| **প্যারামিটার** | শুধু **x** অথবা **y** লাগে। | **x** এবং **y** দুটোই লাগে। |
| **Error Bar** | থাকে না। | ডিফল্টভাবে থাকে (Confidence Interval)। |

### কাস্টমাইজ করার প্যারামিটারসমূহ:

* **`hue`**: আলাদা ক্যাটাগরি অনুযায়ী রঙ ভাগ করতে।
* **`palette`**: গ্রাফের কালার স্কিম পরিবর্তন করতে (যেমন: `'viridis'`, `'magma'`)।
* **`order`**: বারগুলো কোন ক্রমে থাকবে তা নির্দিষ্ট করতে।

# 16.7 Reg plot using seaborn


**Reg Plot** (Regression Plot) হলো **Seaborn**-এর এমন একটি গ্রাফ যা দুটি সংখ্যাবাচক (**Numerical**) ভেরিয়েবলের মধ্যে সম্পর্ক দেখানোর পাশাপাশি একটি **Regression Line** বা সরলরেখা টেনে দেয়। এটি মূলত ডেটার ট্রেন্ড বা প্রবণতা বোঝার জন্য ব্যবহৃত হয়।

### **Reg Plot** এর মূল বৈশিষ্ট্যসমূহ:

* **Functionality:** এটি একটি **Scatter Plot** তৈরি করে এবং তার ওপর দিয়ে একটি **Linear Regression** লাইন এঁকে দেয়।
* **Primary Function:** এটি তৈরি করতে `sns.regplot()` ফাংশন ব্যবহার করা হয়।
* **Correlation:** এই গ্রাফটি দেখে সহজে বোঝা যায় একটি ভেরিয়েবল বাড়লে অন্যটি বাড়ছে (**Positive**) নাকি কমছে (**Negative**)।

### কাস্টমাইজ করার প্যারামিটারসমূহ:

| প্যারামিটার | কাজ |
| --- | --- |
| **`x`, `y**` | অক্ষে কোন কলামের মান থাকবে তা নির্দিষ্ট করে। |
| **`data`** | আপনার **Pandas DataFrame**-টির নাম। |
| **`marker`** | ডেটা পয়েন্টের স্টাইল পরিবর্তন করতে (যেমন: `'x'`, `'+'`)। |
| **`scatter_kws`** | স্ক্যাটার পয়েন্টের রঙ বা আকার পরিবর্তন করতে (ডিকশনারি ফরম্যাটে)। |
| **`line_kws`** | রিগ্রেশন লাইনের রঙ বা পুরুত্ব পরিবর্তন করতে। |

### কেন এটি ব্যবহার করবেন?

১. **Trend Analysis:** ডেটার ডিস্ট্রিবিউশন দেখে ভবিষ্যৎবাণী বা প্রেডিকশন করার প্রাথমিক ধারণা পাওয়া যায়।
২. **Confidence Interval:** লাইনের চারপাশে একটি হালকা ছায়া দেখা যায়, যা ডেটার অনিশ্চয়তা বা **Confidence Interval** নির্দেশ করে।

# 16.8 Pair plot and Joint plot using Seaborn

**Seaborn**-এর **Joint Plot** এবং **Pair Plot** হলো মাল্টিভ্যারিয়েট ডেটা অ্যানালাইসিসের জন্য অত্যন্ত শক্তিশালী দুটি টুল। এগুলো মূলত একাধিক ভেরিয়েবলের মধ্যে সম্পর্ক এবং তাদের নিজস্ব ডিস্ট্রিবিউশন একসাথে দেখার জন্য ব্যবহৃত হয়।

### **১. Joint Plot**

এটি মূলত দুটি নির্দিষ্ট ভেরিয়েবলের মধ্যে দ্বিমুখী সম্পর্ক দেখায়।
* **Functionality:** মাঝখানে একটি **Scatter Plot** বা **Reg Plot** থাকে এবং অক্ষগুলোর পাশে প্রতিটি ভেরিয়েবলের আলাদা **Histogram** বা **KDE Plot** থাকে।
* **Primary Function:** `sns.jointplot()`।
* **Kind:** `kind` প্যারামিটার ব্যবহার করে মাঝখানের গ্রাফটি পরিবর্তন করা যায় (যেমন: `'scatter'`, `'reg'`, `'resid'`, `'kde'`, `'hex'`)।

### **২. Pair Plot**

এটি একটি পুরো **DataFrame**-এর সমস্ত সংখ্যাবাচক (**Numerical**) ভেরিয়েবলের মধ্যে জোড়ায় জোড়ায় সম্পর্ক দেখায়।
* **Functionality:** এটি একটি গ্রিড তৈরি করে যেখানে প্রতিটি ভেরিয়েবলের সাথে অন্য সব ভেরিয়েবলের সম্পর্ক দেখা যায়। ডায়াগোনাল বা কোনাকুনি ঘরে প্রতিটি ভেরিয়েবলের নিজস্ব **Histogram** বা **KDE** থাকে।
* **Primary Function:** `sns.pairplot()`।
* **Key Point:** ডেটাসেটে অনেক বেশি কলাম থাকলে এটি অনেক বড় হয়ে যায়, তাই নির্দিষ্ট কিছু কলাম নিয়ে এটি করা ভালো।

### **পার্থক্য একনজরে:**

| ফিচার | Joint Plot | Pair Plot |
| --- | --- | --- |
| **ভেরিয়েবল সংখ্যা** | শুধুমাত্র ২টির মধ্যে সম্পর্ক। | পুরো ডেটাসেটের সবগুলোর মধ্যে সম্পর্ক। |
| **ব্যবহার** | নির্দিষ্ট দুটি ভেরিয়েবলের গভীর বিশ্লেষণ। | পুরো ডেটার ওপর একটি দ্রুত ওভারভিউ পেতে। |
| **লেআউট** | ১টি মেইন গ্রাফ + ২লি মার্জিনাল গ্রাফ। | একটি বিশাল গ্রিড (Matrix)। |

**Plotly** হলো একটি আধুনিক এবং শক্তিশালী **Python Library**, যা মূলত **Interactive Visualization** তৈরির জন্য বিখ্যাত। এটি দিয়ে এমন সব গ্রাফ তৈরি করা যায় যা ইউজাররা জুম করতে পারে, মাউস দিয়ে ডেটা পয়েন্ট চেক করতে পারে এবং ফিল্টার করতে পারে।

### **Plotly** কী?

* এটি একটি **Open-source Library** যা ওয়েব-বেসড এবং ইন্টারঅ্যাক্টিভ চার্ট তৈরি করে।
* এটি মূলত **JavaScript (Plotly.js)**-এর ওপর ভিত্তি করে তৈরি, কিন্তু **Python** দিয়ে খুব সহজে এটি নিয়ন্ত্রণ করা যায়।

### কেন এবং কী কী কাজে ব্যবহৃত হয়?

* **Interactivity:** এর প্রতিটি গ্রাফে মাউস নিয়ে গেলে নির্দিষ্ট ডেটা দেখা যায় (**Hover**), জুম করা যায় এবং ইমেজ হিসেবে সেভ করা যায়।
* **Plotly Express:** এটি একটি উচ্চস্তরের (**High-level**) মডিউল যা দিয়ে মাত্র এক লাইনের কোডে জটিল সব গ্রাফ তৈরি করা যায়।
* **Dashboards:** এটি **Dash** ফ্রেমওয়ার্কের সাথে ব্যবহার করে প্রফেশনাল লেভেলের ডেটা ড্যাশবোর্ড তৈরি করতে সাহায্য করে।
* **3D Plots:** এতে **3D Scatter Plot** বা **Terrain Map**-এর মতো উন্নত ভিজ্যুয়ালাইজেশন খুব চমৎকারভাবে করা যায়।
